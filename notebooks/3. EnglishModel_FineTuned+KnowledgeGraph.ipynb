```python
import ast

def fast_parse_list(x):
    """Parse string/list values, and split by newline if needed."""
    if isinstance(x, str):
        x_strip = x.strip()
        # Try parse if it looks like a Python list
        if x_strip.startswith("[") and x_strip.endswith("]"):
            try:
                parsed = ast.literal_eval(x_strip)
                return fast_parse_list(parsed)  # handle nested
            except Exception:
                return [s.strip() for s in x_strip.split("\n") if s.strip()]
        # Otherwise, split by newline directly
        return [s.strip() for s in x_strip.split("\n") if s.strip()]

    elif isinstance(x, list):
        # flatten, split by newline if needed
        out = []
        for i in x:
            if isinstance(i, str):
                out.extend([s.strip() for s in i.split("\n") if s.strip()])
        return out

    else:
        return [str(x)]

def process_data2(df):
    df = df.copy()
    df["aspect"] = df["aspect"].apply(fast_parse_list)
    df["paraphrases"] = df["paraphrases"].apply(fast_parse_list)
    df.reset_index(drop=True, inplace=True)
    return df
```

```python
import re
from typing import List, Tuple, Dict, Iterable

# -----------------------------
# 1) Parse model string -> list
# -----------------------------
_SEP_PAT = re.compile(
    r"\s*(?:\[sep\]|\[SEP\]|ÿå|,|;|ÿõ|\||\n|\t|\/|\\|:|\s{2,})\s*", re.IGNORECASE
)

def parse_aspect_string(s) -> List[str]:
    """
    Robustly parse a model's string into a list of aspect terms.
    Handles Arabic/English commas, ; ÿõ | [sep], brackets, quotes, whitespace.
    If input is already a list, returns a cleaned version.
    """
    if s is None:
        return []
    if isinstance(s, list):
        items = s
    else:
        # strip surrounding brackets/quotes if present
        s = str(s).strip()
        s = s.strip("[](){}")
        s = s.replace("'", "").replace('"', "")
        # split on many possible separators
        items = _SEP_PAT.split(s)

    # clean tokens: remove empties/stray punctuation
    cleaned = []
    for t in items:
        t = t.strip()
        if not t:
            continue
        # remove leading/trailing punctuation
        t = re.sub(r"^[\s\-\‚Äì\‚Äî\¬∑\‚Ä¢\*\.]+|[\s\-\‚Äì\‚Äî\¬∑\‚Ä¢\*\.]+$", "", t)
        if t:
            cleaned.append(t)
    # de-duplicate while preserving order
    seen = set()
    dedup = []
    for w in cleaned:
        if w not in seen:
            seen.add(w)
            dedup.append(w)
    return dedup

# ----------------------------------------
# 2) Light Arabic normalization for eval
# ----------------------------------------
_DIACRITICS = re.compile(r'[\u0617-\u061A\u064B-\u0652]')
# common clitic prefixes and pronominal suffixes
_PREFIXES = re.compile(r'^(?:ÿßŸÑ|Ÿà|ŸÅ|ÿ®|ŸÑ|ŸÉ)+')
_SUFFIXES = re.compile(r'(?:ŸáŸÖÿß|ŸÉŸÖÿß|Ÿáÿß|ŸáŸÖ|ŸÜÿß|ŸÉŸÖ|ŸáŸÜ|ŸÉŸÜ|Ÿá|Ÿä|ÿ©|ÿßÿ™)$')

def normalize_token(tok: str) -> str:
    t = str(tok)
    # unify common letters
    t = t.replace("ÿ£", "ÿß").replace("ÿ•", "ÿß").replace("ÿ¢", "ÿß")
    t = t.replace("Ÿâ", "Ÿä").replace("ÿ©", "Ÿá").replace("ÿ§", "ÿ°").replace("ÿ¶", "ÿ°")
    # remove diacritics & tatweel
    t = _DIACRITICS.sub("", t)
    t = t.replace("ŸÄ", "")
    # trim clitics (light)
    t = _PREFIXES.sub("", t)
    t = _SUFFIXES.sub("", t)
    return t.strip()

def normalize_phrase(phrase: str) -> str:
    # token-wise normalization, preserve spaces between meaningful tokens
    toks = [w for w in re.split(r"\s+", phrase.strip()) if w]
    return " ".join(normalize_token(w) for w in toks)

# ----------------------------------------------------
# 3) Flexible matching: exact, substring, Jaccard(IoU)
# ----------------------------------------------------
def jaccard(a: str, b: str) -> float:
    A = set(a.split())
    B = set(b.split())
    if not A and not B:
        return 1.0
    if not A or not B:
        return 0.0
    return len(A & B) / len(A | B)

def similarity(gold: str, pred: str) -> float:
    """Return similarity score in [0,1] after normalization.
    1.0 if exact or substring; else token Jaccard."""
    g = normalize_phrase(gold)
    p = normalize_phrase(pred)
    if not g or not p:
        return 0.0
    if g == p:
        return 1.0
    if g in p or p in g:
        return 1.0
    return jaccard(g, p)

# ---------------------------------------------------
# 4) Greedy bipartite matching above a threshold
# ---------------------------------------------------
def best_alignment(gold: List[str], pred: List[str], thresh: float = 0.6) -> Tuple[int, List[Tuple[int,int,float]]]:
    """
    Greedy maximum-weight matching:
    - Build all pairs with sim >= thresh
    - Sort by similarity desc
    - Pick non-conflicting pairs
    Returns TP count and list of matches (gi, pj, sim)
    """
    pairs = []
    for i, g in enumerate(gold):
        for j, p in enumerate(pred):
            s = similarity(g, p)
            if s >= thresh:
                pairs.append((i, j, s))
    pairs.sort(key=lambda x: x[2], reverse=True)

    matched_g = set()
    matched_p = set()
    matches = []
    for gi, pj, s in pairs:
        if gi not in matched_g and pj not in matched_p:
            matched_g.add(gi)
            matched_p.add(pj)
            matches.append((gi, pj, s))
    return len(matches), matches

# ----------------------------------------------------------------
# 5) Precision/Recall/F1 for a single example and aggregated stats
# ----------------------------------------------------------------
# def prf1_for_example(gold_terms: Iterable[str], pred_terms: Iterable[str], thresh: float = 0.6) -> Dict[str, float]:
#     gold = [t for t in map(str, gold_terms) if t and t.strip()]
#     pred = [t for t in map(str, pred_terms) if t and t.strip()]

#     tp, _ = best_alignment(gold, pred, thresh=thresh)
#     fp = max(0, len(pred) - tp)
#     fn = max(0, len(gold) - tp)

#     prec = tp / (tp + fp) if (tp + fp) > 0 else 0.0
#     rec  = tp / (tp + fn) if (tp + fn) > 0 else 0.0
#     f1   = 2 * prec * rec / (prec + rec) if (prec + rec) > 0 else 0.0

#     return {"tp": tp, "fp": fp, "fn": fn, "precision": prec, "recall": rec, "f1": f1}

def prf1_for_example(gold_terms: Iterable[str], pred_terms: Iterable[str], thresh: float = 0.6) -> Dict[str, float]:
    gold = [t for t in map(str, gold_terms) if t and t.strip()]
    pred = [t for t in map(str, pred_terms) if t and t.strip()]

    # Special case: both empty ‚Üí perfect prediction
    if not gold and not pred:
        return {"tp": 0, "fp": 0, "fn": 0,
                "precision": 1.0, "recall": 1.0, "f1": 1.0}

    tp, _ = best_alignment(gold, pred, thresh=thresh)
    fp = max(0, len(pred) - tp)
    fn = max(0, len(gold) - tp)

    prec = tp / (tp + fp) if (tp + fp) > 0 else 0.0
    rec  = tp / (tp + fn) if (tp + fn) > 0 else 0.0
    f1   = 2 * prec * rec / (prec + rec) if (prec + rec) > 0 else 0.0

    return {"tp": tp, "fp": fp, "fn": fn,
            "precision": prec, "recall": rec, "f1": f1}


def evaluate_aspects(
    gold_lists: List[List[str]],
    pred_strings: List[str],
    parse_threshold: float = 0.6
) -> Dict[str, object]:
    """
    Evaluate lists of gold aspects vs model string predictions.
    - gold_lists: list of gold lists, e.g., [['ÿßŸÑŸÅŸÜÿØŸÇ','ŸÖŸàŸÇŸÅ'], ...]
    - pred_strings: list of raw model strings for each example
    - parse_threshold: similarity threshold for a match (default 0.6)
    Returns dict with micro/macro P/R/F1 and per-example scores.
    """
    assert len(gold_lists) == len(pred_strings), "gold and pred sizes must match"

    per_example = []
    tot_tp = tot_fp = tot_fn = 0

    for gold, pred_str in zip(gold_lists, pred_strings):
        pred_list = parse_aspect_string(pred_str)
        metrics = prf1_for_example(gold, pred_list, thresh=parse_threshold)
        per_example.append(metrics)
        tot_tp += metrics["tp"]
        tot_fp += metrics["fp"]
        tot_fn += metrics["fn"]

    # micro
    micro_p = tot_tp / (tot_tp + tot_fp) if (tot_tp + tot_fp) > 0 else 0.0
    micro_r = tot_tp / (tot_tp + tot_fn) if (tot_tp + tot_fn) > 0 else 0.0
    micro_f = 2 * micro_p * micro_r / (micro_p + micro_r) if (micro_p + micro_r) > 0 else 0.0

    # macro
    if per_example:
        macro_p = sum(x["precision"] for x in per_example) / len(per_example)
        macro_r = sum(x["recall"] for x in per_example) / len(per_example)
        macro_f = sum(x["f1"] for x in per_example) / len(per_example)
    else:
        macro_p = macro_r = macro_f = 0.0

    return {
        "micro": {"precision": micro_p, "recall": micro_r, "f1": micro_f, "tp": tot_tp, "fp": tot_fp, "fn": tot_fn},
        "macro": {"precision": macro_p, "recall": macro_r, "f1": macro_f},
        "per_example": per_example,
    }
```

```python
import json
from pathlib import Path
import networkx as nx
from typing import Dict, Any, List

class KnowledgeGraph:
    """
    Reusable Knowledge Graph wrapper (NetworkX).
    - Handles both structured aspects (dicts) and string lists.
    - Normalizes IDs but preserves original labels.
    - IMPORTANT: Ignores any JSON edges of the form aspect -> is_relevant_to_domain -> domain.
                 Only uses:
                   adjective -> describes_aspect -> aspect
                   aspect    -> belongs_to_entity -> entity
                   entity    -> is_relevant_to_domain -> domain
    """
    def __init__(self):
        self.G = nx.DiGraph()
        self.meta: Dict[str, Any] = {}
        # Quick indexes for convenience
        self._id2label: Dict[str, str] = {}
        self._label2id: Dict[str, str] = {}

    @staticmethod
    def _norm(s):
        return s.strip().lower() if isinstance(s, str) else s

    @staticmethod
    def _pred_str(d) -> str:
        return (d.get("predicate") or "").lower().replace(" ", "_")

    @classmethod
    def from_json(cls, path: str | Path) -> "KnowledgeGraph":
        kg = cls()
        kg._load(Path(path))
        return kg

    def _add_node(self, nid: str, label: str, ntype: str, **attrs):
        if nid not in self.G:
            self.G.add_node(nid, type=ntype, label=label, **attrs)
            if label and label not in self._label2id:
                self._label2id[label] = nid
            self._id2label[nid] = label
        else:
            # upgrade type if unknown
            if ntype and self.G.nodes[nid].get("type") == "unknown":
                self.G.nodes[nid]["type"] = ntype
            if "label" not in self.G.nodes[nid]:
                self.G.nodes[nid]["label"] = label

    def _load(self, path: Path):
        with open(path, "r", encoding="utf-8") as f:
            data = json.load(f)
        self.meta["source_path"] = str(path)

        # 1) Domains
        for d in data.get("domains", []):
            if isinstance(d, dict) and "name" in d:
                did = self._norm(d["name"])
                self._add_node(did, d["name"], "domain", **{k: v for k, v in d.items() if k != "name"})

        # 2) Entities (+ entity -> domain)
        for e in data.get("entities", []):
            if isinstance(e, dict) and "name" in e:
                eid = self._norm(e["name"])
                self._add_node(eid, e["name"], "entity", **{k: v for k, v in e.items() if k != "name"})
                dom = e.get("domain")
                if isinstance(dom, str) and dom.strip():
                    did = self._norm(dom)
                    if did not in self.G:
                        self._add_node(did, dom, "domain")
                    self.G.add_edge(
                        eid, did,
                        predicate="is_relevant_to_domain",
                        subject=e["name"], object=dom,
                        sub_lang=e.get("language"), obj_lang=e.get("language")
                    )

        # 3) Aspects (+ aspect -> entity)
        aspects = data.get("aspects", [])
        if aspects and isinstance(aspects[0], dict):
            for a in aspects:
                if not isinstance(a, dict) or "name" not in a:
                    continue
                aid = self._norm(a["name"])
                self._add_node(aid, a["name"], "aspect", **{k: v for k, v in a.items() if k != "name"})
                ents = a.get("entity", []) or []
                for ent in ents:
                    eid = self._norm(ent)
                    if eid not in self.G:
                        self._add_node(eid, ent, "entity")
                    self.G.add_edge(
                        aid, eid,
                        predicate="belongs_to_entity",
                        subject=a["name"], object=ent,
                        sub_lang=a.get("language"), obj_lang=a.get("language")
                    )
        else:
            # Aspects as strings
            for a in aspects:
                if isinstance(a, str):
                    aid = self._norm(a)
                    self._add_node(aid, a, "aspect")

        # 4) Adjectives
        for adj in data.get("adjectives", []):
            if isinstance(adj, str):
                aid = self._norm(adj)
                self._add_node(aid, adj, "adjective")

        # 5) Relationships (ignore aspect -> domain)
        for rel in data.get("relationships", []):
            subj = self._norm(rel.get("subject"))
            obj  = self._norm(rel.get("object"))
            pred = self._pred_str(rel)
            if not subj or not obj or not pred:
                continue

            # ensure nodes exist
            if subj not in self.G:
                self._add_node(subj, rel.get("subject", subj), "unknown")
            if obj not in self.G:
                self._add_node(obj, rel.get("object", obj), "unknown")

            # HARD RULE: JSON must NOT define aspect -> is_relevant_to_domain -> domain
            if pred == "is_relevant_to_domain" and self.G.nodes[subj].get("type") == "aspect":
                continue

            # add allowed edge as-is
            self.G.add_edge(subj, obj, **rel)

        # Summary
        self.meta["counts"] = self.counts()

    # ------------------- Queries -------------------
    def counts(self) -> dict:
        types = {}
        for _, d in self.G.nodes(data=True):
            t = d.get("type", "unknown")
            types[t] = types.get(t, 0) + 1
        return {
            "nodes": self.G.number_of_nodes(),
            "edges": self.G.number_of_edges(),
            **{f"n_{k}": v for k, v in types.items()},
        }

    def get_traces_for_adjective(self, adjective: str) -> List[dict]:
        """Return all adjective -> aspect -> entity -> domain traces."""
        adj_id = self._norm(adjective)
        if adj_id not in self.G:
            return []
        traces = []
        for _, aspect_id, d1 in self.G.out_edges(adj_id, data=True):
            if self._pred_str(d1) != "describes_aspect":
                continue
            for _, ent_id, d2 in self.G.out_edges(aspect_id, data=True):
                if self._pred_str(d2) != "belongs_to_entity":
                    continue
                for _, dom_id, d3 in self.G.out_edges(ent_id, data=True):
                    if self._pred_str(d3) != "is_relevant_to_domain":
                        continue
                    traces.append({
                        "adjective": self.G.nodes[adj_id].get("label", adj_id),
                        "aspect":    self.G.nodes[aspect_id].get("label", aspect_id),
                        "entity":    self.G.nodes[ent_id].get("label", ent_id),
                        "domain":    self.G.nodes[dom_id].get("label", dom_id),
                    })
        return traces

    def get_aspects_by_entity(self, entity: str) -> List[str]:
        """Aspects that belong to a given entity."""
        eid = self._norm(entity)
        if eid not in self.G:
            return []
        aspects = [
            u for u, v, d in self.G.in_edges(eid, data=True)
            if self._pred_str(d) == "belongs_to_entity"
        ]
        return sorted({self.G.nodes[a].get("label", a) for a in aspects})

    def get_entities_by_domain(self, domain: str) -> List[str]:
        """Entities that are relevant to a given domain."""
        did = self._norm(domain)
        if did not in self.G:
            return []
        ents = [
            u for u, v, d in self.G.in_edges(did, data=True)
            if self._pred_str(d) == "is_relevant_to_domain"
        ]
        return sorted({self.G.nodes[e].get("label", e) for e in ents})

    def get_adjectives_for_aspect(self, aspect: str) -> List[str]:
        """Adjectives that describe a given aspect."""
        aid = self._norm(aspect)
        if aid not in self.G:
            return []
        adjs = [
            u for u, v, d in self.G.in_edges(aid, data=True)
            if self._pred_str(d) == "describes_aspect"
        ]
        return sorted({self.G.nodes[x].get("label", x) for x in adjs})

    # ---------- NEW convenience helpers (names as requested) ----------
    def get_aspects(self, adjective: str) -> List[str]:
        """Aspects linked from an adjective via 'describes_aspect'."""
        adj_id = self._norm(adjective)
        if adj_id not in self.G:
            return []
        aspects = [
            v for _, v, d in self.G.out_edges(adj_id, data=True)
            if self._pred_str(d) == "describes_aspect"
        ]
        return sorted({self.G.nodes[a].get("label", a) for a in aspects})

    def get_entities(self, aspect: str) -> List[str]:
        """Entities linked from an aspect via 'belongs_to_entity'."""
        aid = self._norm(aspect)
        if aid not in self.G:
            return []
        ents = [
            v for _, v, d in self.G.out_edges(aid, data=True)
            if self._pred_str(d) == "belongs_to_entity"
        ]
        return sorted({self.G.nodes[e].get("label", e) for e in ents})

    def get_domains(self, entity: str) -> List[str]:
        """Domains linked from an entity via 'is_relevant_to_domain'."""
        eid = self._norm(entity)
        if eid not in self.G:
            return []
        doms = [
            v for _, v, d in self.G.out_edges(eid, data=True)
            if self._pred_str(d) == "is_relevant_to_domain"
        ]
        return sorted({self.G.nodes[d].get("label", d) for d in doms})

    # ------------------- Persistence -------------------
    def save_pickle(self, path: str | Path):
        import pickle
        with open(path, "wb") as f:
            pickle.dump({"graph": self.G, "meta": self.meta}, f)

    @classmethod
    def load_pickle(cls, path: str | Path) -> "KnowledgeGraph":
        import pickle
        with open(path, "rb") as f:
            payload = pickle.load(f)
        kg = cls()
        kg.G = payload["graph"]
        kg.meta = payload.get("meta", {})
        # rebuild label caches
        for n, d in kg.G.nodes(data=True):
            label = d.get("label", n)
            kg._id2label[n] = label
            if label not in kg._label2id:
                kg._label2id[label] = n
        return kg
```

```python
def post_process_aspects_list(aspects, original_sentence=None,
                              enforce_verbatim=False,
                              sim_dedupe_threshold=0.80,
                              lexicon=None,
                              empty_to_none=True):
    """
    Post-process a single prediction list: list[str] -> list[str]
    Uses refine_aspects(), then optionally replaces empty with ['none'].
    """
    cleaned = refine_aspects(
        aspects,
        original_sentence=original_sentence,
        enforce_verbatim=enforce_verbatim,
        sim_dedupe_threshold=sim_dedupe_threshold,
        lexicon=lexicon
    )

    if empty_to_none and (not cleaned):
        return ["none"]
    return cleaned


def post_process_pred_para_list(pred_para_list, original_sentence=None,
                                enforce_verbatim=False,
                                sim_dedupe_threshold=0.80,
                                lexicon=None,
                                empty_to_none=True):
    """
    Post-process pred_para_list which is list[list[str]].
    Returns list[list[str]] with cleaning applied to each inner list.
    """
    if pred_para_list is None:
        return []

    # ensure it is a list of lists
    if not isinstance(pred_para_list, list):
        return []

    out = []
    for lst in pred_para_list:
        # each lst should be list[str]
        if lst is None:
            lst = []
        elif not isinstance(lst, list):
            # if somehow a string slipped in, wrap it
            lst = [lst]

        out.append(
            post_process_aspects_list(
                lst,
                original_sentence=original_sentence,
                enforce_verbatim=enforce_verbatim,
                sim_dedupe_threshold=sim_dedupe_threshold,
                lexicon=lexicon,
                empty_to_none=empty_to_none
            )
        )
    return out
```

```python
import re

def normalize_aspect(aspect: str) -> str:
    """
    Basic normalization to reduce duplicates:
    - lowercase
    - remove surrounding punctuation
    - strip plural 's' (English simple rule)
    - strip extra spaces
    """
    asp = aspect.lower().strip()
    asp = re.sub(r"^\W+|\W+$", "", asp)   # remove punctuation at edges
    if asp.endswith("s") and len(asp) > 3:  # crude singularization
        asp = asp[:-1]
    return asp


def filter_to_sentence(aspects, original_sentence):
    """
    Keep only aspects that appear in the original sentence (case-insensitive).
    """
    sent = original_sentence.lower()
    return [a for a in aspects if a.lower() in sent]


# ------------------ NEW HELPERS FOR IG ------------------ #

def _ig_norm_for_overlap(s: str) -> str:
    """
    Normalize aspect strings for overlap comparison.
    Prefer the global _normalize_for_verbatim if available
    (already defined in the paraphrase component), otherwise
    fall back to simple lowercase + whitespace collapse.
    """
    if "_normalize_for_verbatim" in globals():
        return _normalize_for_verbatim(s)
    s = s.lower()
    s = re.sub(r"\s+", " ", s).strip()
    return s


def _ig_is_valid_span(asp: str) -> bool:
    """
    POS-based filter for IG candidates:
      - Reject single-token ADJ / ADV / VERB / AUX (pure adjectives/actions).
      - Require at least one NOUN/PROPN somewhere in the span.
    If spaCy nlp is not available for any reason, we fall back to keeping the span.
    """
    asp = asp.strip()
    if not asp:
        return False

    # If spaCy isn't available, don't block anything here.
    if "nlp" not in globals():
        return True

    doc = nlp(asp)

    # Single-word adjective/verb/etc ‚Üí drop (e.g. "speedy", "cooked", "works")
    if len(doc) == 1 and doc[0].pos_ in {"ADJ", "ADV", "VERB", "AUX"}:
        return False

    # Require at least one noun/proper noun somewhere in the phrase
    has_noun = any(t.pos_ in {"NOUN", "PROPN"} for t in doc)
    if not has_noun:
        return False

    return True


def _ig_overlap_cleanup(aspects):
    """
    Simple overlap cleanup inside IG predictions:
      - If one normalized aspect is a substring of another, keep only the longer one.
      - Uses _ig_norm_for_overlap for robust comparison.
    This is purely intra-IG (does not look at baseline/KG).
    """
    if not aspects:
        return []

    norms = [_ig_norm_for_overlap(a) for a in aspects]
    keep = [True] * len(aspects)

    for i, ni in enumerate(norms):
        if not keep[i] or not ni:
            continue
        for j, nj in enumerate(norms):
            if i == j or not keep[j] or not nj:
                continue
            # substring relation ‚Üí drop the shorter surface form
            if ni in nj or nj in ni:
                if len(aspects[i]) < len(aspects[j]):
                    keep[i] = False
                    break
                else:
                    keep[j] = False

    return [asp for asp, k in zip(aspects, keep) if k]


# ------------------ UPDATED IG COMPONENT ------------------ #

def iterative_residual_decoding2(sentence, max_rounds=3):
    """
    Iterative Residual Decoding with:
      - filtering to aspects that appear in the original sentence
      - POS-based removal of adjective/verb-only spans
      - simple overlap cleanup at the end
    """
    original_sentence = sentence
    remaining_sentence = sentence
    all_aspects = []
    seen_norm = set()

    for round_id in range(max_rounds):
        aspects = extract_aspect(remaining_sentence)  # returns list
        if not aspects:
            break

        # basic cleanup
        new_aspects = [a.strip() for a in aspects if a and a.strip()]
        if not new_aspects or all(a.lower() == "none" for a in new_aspects):
            break

        # keep only those that actually occur in the original sentence
        new_aspects = filter_to_sentence(new_aspects, original_sentence)

        # POS-based filtering: drop adjective-only / verb-only spans
        new_aspects = [a for a in new_aspects if _ig_is_valid_span(a)]

        if not new_aspects:
            # nothing useful this round
            break

        for asp in new_aspects:
            norm = normalize_aspect(asp)
            if norm not in seen_norm:
                seen_norm.add(norm)
                all_aspects.append(asp)  # keep original surface form

            # remove exact form from remaining sentence (case-insensitive)
            pattern = r"\b{}\b".format(re.escape(asp))
            remaining_sentence = re.sub(
                pattern, "", remaining_sentence, flags=re.IGNORECASE
            ).strip()

        if not remaining_sentence:
            break

    # drop literal "none" unless it's the only thing we ever saw
    final_aspects = [
        a for a in all_aspects
        if a.lower() != "none" or len(all_aspects) == 1
    ]

    # overlap cleanup inside IG outputs
    final_aspects = _ig_overlap_cleanup(final_aspects)

    return final_aspects
```

```python
import pandas as pd
from collections import Counter
from sentence_transformers import SentenceTransformer, util
import re, spacy
import math

# load once (global)
_sbert = SentenceTransformer("all-MiniLM-L6-v2")
_nlp = spacy.load("en_core_web_sm")

# ------------ helpers (unchanged) ------------
def _noun_head_lemma(text: str) -> str:
    doc = _nlp(text.lower().strip())
    heads = [t.lemma_ for t in doc if t.pos_ in ("NOUN","PROPN")]
    return heads[-1] if heads else text.lower().strip()


def min_consensus_for_k(k: int) -> int:
    return max(math.floor(k / 3), 1)

def _canonicalize_list(lst):
    return [(_noun_head_lemma(a), a) for a in lst]  # (canon, raw)

def _consensus_on_canonical(aspects_by_input, min_votes=2):
    """Vote using canonical (noun-head lemma) forms, across lists of aspects."""
    cnt = Counter()
    for lst in aspects_by_input:
        canon_set = set(_noun_head_lemma(a) for a in lst)
        for c in canon_set:
            cnt[c] += 1
    return {c for c, v in cnt.items() if v >= min_votes}

def _normalize_for_verbatim(s: str) -> str:
    s = s.lower()
    s = re.sub(r"[\"'`]+", "", s)
    s = re.sub(r"[\(\)\[\]\{\},;:]+", " ", s)
    s = re.sub(r"\s+", " ", s).strip()
    return s

def _explicitness_filter(aspects, original_sentence, jaccard_threshold=0.5):
    sent_norm = _normalize_for_verbatim(original_sentence)
    sent_tokens = set(sent_norm.split())

    kept = []
    for a in aspects:
        low = _normalize_for_verbatim(a)
        if not low:
            continue
        if low in sent_norm:
            kept.append(a); continue
        head = _noun_head_lemma(a)
        if head and head in sent_norm:
            kept.append(a); continue
        a_tokens = set(low.split())
        inter = len(a_tokens & sent_tokens)
        union = len(a_tokens | sent_tokens) or 1
        jacc = inter / union
        if jacc >= jaccard_threshold:
            kept.append(a)
    return kept

def _dedup_preserve(seq):
    seen, out = set(), []
    for x in seq:
        k = _normalize_for_verbatim(x)
        if k not in seen and k:
            seen.add(k)
            out.append(x)
    return out


# ------------ main (edited) ------------
# def extract_aspects_with_paraphrases(
#     sentences,
#     top_k=2,
#     similarity_threshold=0.58,
#     min_consensus=1,
#     include_original_always=True,
#     jaccard_explicitness=0.50,
#     verbose=False,
# ):
#     """
#     Edited version:
#     - sentences must be a list of 11 strings: [original, para1, ..., para10]
#     - runs iterative_residual_decoding ONCE per sentence (11 calls total)
#     - then materializes predictions for k in {0,2,4,6,8,10} WITHOUT extra model calls
#     - returns a 1-row DataFrame with 6 columns:
#         k_0Preds, k_2Preds, k_4Preds, k_6Preds, k_8Preds, k_10Preds
#     """
#     # -------- validation --------
#     if not sentences or not isinstance(sentences, list):
#         return pd.DataFrame([{
#             "k_0Preds": [], "k_2Preds": [], "k_4Preds": [],
#             "k_6Preds": [], "k_8Preds": [], "k_10Preds": []
#         }])

#     original_sentence = sentences[0]
#     paraphrases = sentences[1:]  # could be <10, but expected 10
#     # Ensure we only consider up to 10 paraphrases
#     paraphrases = paraphrases[:10]

#     # -------- 1) EXPENSIVE ONCE: iterative decoding for original + each paraphrase --------
#     per_input = []
#     for s in [original_sentence] + paraphrases:
#         per_input.append(iterative_residual_decoding(s, max_rounds=2))

#     pred_orig = per_input[0]         # list[str]
#     pred_para_list = per_input[1:]   # list[list[str]] length up to 10

#     # Prepare pinned originals once
#     pinned_originals = _dedup_preserve(pred_orig) if include_original_always else []

#     # -------- helper: aggregate for a given k (CHEAP) --------
#     def _aggregate_for_k(k: int):

#         # k=0 means strictly original-only, no paraphrase logic
#         if k == 0 or not pred_para_list:
#             return _dedup_preserve(pinned_originals)

#         para_lists_k = pred_para_list[:k]

#         # 3) paraphrase-derived candidates
#         # 3a) consensus on canonical FOR PARAPHRASES ONLY
#         # A. Here, the min_votes changes accoring to the value of k:
#         mc = min_consensus_for_k(k)
#         voted_canon = _consensus_on_canonical(para_lists_k, min_votes=mc)
#         # B.  the min_consensus is fixed as received in parameters:
#         # voted_canon = _consensus_on_canonical(para_lists_k, min_votes=min_consensus)

#         # map canon -> representative raw (prefer longest)
#         canon_to_raw = {}
#         for lst in para_lists_k:
#             for c, raw in _canonicalize_list(lst):
#                 if c in voted_canon:
#                     if c not in canon_to_raw or len(raw) > len(canon_to_raw[c]):
#                         canon_to_raw[c] = raw

#         candidates = list(canon_to_raw.values())

#         # fallback: if no candidates survived voting, use union of paraphrases (subset)
#         if not candidates:
#             candidates = list({a for lst in para_lists_k for a in lst})

#         # 3b) similarity filter & rank vs original
#         kept_from_para = []
#         if candidates:
#             orig_emb = _sbert.encode(original_sentence, convert_to_tensor=True)
#             asp_embs = _sbert.encode(candidates, convert_to_tensor=True)
#             sims = util.cos_sim(asp_embs, orig_emb).cpu().numpy().flatten().tolist()
#             ranked = sorted(zip(candidates, sims), key=lambda x: x[1], reverse=True)
#             ranked = [(a, s) for a, s in ranked if s >= similarity_threshold]
#             if top_k is not None:
#                 ranked = ranked[:top_k]
#             kept_from_para = [a for a, _ in ranked]

#         # 3c) explicitness vs original (soft)
#         kept_from_para = _explicitness_filter(
#             kept_from_para, original_sentence, jaccard_threshold=jaccard_explicitness
#         )

#         # 4) union + dedup
#         final = _dedup_preserve(pinned_originals + kept_from_para)

#         if verbose:
#             print(f"[PARA-k={k}] orig={len(pinned_originals)}, "
#                   f"para_cands={len(candidates)}, kept_para={len(kept_from_para)}, final={len(final)}")
#         return final

#     # -------- 2) Materialize all requested k columns --------
#     out = {
#         "k_0Preds":  _aggregate_for_k(0),
#         "k_2Preds":  _aggregate_for_k(2),
#         "k_4Preds":  _aggregate_for_k(4),
#         "k_6Preds":  _aggregate_for_k(6),
#         "k_8Preds":  _aggregate_for_k(8),
#         "k_10Preds": _aggregate_for_k(10),
#     }

#     return pd.DataFrame([out])
def extract_aspects_with_paraphrases(
    sentences,
    top_k=2,
    similarity_threshold=0.58,
    include_original_always=True,
    jaccard_explicitness=0.50,
    verbose=False,
):
    """
    - sentences: list[str] expected [original, para1, ..., para10]
    - Runs iterative_residual_decoding ONCE per input sentence (original + paraphrases)
    - Caches:
        pred_orig: list[str]
        pred_para_list: list[list[str]] length up to 10
    - Materializes predictions for k in {0,2,4,6,8,10} WITHOUT extra model calls
    - Returns a 1-row DataFrame with:
        pred_orig, pred_para_list,
        k_0Preds, k_2Preds, k_4Preds, k_6Preds, k_8Preds, k_10Preds
    """
    # -------- validation --------
    empty = {
        "pred_orig": [],
        "pred_para_list": [],
        "k_0Preds": [], "k_2Preds": [], "k_4Preds": [],
        "k_6Preds": [], "k_8Preds": [], "k_10Preds": []
    }
    if not sentences or not isinstance(sentences, list):
        return pd.DataFrame([empty])

    original_sentence = sentences[0]
    paraphrases = sentences[1:] if len(sentences) > 1 else []
    paraphrases = paraphrases[:10]  # enforce max 10

    # -------- 1) EXPENSIVE ONCE: iterative decoding for original + each paraphrase --------
    per_input = []
    for s in [original_sentence] + paraphrases:
        per_input.append(iterative_residual_decoding(s, max_rounds=2))

    pred_orig = per_input[0]          # list[str]
    pred_para_list = per_input[1:]    # list[list[str]] length up to 10

    # PIN original aspects (if enabled)
    pinned_originals = _dedup_preserve(pred_orig) if include_original_always else []

    # -------- helper: aggregate for a given k (CHEAP) --------
    def _aggregate_for_k(k: int):
        # k=0 means strictly original-only, no paraphrase logic
        if k == 0 or not pred_para_list:
            return _dedup_preserve(pinned_originals)

        para_lists_k = pred_para_list[:k]

        # A) dynamic min_consensus based on k
        mc = min_consensus_for_k(k)

        # 3a) consensus on canonical FOR PARAPHRASES ONLY
        voted_canon = _consensus_on_canonical(para_lists_k, min_votes=mc)

        # map canon -> representative raw (prefer longest)
        canon_to_raw = {}
        for lst in para_lists_k:
            for c, raw in _canonicalize_list(lst):
                if c in voted_canon:
                    if c not in canon_to_raw or len(raw) > len(canon_to_raw[c]):
                        canon_to_raw[c] = raw

        candidates = list(canon_to_raw.values())

        # fallback: if no candidates survived voting, use union of paraphrases (subset)
        if not candidates:
            candidates = list({a for lst in para_lists_k for a in lst})

        # 3b) similarity filter & rank vs original
        kept_from_para = []
        if candidates:
            orig_emb = _sbert.encode(original_sentence, convert_to_tensor=True)
            asp_embs = _sbert.encode(candidates, convert_to_tensor=True)
            sims = util.cos_sim(asp_embs, orig_emb).cpu().numpy().flatten().tolist()
            ranked = sorted(zip(candidates, sims), key=lambda x: x[1], reverse=True)
            ranked = [(a, s) for a, s in ranked if s >= similarity_threshold]
            if top_k is not None:
                ranked = ranked[:top_k]
            kept_from_para = [a for a, _ in ranked]

        # 3c) explicitness vs original (soft)
        kept_from_para = _explicitness_filter(
            kept_from_para, original_sentence, jaccard_threshold=jaccard_explicitness
        )

        # 4) union + dedup (originals protected)
        final = _dedup_preserve(pinned_originals + kept_from_para)

        if verbose:
            print(f"[PARA-k={k}] mc={mc} orig={len(pinned_originals)}, "
                  f"cands={len(candidates)}, kept_para={len(kept_from_para)}, final={len(final)}")

        return final

    # -------- 2) Materialize all requested k columns --------
    out = {
        "pred_orig": pred_orig,
        "pred_para_list": pred_para_list,

        "k_0Preds":  _aggregate_for_k(0),
        "k_2Preds":  _aggregate_for_k(2),
        "k_4Preds":  _aggregate_for_k(4),
        "k_6Preds":  _aggregate_for_k(6),
        "k_8Preds":  _aggregate_for_k(8),
        "k_10Preds": _aggregate_for_k(10),
    }

    return pd.DataFrame([out])
```

#### 3. Knowledge graph for Implicit aspects extraction:

```python
import json
from collections import defaultdict
from typing import Dict, List, Tuple, Optional

import torch
from sentence_transformers import SentenceTransformer, util

# Multilingual so you can later use Arabic
embedder = SentenceTransformer("sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2")

def list_domains(kg) -> List[str]:
    """List all domain node labels from the KG."""
    out = []
    for nid, attrs in kg.G.nodes(data=True):
        if attrs.get("type") == "domain":
            out.append(attrs.get("label", nid))
    return sorted(set(out))

def build_domain_keywords(kg) -> Dict[str, List[str]]:
    """
    Build keywords per domain using your helpers:
      - domain label itself
      - all entities of that domain
      - all aspects belonging to those entities
    """
    domain_kws = defaultdict(set)
    domains = list_domains(kg)
    for dom in domains:
        domain_kws[dom].add(dom)
        ents = kg.get_entities_by_domain(dom)
        for e in ents:
            domain_kws[dom].add(e)
            # add aspects for each entity
            aspects = kg.get_aspects_by_entity(e)
            for a in aspects:
                domain_kws[dom].add(a)
    # convert to sorted lists for readability
    return {d: sorted(list(s)) for d, s in domain_kws.items()}

def create_domain_embeddings_from_keywords(domain_keywords: Dict[str, List[str]]):
    """Create centroid embeddings and print debug info."""
    domain_embs = {}
    # print("\n[DBG] Building domain embeddings:")
    for dom, kws in domain_keywords.items():
        # print(f"  - {dom}: {len(kws)} keywords -> {kws[:10]}{' ...' if len(kws)>10 else ''}")
        if not kws:
            continue
        embs = embedder.encode(kws, convert_to_tensor=True)
        domain_embs[dom.lower()] = torch.mean(embs, dim=0)
    # print("[DBG] Domain embeddings ready for:", list(domain_embs.keys()))
    return domain_embs

def infer_domain_with_centroid_dbg(sentence: str, domain_embs) -> Tuple[Optional[str], float]:
    """Infer domain and print per-domain scores."""
    if not domain_embs:
        # print("[DBG] No domain embeddings available.")
        return None, 0.0
    s_emb = embedder.encode(sentence, convert_to_tensor=True)
    scores = {d: util.cos_sim(s_emb, e).item() for d, e in domain_embs.items()}
    if not scores:
        # print("[DBG] Empty domain score map.")
        return None, 0.0
    # print("\n[DBG] Domain scores:")
    # for d, sc in sorted(scores.items(), key=lambda x: x[1], reverse=True):
    #     print(f"  - {d}: {sc:.4f}")
    best = max(scores, key=scores.get)
    return best, scores[best]


import spacy
nlp = spacy.load("en_core_web_sm")  # swap later for Arabic

def find_lonely_adjectives_dbg(sentence: str) -> List[str]:
    doc = nlp(sentence)
    # print("\n[DBG] Token/POS/Head:")
    for t in doc:
        print(f"  - '{t.text}' POS={t.pos_}  head='{t.head.text}'({t.head.pos_})  children={[c.text for c in t.children]}")
    lonely = []
    for tok in doc:
        if tok.pos_ == "ADJ":
            has_noun_child = any(ch.pos_ in ("NOUN","PROPN") for ch in tok.children)
            head_is_noun = tok.head.pos_ in ("NOUN","PROPN")
            is_lonely = (not has_noun_child) and (not head_is_noun)
            # print(f"[DBG] ADJ '{tok.text}' lemma='{tok.lemma_.lower()}': "
            #       f"has_noun_child={has_noun_child}, head_is_noun={head_is_noun} -> lonely={is_lonely}")
            if is_lonely:
                lonely.append(tok.lemma_.lower())
    # print("[DBG] Lonely adjectives:", lonely)
    return lonely

def debug_traces_for_adj(kg, adjective: str):
    """Use your get_traces_for_adjective to print full paths."""
    traces = kg.get_traces_for_adjective(adjective)
    if not traces:
        # print(f"[DBG] No KG traces for adjective '{adjective}'.")
        # also show direct aspects if any
        aspects = kg.get_aspects(adjective)
        # print(f"[DBG] Direct aspects via get_aspects('{adjective}'): {aspects}")
        return
    # print(f"[DBG] Traces for '{adjective}':")
    # for tr in traces:
        # print(f"   ADJ='{tr['adjective']}' -> ASPECT='{tr['aspect']}' "
        #       f"-> ENTITY='{tr['entity']}' -> DOMAIN='{tr['domain']}'")


def choose_best_aspect_for_adj(adj: str, aspects: List[str], top_k=1):
    """
    From candidate aspects for an adjective, choose the most semantically relevant one.
    """
    if not aspects:
        return []

    # Embed adjective
    adj_emb = embedder.encode(adj, convert_to_tensor=True)
    # Embed candidate aspects
    asp_embs = embedder.encode(aspects, convert_to_tensor=True)

    # Similarities: (1 x n) tensor
    sims = util.cos_sim(adj_emb, asp_embs)  # shape [1, n]
    sims = sims.cpu().numpy().flatten().tolist()  # always a list of floats

    # Pair and sort
    ranked = sorted(zip(aspects, sims), key=lambda x: x[1], reverse=True)

    # Debug
    # print(f"[DBG] Ranking aspects for adjective '{adj}':")
    # for asp, sc in ranked:
    #     print(f"   - {asp} (sim={sc:.3f})")

    return [asp for asp, _ in ranked[:top_k]]

def map_aspect_to_category(aspect: str, domain: str, domain_categories: Dict[str, List[str]], top_k: int = 1) -> List[str]:
    """
    Map a KG aspect to the closest category in the given domain.
    Returns top_k categories ranked by embedding similarity.
    """
    if domain not in domain_categories:
        return []

    candidates = domain_categories[domain]
    if not candidates:
        return []

    # Encode
    asp_emb = embedder.encode(aspect, convert_to_tensor=True)
    cat_embs = embedder.encode(candidates, convert_to_tensor=True)

    # Similarities
    sims = util.cos_sim(asp_emb, cat_embs).cpu().numpy().flatten().tolist()
    ranked = sorted(zip(candidates, sims), key=lambda x: x[1], reverse=True)

    return [cat for cat, _ in ranked[:top_k]]

def infer_implicit_with_category(sentence, kg, domain_embs, domain_categories, domain_conf_threshold=0.15):
    lonely = find_lonely_adjectives_dbg(sentence)
    if not lonely:
        return []

    dom, dom_score = infer_domain_with_centroid_dbg(sentence, domain_embs)
    if not dom or dom_score < domain_conf_threshold:
        return []

    results = []
    for adj in lonely:
        aspects = kg.get_aspects(adj)

        # filter aspects by domain
        valid_aspects = []
        for asp in aspects:
            ents = kg.get_entities(asp)
            doms = {d.lower() for e in ents for d in kg.get_domains(e)}
            if dom.lower() in doms:
                valid_aspects.append(asp)

        best_aspects = choose_best_aspect_for_adj(adj, valid_aspects, top_k=1)

        for asp in best_aspects:
            cats = map_aspect_to_category(asp, dom, domain_categories, top_k=1)
            if cats:
                results.append(f"{asp}#{cats[0]}")   # return aspect#category

    # Deduplicate
    seen, final = set(), []
    for item in results:
        if item.lower() not in seen:
            seen.add(item.lower())
            final.append(item)
    return final
```

```python
# ========== 2. Apply KG inference only where baseline == ['none'] or [] ==========
def apply_kg_if_none(row):
    base = row["baseline"]

    # Normalize to a list of strings
    if not isinstance(base, list):
        base = [str(base).strip().lower()]
    else:
        base = [str(x).strip().lower() for x in base]

    # Trigger KG if it's ['none'] or empty list
    if base == ["none"] or len(base) == 0:
        return infer_implicit_with_category(
            row["sentence"],
            kg,
            domain_embs,
            domain_categories,
            domain_conf_threshold=0.1
        )
    else:
        return []
```

#### Final KG Block:

```python
# ====================== KG-only inference, gated by baseline=='none' ======================
# - Triggers KG inference ONLY when row['baseline'] == 'none' (case-insensitive)
# - Uses KG adjectives/terms (no hardcoded clue lists) + paraphrases
# - "general" is strict fallback; never mixed with specifics
# - Prunes to top-K with tie keeping
# - Writes only one column to `test`: test["kg_final_preds"] (list)
# - Full traces go to separate DF: KG_debug
#
# ---- UPDATE (post-processing) ----
# * Canonicalize near-duplicate labels (domain-aware)
# * Collapse to generic: if both generic & its family specifics are present, keep only the generic
# * (Optional, not invoked): collapse to specific (inverse policy), left commented
# * Added detailed post-processing trace in KG_debug["pp_trace"]
# ----------------------------------

from typing import List, Dict, Tuple, Optional, Any
from collections import defaultdict
import pandas as pd
import re, ast
from difflib import SequenceMatcher

# --------------------------- Config ---------------------------
MAX_KEEP = 2                     # keep up to K aspects (ties allowed) -- unchanged
KEEP_TIES = True                 # include ties with the Kth score         -- unchanged
CAP_PARAPHRASES = 2              # inspect at most 2 paraphrases per row   -- unchanged
GENERAL_NAME = "general"         # strict fallback label                   -- unchanged

# Make "location" a bit stricter to reduce false positives
MIN_HITS_BY_ASPECT_DEFAULT = {
    "location": 2,               # require >= 2 distinct term hits across all sources
}
MIN_HITS_BY_ASPECT_BY_DOMAIN = {
    "restaurant": {"location": 2},
    "hotel": {"location": 2},
}

# Generic sentiment terms -> never counted for specific aspects (reserved for fallback/general)
GENERIC_SENT = {
    "good","great","amazing","awesome","fantastic","fabulous","perfect","excellent","nice","decent","ok","okay",
    "bad","awful","terrible","horrible","disappointing","meh","ugh","wow","love","hate","liked","disliked",
    "best","better","worse","worst","fine","average","mediocre","overrated","underrated","superb","pleasant",
    "wonderful","horrid","lovely","lousy","solid","mid"
}

# --------------------------- Post-processing toggles ---------------------------
ENABLE_CANON_LABELS        = True   # map labels to a canonical spelling (domain-aware)
ENABLE_COLLAPSE_TO_GENERIC = True   # if both "quality" & {"food quality","taste"}: keep "quality"
ENABLE_PREFER_SPECIFICS    = False  # inverse policy; KEEP False as per preference
ENABLE_SCORE_FLOOR         = True   # implemented as NO-OP by default (thresholds set to None/0)

# Score floor settings (NO-OP unless you change these)
SCORE_RELATIVE_BAND = 0.0          # keep aspects within 0% of top (i.e., keep all)
SCORE_ABSOLUTE_FLOOR = None        # if set (e.g., 1.0), drops weak aspects

# Canonical label map (domain-aware)
CANON_MAP = {
    "restaurant": {
        "noise level": "noise",
        "portion size": "portion",
        "food cleanliness": "cleanliness",
        "taste/flavor": "taste",
        "flavor": "taste",
        "spices": "spice level",
    },
    "hotel": {
        "noise level": "noise",
        "room cleanliness": "cleanliness",
        "bed": "bed comfort",
        "beds": "bed comfort",
        "mattress": "bed comfort",
        "pillow": "bed comfort",
        "pillows": "bed comfort",
        "ac": "air conditioning",
        "a/c": "air conditioning",
        "wi-fi": "wifi",
        "wi fi": "wifi",
    },
    "laptop": {
        "build quality": "quality",
        "thermals/fan": "thermals",
        "battery": "battery life",
        "screen": "display",
    },
}

# Collapse families (domain-aware): keep the GENERIC key if any of its family members are present.
# IMPORTANT: This is why ENABLE_COLLAPSE_TO_GENERIC might have looked like a no-op earlier:
# you must list all specifics you want to fold into the generic (e.g., include "taste" under "quality").
GENERIC_FAMILIES = {
    "restaurant": {
        "quality": {"food quality", "taste"},   # <‚Äî now includes "taste"
        # add more if you need to fold specifics into a generic
    },
    "hotel": {
        "cleanliness": {"room cleanliness"},
        "comfort": {"bed comfort"},             # if you ever produce "comfort"
    },
    "laptop": {
        "connectivity": {"wifi", "bluetooth"},
        "quality": {"build quality"},
    },
}

# For the optional ‚Äúprefer specifics‚Äù inverse policy (not invoked)
SPECIFIC_FAMILIES = {
    "restaurant": {
        "quality": {"taste", "food quality"},
    },
    "hotel": {
        "cleanliness": {"room cleanliness"},
        "comfort": {"bed comfort"},
    },
    "laptop": {
        "connectivity": {"wifi", "bluetooth"},
        "quality": {"build quality"},
    },
}

# --------------------------- Small utilities ---------------------------
_DOMAIN_MAP = {
    "restaurants":"restaurant","restaurant":"restaurant",
    "hotels":"hotel","hotel":"hotel",
    "laptops":"laptop","laptop":"laptop"
}
def _norm_domain(d: Optional[str]) -> Optional[str]:
    if d is None: return None
    d = str(d).strip().lower()
    return _DOMAIN_MAP.get(d, d)

def _parse_listlike(x) -> List[str]:
    if isinstance(x, list): return [str(t) for t in x if isinstance(t, (str,int,float))]
    if x is None: return []
    s = str(x).strip()
    if not s: return []
    try:
        v = ast.literal_eval(s)
        if isinstance(v, list): return [str(t) for t in v if isinstance(t, (str,int,float))]
    except Exception:
        pass
    return [s]

def _texts_from_row(row, cap=CAP_PARAPHRASES) -> List[Tuple[str,str]]:
    sent = str(row.get("sentence","") or "")
    out = [("orig", sent)]
    paras = _parse_listlike(row.get("paraphrases"))
    for i, p in enumerate(paras[:cap]):
        ps = str(p).strip()
        if ps:
            out.append((f"para{i+1}", ps))
    return out

def _baseline_is_none(b):
    """Trigger KG only if baseline is string 'none' (case-insensitive)."""
    if isinstance(b, str) and b.strip().lower() == "none":
        return True
    return False

# --------------------------- KG helpers (robust to API shape) ---------------------------
def aspects_for_domain(kg, domain: str) -> List[str]:
    aspects = set()
    # direct accessor
    if hasattr(kg, "get_aspects_by_domain"):
        try:
            for a in kg.get_aspects_by_domain(domain) or []:
                if a: aspects.add(str(a).strip().lower())
        except Exception:
            pass
    # via entities
    if hasattr(kg, "get_entities_by_domain") and hasattr(kg, "get_aspects_by_entity"):
        try:
            for e in kg.get_entities_by_domain(domain) or []:
                for a in kg.get_aspects_by_entity(e) or []:
                    if a: aspects.add(str(a).strip().lower())
        except Exception:
            pass
    # never allow "general" among specifics
    aspects.discard(GENERAL_NAME)
    return sorted(aspects)

def _terms_for_aspect(kg, domain: str, aspect: str) -> List[str]:
    """
    Pull terms (adjectives/keywords) for an aspect.
    Primary: kg.get_adjectives_for_aspect(aspect)
    Optional: kg.lexicon[domain][aspect] if present
    Optional: other API names if you add them later.
    """
    terms = set()

    # 1) adjectives linked to aspect
    if hasattr(kg, "get_adjectives_for_aspect"):
        try:
            for t in kg.get_adjectives_for_aspect(aspect) or []:
                if t: terms.add(str(t).strip().lower())
        except Exception:
            pass

    # 2) optional lexicon dict
    if hasattr(kg, "lexicon"):
        try:
            for t in kg.lexicon.get(domain, {}).get(aspect, []) or []:
                if t: terms.add(str(t).strip().lower())
        except Exception:
            pass

    # 3) optional additional APIs
    for meth in ("get_terms_for_aspect","get_lexicon_for_aspect","get_keywords_for_aspect",
                 "get_triggers_for_aspect","get_aspect_terms"):
        if hasattr(kg, meth):
            fn = getattr(kg, meth)
            try:
                for t in (fn(domain, aspect) or []):
                    if t: terms.add(str(t).strip().lower())
            except TypeError:
                try:
                    for t in (fn(aspect) or []):
                        if t: terms.add(str(t).strip().lower())
                except Exception:
                    pass
            except Exception:
                pass

    # Filter out generic sentiment ‚Äî belongs to GENERAL fallback only
    terms = {t for t in terms if t not in GENERIC_SENT and 1 <= len(t) <= 64}
    return sorted(terms, key=len, reverse=True)

def _compile_terms(terms: List[str]) -> List[Tuple[str, re.Pattern]]:
    out = []
    for t in terms:
        out.append((t, re.compile(rf"(?<!\w){re.escape(t)}(?!\w)", re.I)))
    return out

# --------------------------- Core scanning & scoring ---------------------------
def _scan_with_index(text: str, index: Dict[str, List[Tuple[str, re.Pattern]]]) -> Dict[str, List[str]]:
    """Returns: aspect -> list of matched terms (distinct)."""
    text_l = (text or "").lower()
    hits: Dict[str, set] = defaultdict(set)
    for aspect, pats in index.items():
        for term, rx in pats:
            if rx.search(text_l):
                hits[aspect].add(term)
    return {a: sorted(list(s)) for a, s in hits.items()}

def _aggregate_sources(hits_by_src: List[Dict[str, List[str]]]) -> Tuple[Dict[str, List[str]], Dict[str, float], Dict[str,int]]:
    """
    Aggregate across sources (orig + paraphrases).
    We use MAX across sources to avoid inflating weak signals.
    Returns:
      all_hits: aspect -> unique list of terms across all sources
      max_scores: aspect -> float score (one source's strength)
      src_count: aspect -> #sources where aspect fired
    """
    TYPE_BASE = 1.0  # all terms equal; multiword gets a tiny bump
    all_hits = defaultdict(set)
    max_scores = defaultdict(float)
    src_count = defaultdict(int)

    for src_hits in hits_by_src:
        src_score = {}
        for a, terms in src_hits.items():
            sc = 0.0
            for t in terms:
                sc += TYPE_BASE + (0.05 if (" " in t or "-" in t) else 0.0)
            src_score[a] = sc
            if terms:
                all_hits[a] |= set(terms)
                src_count[a] += 1
        for a, sc in src_score.items():
            if sc > max_scores[a]:
                max_scores[a] = sc

    return {a: sorted(list(ts)) for a, ts in all_hits.items()}, dict(max_scores), dict(src_count)

def _apply_thresholds(domain: str, all_hits: Dict[str, List[str]], max_scores: Dict[str, float]) -> Dict[str, float]:
    # base minima
    mins = dict(MIN_HITS_BY_ASPECT_DEFAULT)
    # domain-specific overrides
    d = _norm_domain(domain)
    if d in MIN_HITS_BY_ASPECT_BY_DOMAIN:
        for k, v in MIN_HITS_BY_ASPECT_BY_DOMAIN[d].items():
            mins[k] = v
    filtered = {}
    for a, terms in all_hits.items():
        need = mins.get(a, 1)
        if len(set(terms)) >= need:
            filtered[a] = max_scores.get(a, 0.0)
    return filtered

def _rank_and_prune(scores: Dict[str, float], k: int = MAX_KEEP, keep_ties: bool = KEEP_TIES) -> List[str]:
    if not scores:
        return []
    items = sorted(scores.items(), key=lambda x: x[1], reverse=True)
    if k is None or k <= 0:
        return [a for a,_ in items]
    kept = items[:k]
    if keep_ties and len(items) > k:
        kth = kept[-1][1]
        for a, sc in items[k:]:
            if abs(sc - kth) < 1e-9:
                kept.append((a, sc))
            else:
                break
    return [a for a,_ in kept]

# --------------------------- Post-processing helpers (NEW) ---------------------------
def _canonize_labels(labels: List[str], domain: str) -> List[str]:
    """Domain-aware canonicalization + dedup while preserving order."""
    dom = _norm_domain(domain)
    cmap = CANON_MAP.get(dom, {})
    seen = set()
    out = []
    for a in labels:
        a0 = (a or "").strip().lower()
        a1 = cmap.get(a0, a0)
        if a1 not in seen:
            seen.add(a1)
            out.append(a1)
    return out

def _collapse_generic_when_specific_present(labels: List[str], domain: str) -> List[str]:
    """
    If a generic and one of its family specifics co-exist, keep the generic and drop its specifics.
    'general' is never touched here.
    """
    dom = _norm_domain(domain)
    fam = GENERIC_FAMILIES.get(dom, {})
    label_set = set(labels)
    to_drop = set()
    for generic, specifics in fam.items():
        if (generic in label_set) and (label_set & specifics):
            to_drop |= (label_set & specifics)
    if not to_drop:
        return labels
    out = [a for a in labels if a not in to_drop]
    return out

def _prefer_specifics_over_generic(labels: List[str], domain: str) -> List[str]:
    """
    Optional inverse policy (NOT INVOKED): if a generic and specifics co-exist, drop the generic.
    """
    dom = _norm_domain(domain)
    fam = SPECIFIC_FAMILIES.get(dom, {})
    label_set = set(labels)
    to_drop = set()
    for generic, specifics in fam.items():
        if (generic in label_set) and (label_set & specifics):
            to_drop.add(generic)
    if not to_drop:
        return labels
    out = [a for a in labels if a not in to_drop]
    return out

def _score_floor_filter(scores: Dict[str, float]) -> Dict[str, float]:
    """
    Keep aspects within a relative band of the top score OR above an absolute floor.
    NO-OP with defaults: SCORE_RELATIVE_BAND=0.0 and SCORE_ABSOLUTE_FLOOR=None.
    """
    if not scores:
        return scores
    top = max(scores.values())
    out = {}
    for a, sc in scores.items():
        keep_rel = (SCORE_RELATIVE_BAND is not None) and (sc >= (1.0 - SCORE_RELATIVE_BAND) * top)
        keep_abs = (SCORE_ABSOLUTE_FLOOR is not None) and (sc >= SCORE_ABSOLUTE_FLOOR)
        if SCORE_ABSOLUTE_FLOOR is None and SCORE_RELATIVE_BAND == 0.0:
            # default: keep all (no-op)
            out[a] = sc
        else:
            if keep_rel or keep_abs:
                out[a] = sc
    return out

def _tokset(s: str) -> str:
    # token-set normalization: lowercase, word tokens, unique, sorted
    toks = re.findall(r"\w+", s.lower())
    return " ".join(sorted(set(toks)))

def _jaccard(a: str, b: str) -> float:
    sa, sb = set(a.split()), set(b.split())
    if not sa and not sb: return 1.0
    if not sa or not sb:  return 0.0
    inter = len(sa & sb)
    union = len(sa | sb)
    return inter / union

def dedupe_near_duplicates(labels, sim_floor=0.75, jaccard_floor=0.5):
    """
    Collapses near-duplicates like 'food quality' vs 'quality'.
    - token-set Jaccard first (order-insensitive)
    - char-level SequenceMatcher as backup
    - prefers the *shorter canonical* (tends to be the generic)
    """
    finals = []
    for cand in labels:
        add = True
        c_norm = _tokset(cand)
        for i, ex in enumerate(finals):
            e_norm = _tokset(ex)
            jac = _jaccard(c_norm, e_norm)
            sim = SequenceMatcher(None, c_norm, e_norm).ratio()
            if jac >= jaccard_floor or sim >= sim_floor:
                # prefer the shorter normalized form (often the generic)
                if len(c_norm) < len(e_norm):
                    finals[i] = cand
                # else keep existing
                add = False
                break
        if add:
            finals.append(cand)
    return finals

# --------------------------- Row driver (with baseline gate) ---------------------------
def _infer_row_with_debug(row: pd.Series, kg) -> Dict[str, Any]:
    # Gate: run KG only if baseline == 'none'
    if not _baseline_is_none(row.get("baseline")):
        return {
            "kg_final_preds": [],  # no KG inference for explicit rows
            "kg_debug": {
                "skipped": True,
                "reason": "skip_not_implicit_baseline",
                "baseline": row.get("baseline"),
            },
        }

    sentence = str(row.get("sentence", "") or "")
    dom = _norm_domain(row.get("domain"))
    if not sentence or not dom:
        return {
            "kg_final_preds": [GENERAL_NAME],
            "kg_debug": {
                "skipped": False,
                "domain": dom,
                "reason": "fallback_no_text_or_domain",
                "all_hits": {},
                "max_scores": {},
                "src_count": {},
                "kept": [GENERAL_NAME],
                "pp_trace": {},
            },
        }

    # allowed aspects from KG
    allowed = aspects_for_domain(kg, dom)
    if not allowed:
        return {
            "kg_final_preds": [GENERAL_NAME],
            "kg_debug": {
                "skipped": False,
                "domain": dom,
                "reason": "fallback_no_aspects_for_domain",
                "all_hits": {},
                "max_scores": {},
                "src_count": {},
                "kept": [GENERAL_NAME],
                "pp_trace": {},
            },
        }

    # build index: aspect -> [(term, regex)]
    index = {}
    for a in allowed:
        terms = _terms_for_aspect(kg, dom, a)
        if terms:
            index[a] = _compile_terms(terms)

    # scan original + paraphrases
    texts = _texts_from_row(row, cap=CAP_PARAPHRASES)
    hits_by_src = []
    for _, txt in texts:
        hits_by_src.append(_scan_with_index(txt, index))

    all_hits, max_scores, src_count = _aggregate_sources(hits_by_src)

    # nothing fired => general
    if not all_hits:
        return {
            "kg_final_preds": [GENERAL_NAME],
            "kg_debug": {
                "skipped": False,
                "domain": dom,
                "reason": "fallback_no_kg_hits",
                "texts_scanned": {src: txt for src, txt in texts},
                "all_hits": {},
                "max_scores": {},
                "src_count": {},
                "kept": [GENERAL_NAME],
                "pp_trace": {},
            },
        }

    # thresholds
    filtered = _apply_thresholds(dom, all_hits, max_scores)

    # (optional) score floor ‚Äî default is effectively a NO-OP with current settings
    if ENABLE_SCORE_FLOOR:
        filtered = _score_floor_filter(filtered)

    # rank & prune (unchanged)
    kept = _rank_and_prune(filtered, k=MAX_KEEP, keep_ties=KEEP_TIES)

    # final safety from original logic: if specifics exist, don't include 'general'
    if not kept:
        kept = [GENERAL_NAME]
        reason = "fallback_after_thresholds"
    else:
        kept = [a for a in kept if a != GENERAL_NAME]
        if not kept:
            kept = [GENERAL_NAME]
            reason = "fallback_only_general_left"
        else:
            reason = "ok_specific"

    # ---------------- POST-PROCESSING (NEW) ----------------
    pp_trace = {"pre_canon": kept.copy()}

    if ENABLE_CANON_LABELS:
        kept = _canonize_labels(kept, dom)
    pp_trace["post_canon"] = kept.copy()

    if ENABLE_COLLAPSE_TO_GENERIC:
        kept = _collapse_generic_when_specific_present(kept, dom)
    pp_trace["post_collapse_generic"] = kept.copy()


    # Optional inverse policy (NOT INVOKED):
    # if ENABLE_PREFER_SPECIFICS:
    #     kept = _prefer_specifics_over_generic(kept, dom)
    # pp_trace["post_prefer_specifics"] = kept.copy()

    kept=dedupe_near_duplicates(kept)
    pp_trace["post_dedupe"] = kept.copy()

    # -------------------------------------------------------

    return {
        "kg_final_preds": kept,
        "kg_debug": {
            "skipped": False,
            "domain": dom,
            "reason": reason,
            "texts_scanned": {src: txt for src, txt in texts},
            "all_hits": all_hits,         # aspect -> matched terms (unique across sources)
            "max_scores": filtered,       # aspect -> score after thresholds & floor
            "src_count": src_count,       # aspect -> in how many sources it fired
            "kept": kept,                 # final kept aspects (after post-processing)
            "pp_trace": pp_trace,         # NEW: shows how post-processing changed the list
        },
    }

# --------------------------- Run over your DataFrame ---------------------------
_result_series = test.apply(lambda r: _infer_row_with_debug(r, kg), axis=1)

# keep just the predictions in `test`
test["kg_final_preds"] = _result_series.apply(lambda d: d["kg_final_preds"])

# everything else (rich trace) goes to KG_debug
KG_debug = pd.json_normalize(_result_series.apply(lambda d: d["kg_debug"]))
test[test['type']=='implicit']
```

##### KG Evaluation:

```python
# =================== Evaluation: strict vs. relaxed (+ recall & row result) ===================
# Expects:
#   test["kg_final_preds"] : list of predicted aspect strings (or stringified list)
#   test["category"]       : "ENTITY#ASPECT" string or list of such strings

import re, ast
import pandas as pd
from typing import List, Tuple, Dict, Any

# ---------------- Normalization helpers ----------------
GENERAL_SYNONYMS = {"misc", "miscellaneous", "anecdotes/miscellaneous", "miscellany", "other", "general"}

def _to_list(x):
    if isinstance(x, list):
        return x
    if x is None:
        return []
    s = str(x).strip()
    if not s:
        return []
    try:
        v = ast.literal_eval(s)
        if isinstance(v, list):
            return v
    except Exception:
        pass
    return [s]

def _norm_phrase(s: str) -> str:
    if s is None:
        return ""
    s = str(s).lower().strip()
    s = s.replace("&", " and ")
    s = re.sub(r"[-_/]+", " ", s)
    s = re.sub(r"[^a-z0-9\s]", " ", s)
    s = re.sub(r"\s+", " ", s).strip()
    if s in GENERAL_SYNONYMS:
        return "general"
    if s == "prices":  # common canonicalization
        return "price"
    if s == "rooms":
        return "room"
    if s == "drinks":
        return "drink"
    return s

_STOPWORDS = {
    "the","a","an","of","and","to","for","in","on","at","with","without","very",
    "more","most","less","overall","general","misc","miscellaneous","level","size",
    "options","style","option","variety","selection"
}
# often modifiers rather than informative heads
_MODIFIERS = {"quality","comfort","cleanliness","performance","service","condition"}

def _singularize(w: str) -> str:
    if w.endswith("ies") and len(w) > 3:
        return w[:-3] + "y"
    if w.endswith("ses") and len(w) > 3:
        return w[:-2]
    if w.endswith("s") and len(w) > 3:
        return w[:-1]
    return w

def _tokens(s: str) -> List[str]:
    s = _norm_phrase(s)
    toks = [t for t in s.split() if t and t not in _STOPWORDS]
    return toks

def _head_noun(s: str) -> str:
    toks = [t for t in _tokens(s) if t not in _MODIFIERS]
    if not toks:
        toks = _tokens(s)
    if not toks:
        return ""
    return _singularize(toks[-1])

def _token_set(s: str) -> set:
    return set(_tokens(s))

# ---------------- Category parsing ----------------
def _parse_category_item(item: str) -> Tuple[str, str]:
    """
    'ENTITY#ASPECT' ‚Üí (entity_norm, aspect_norm).
    If no '#', treat as aspect-only on the right.
    """
    if item is None:
        return ("", "")
    s = str(item).strip()
    if "#" in s:
        ent, asp = s.split("#", 1)
        ent, asp = _norm_phrase(ent), _norm_phrase(asp)
    else:
        ent, asp = "", _norm_phrase(s)
    if ent in GENERAL_SYNONYMS: ent = "general"
    if asp in GENERAL_SYNONYMS: asp = "general"
    return ent, asp

def _get_gold_pairs(row) -> List[Tuple[str,str]]:
    items = _to_list(row.get("category"))
    out = []
    for it in items:
        ent, asp = _parse_category_item(it)
        if ent or asp:
            out.append((ent, asp))
    return out

# ---------------- Matching logic (per prediction vs any gold pair) ----------------
def _exact_match(pred: str, ent: str, asp: str) -> bool:
    p = _norm_phrase(pred)
    return p and (p == ent or p == asp)

def _partial_match(pred: str, ent: str, asp: str) -> bool:
    """Overlap but not exact (token overlap / substring either way)."""
    p = _norm_phrase(pred)
    if not p:
        return False
    pe = _token_set(p) & _token_set(ent)
    pa = _token_set(p) & _token_set(asp)
    if (p in ent or ent in p) and p != ent:
        return True
    if (p in asp or asp in p) and p != asp:
        return True
    return bool(pe or pa) and not (p == ent or p == asp)

def _relaxed_head_match(pred: str, ent: str, asp: str) -> bool:
    hp = _head_noun(pred)
    he = _head_noun(ent)
    ha = _head_noun(asp)
    return bool(hp) and (hp == he or hp == ha)

def evaluate_prediction_list(preds: List[str], gold_pairs: List[Tuple[str,str]], mode: str = "strict") -> Dict[str, Any]:
    """
    Returns:
      - per-pred verdicts ("TP"/"FP"/"FN")
      - counts tp/fp/fn
      - any_tp flag
      - gold coverage (for recall): how many gold pairs are correctly covered by at least one prediction
    """
    preds = [str(p) for p in preds if str(p).strip()]
    verdicts = []  # list of ("label", "TP/FP/FN")
    tp = fp = fn = 0

    # Per-prediction classification
    for p in preds:
        matched_exact = any(_exact_match(p, e, a) for e, a in gold_pairs)
        if matched_exact:
            verdicts.append((p, "TP")); tp += 1; continue

        if mode == "relaxed":
            matched_relaxed = any(_relaxed_head_match(p, e, a) for e, a in gold_pairs)
            if matched_relaxed:
                verdicts.append((p, "TP")); tp += 1; continue

        matched_partial = any(_partial_match(p, e, a) for e, a in gold_pairs)
        if matched_partial:
            verdicts.append((p, "FP")); fp += 1
        else:
            verdicts.append((p, "FN")); fn += 1

    # Gold coverage for recall (count gold pairs covered by ANY correct prediction)
    def _pred_covers_pair(p, pair):
        e, a = pair
        return _exact_match(p, e, a) or (mode == "relaxed" and _relaxed_head_match(p, e, a))

    covered = 0
    for pair in gold_pairs:
        if any(_pred_covers_pair(p, pair) for p, v in verdicts if p):
            covered += 1

    return {
        "tp": tp, "fp": fp, "fn": fn,
        "verdicts": verdicts,
        "n_preds": len(preds),
        "any_tp": tp > 0,
        "gold_total": len(gold_pairs),
        "gold_covered": covered,
    }

# ---------------- Row-level result label ----------------
def _row_result_from_verdicts(n_preds: int, verdicts: List[Tuple[str,str]]) -> str:
    """
    TP: all predictions are correct (all verdicts == 'TP' and n_preds>0)
    FP: some correct and some incorrect (any 'TP' and any not 'TP')
    FN: predictions are incorrect (no 'TP') -- also used when n_preds == 0
    """
    if n_preds == 0:
        return "FN"
    n_tp = sum(1 for _,v in verdicts if v == "TP")
    if n_tp == n_preds:
        return "TP"
    if n_tp > 0:
        return "FP"
    return "FN"

# -------------- Run evaluation on your dataframe --------------
def run_evaluation(test: pd.DataFrame, preds_col: str = "kg_final_preds"):
    rows = []
    # Per-pred micro tallies
    strict_pred_tot = {"tp":0, "fp":0, "fn":0, "preds":0, "rows":0, "rows_any_tp":0}
    relaxed_pred_tot = {"tp":0, "fp":0, "fn":0, "preds":0, "rows":0, "rows_any_tp":0}
    # Gold coverage for recall (micro)
    strict_gold_tot = {"covered":0, "total":0}
    relaxed_gold_tot = {"covered":0, "total":0}

    for idx, r in test.iterrows():
        # predictions
        preds = r.get(preds_col, [])
        if isinstance(preds, str):
            try:
                preds = ast.literal_eval(preds)
            except Exception:
                preds = [preds]
        preds = [p for p in preds if str(p).strip()]
        # gold
        gold_pairs = _get_gold_pairs(r)

        strict_res  = evaluate_prediction_list(preds, gold_pairs, mode="strict")
        relaxed_res = evaluate_prediction_list(preds, gold_pairs, mode="relaxed")

        # Row-level result labels
        strict_row_res  = _row_result_from_verdicts(strict_res["n_preds"], strict_res["verdicts"])
        relaxed_row_res = _row_result_from_verdicts(relaxed_res["n_preds"], relaxed_res["verdicts"])

        rows.append({
            "idx": idx,
            "sentence": r.get("sentence",""),
            "domain": r.get("domain",""),
            "baseline": r.get("baseline",""),
            "preds": preds,
            "gold_pairs": gold_pairs,

            "strict_verdicts": strict_res["verdicts"],
            "strict_tp": strict_res["tp"], "strict_fp": strict_res["fp"], "strict_fn": strict_res["fn"],
            "strict_any_tp": strict_res["any_tp"],
            "strict_row_result": strict_row_res,
            "strict_gold_covered": strict_res["gold_covered"],
            "strict_gold_total": strict_res["gold_total"],
            "strict_recall_row": (strict_res["gold_covered"]/strict_res["gold_total"]) if strict_res["gold_total"]>0 else float("nan"),
            "strict_partial_row": (strict_row_res=="FP"),

            "relaxed_verdicts": relaxed_res["verdicts"],
            "relaxed_tp": relaxed_res["tp"], "relaxed_fp": relaxed_res["fp"], "relaxed_fn": relaxed_res["fn"],
            "relaxed_any_tp": relaxed_res["any_tp"],
            "relaxed_row_result": relaxed_row_res,
            "relaxed_gold_covered": relaxed_res["gold_covered"],
            "relaxed_gold_total": relaxed_res["gold_total"],
            "relaxed_recall_row": (relaxed_res["gold_covered"]/relaxed_res["gold_total"]) if relaxed_res["gold_total"]>0 else float("nan"),
            "relaxed_partial_row": (relaxed_row_res=="FP"),
        })

        # Totals (per-pred)
        for mode_res, acc in ((strict_res, strict_pred_tot), (relaxed_res, relaxed_pred_tot)):
            acc["tp"] += mode_res["tp"]; acc["fp"] += mode_res["fp"]; acc["fn"] += mode_res["fn"]
            acc["preds"] += mode_res["n_preds"]; acc["rows"] += 1; acc["rows_any_tp"] += (1 if mode_res["any_tp"] else 0)
        strict_gold_tot["covered"] += strict_res["gold_covered"]; strict_gold_tot["total"] += strict_res["gold_total"]
        relaxed_gold_tot["covered"] += relaxed_res["gold_covered"]; relaxed_gold_tot["total"] += relaxed_res["gold_total"]

    eval_df = pd.DataFrame(rows)

    # Macro recall (row-averaged) ‚Äì ignore rows with no gold labels
    def _macro_recall(col_cov, col_tot):
        sub = eval_df[eval_df[col_tot] > 0]
        if sub.empty:
            return 0.0
        return (sub[col_cov] / sub[col_tot]).mean()

    # Micro metrics
    def _pred_metrics(tot, gold_tot):
        tp, fp, fn = tot["tp"], tot["fp"], tot["fn"]
        preds = max(tot["preds"], 0)
        rows = max(tot["rows"], 1)
        prec = tp / (tp + fp) if (tp + fp) > 0 else 0.0
        acc  = tp / preds if preds > 0 else 0.0               # per-pred accuracy
        f1   = (2*prec*acc)/(prec+acc) if (prec+acc) > 0 else 0.0
        hit  = tot["rows_any_tp"] / rows
        # micro recall from gold coverage
        mic_rec = gold_tot["covered"] / gold_tot["total"] if gold_tot["total"] > 0 else 0.0
        return {"TP":tp,"FP":fp,"FN":fn,"#preds":preds,"Rows":rows,"Any-TP@row":hit,
                "Precision":prec,"Acc(pred)":acc,"F1(pred)":f1,"Micro Recall(gold)":mic_rec}

    strict_metrics  = _pred_metrics(strict_pred_tot, strict_gold_tot)
    relaxed_metrics = _pred_metrics(relaxed_pred_tot, relaxed_gold_tot)
    strict_macro_rec  = _macro_recall("strict_gold_covered","strict_gold_total")
    relaxed_macro_rec = _macro_recall("relaxed_gold_covered","relaxed_gold_total")

    print("=== STRICT ===")
    for k,v in strict_metrics.items():
        print(f"{k:>18}: {v:.4f}" if isinstance(v, float) else f"{k:>18}: {v}")
    print(f"{'Macro Recall(rows)':>18}: {strict_macro_rec:.4f}")

    print("\n=== RELAXED (head-noun) ===")
    for k,v in relaxed_metrics.items():
        print(f"{k:>18}: {v:.4f}" if isinstance(v, float) else f"{k:>18}: {v}")
    print(f"{'Macro Recall(rows)':>18}: {relaxed_macro_rec:.4f}")

    # Per-predicted-aspect confusion summary
    def _collapse(verdicts_col):
        out = {}
        for lst in verdicts_col:
            for label, v in lst:
                key = str(label).lower().strip()
                if not key:
                    continue
                if key not in out:
                    out[key] = {"TP":0,"FP":0,"FN":0}
                out[key][v] += 1
        df = pd.DataFrame.from_dict(out, orient="index").sort_values(["TP","FP"], ascending=[False,True])
        df.index.name = "predicted_aspect"
        return df

    by_aspect_strict  = _collapse(eval_df["strict_verdicts"])
    by_aspect_relaxed = _collapse(eval_df["relaxed_verdicts"])

    return eval_df, by_aspect_strict, by_aspect_relaxed

# ---------------- Use it ----------------
eval_df, by_aspect_strict, by_aspect_relaxed = run_evaluation(test, preds_col="kg_final_preds")

# Examples of what you now have:
# eval_df[["sentence","preds","gold_pairs","strict_row_result","strict_recall_row","relaxed_row_result","relaxed_recall_row"]].head()
# by_aspect_strict.head(20)
# by_aspect_relaxed.head(20)
```

###### Evaluation #2: We return predictions as ENTITY#ASPECT pais:

```python

# ===========This cell considers either aspect or entity matching as correct.=====
# === Aspect ‚Üí ENTITY#ASPECT post-processing + strict/relaxed evaluation ===
# - DOES NOT change your KG inference; it only lifts aspect predictions to ENTITY#ASPECT via the KG edges.
# - Strict: TP if predicted entity OR predicted aspect exactly matches at least one gold pair's entity or aspect.
# - Relaxed: same, but aspects also match by head-noun (last token) or token-set inclusion.
# - Adds:
#     test["kg_final_pairs"]     -> ["FOOD#TASTE", ...]
#     test["row_result_strict"]  -> {"TP","FP","FN"}
#     test["row_result_relaxed"] -> {"TP","FP","FN"}
#   and prints micro/macro metrics for both modes.

import ast
import re
import pandas as pd
from collections import Counter, defaultdict

# ---------- helpers ----------
DOMAIN_MAP = {"restaurants":"restaurant","restaurant":"restaurant",
              "hotels":"hotel","hotel":"hotel",
              "laptops":"laptop","laptop":"laptop"}

def _norm_domain(d):
    if d is None: return None
    d = str(d).strip().lower()
    return DOMAIN_MAP.get(d, d)

# Parse gold categories that may be strings like "['FOOD#QUALITY','SERVICE#GENERAL']" or a single "FOOD#QUALITY"
def _parse_gold(val):
    if val is None: return []
    if isinstance(val, list):
        raw = val
    else:
        s = str(val).strip()
        try:
            v = ast.literal_eval(s)
            raw = v if isinstance(v, list) else [s]
        except Exception:
            raw = [s] if s else []
    pairs = []
    for x in raw:
        xs = str(x)
        if "#" in xs:
            a,b = xs.split("#",1)
            pairs.append((a.strip().lower(), b.strip().lower()))
        elif xs:
            # if only one token, treat as aspect-only gold with unknown entity
            pairs.append(("*", xs.strip().lower()))
    # normalize 'miscellaneous' to 'general'
    pairs = [(e, ("general" if a == "miscellaneous" else a)) for (e,a) in pairs]
    return pairs

# Token helpers
_ws = re.compile(r"\s+")
def _tokset(s: str) -> list:
    return re.findall(r"[a-z0-9]+", (s or "").lower())

def _head(s: str) -> str:
    toks = _tokset(s)
    return toks[-1] if toks else ""

def _same_strict(a: str, b: str) -> bool:
    return (a or "").lower() == (b or "").lower()

def _same_relaxed_aspect(a: str, b: str) -> bool:
    # exact OR head-noun equality OR token-set inclusion
    a = (a or "").lower(); b = (b or "").lower()
    if a == b: return True
    if _head(a) and _head(a) == _head(b): return True
    sa, sb = set(_tokset(a)), set(_tokset(b))
    if sa and (sa.issubset(sb) or sb.issubset(sa)): return True
    return False

# Prefer a single entity per (domain, aspect) without changing scoring
ENTITY_PREF = {
    "restaurant": ["food","service","location","ambience","restaurant","drinks","price","cleanliness","amenities","portion"],
    "hotel":      ["rooms","service","location","hotel","cleanliness","value","facilities","internet","parking","breakfast"],
    "laptop":     ["laptop","performance","battery","display","keyboard","trackpad","audio","connectivity","quality","storage","camera"],
}
def _entities_for_aspect(kg, domain: str, aspect: str):
    # First try KG edges aspect -> entity that belong to the same domain
    ents = []
    if hasattr(kg, "get_entities"):
        try:
            for e in kg.get_entities(aspect) or []:
                # Keep only entities that link to the domain (if the API is present)
                ok = True
                if hasattr(kg, "get_domains"):
                    ds = [d.lower() for d in kg.get_domains(e)]
                    ok = (_norm_domain(domain) in ds) if ds else True
                ents.append(str(e).strip())
        except Exception:
            pass
    # Dedup with simple normalization
    seen = set(); out = []
    for e in ents:
        en = e.lower().strip()
        if en not in seen:
            seen.add(en); out.append(e)
    return out

def _choose_entity(kg, domain: str, aspect: str):
    # Use KG-provided entities when available
    ents = _entities_for_aspect(kg, domain, aspect)
    if ents:
        # prefer by domain-specific priority if any priority matches
        prefs = [p.lower() for p in ENTITY_PREF.get(_norm_domain(domain), [])]
        ranked = sorted(ents, key=lambda e: (prefs.index(e.lower()) if e.lower() in prefs else 10_000, len(e)))
        return ranked[0]
    # fallback: top priority name for the domain if it exists, else domain name itself
    prefs = ENTITY_PREF.get(_norm_domain(domain), [])
    return prefs[0] if prefs else str(domain)

def _lift_to_pairs(aspects, domain, kg):
    pairs = []
    for a in (aspects or []):
        ent = _choose_entity(kg, domain, a)
        pairs.append(f"{str(ent).upper()}#{str(a).upper()}")
    return pairs

def _lower_pair(p):
    e,a = p.split("#",1)
    return (e.strip().lower(), a.strip().lower())

def _pred_match_vs_gold(pred_pair, gold_pairs, relaxed=False):
    """
    Returns True if EITHER entity or aspect of prediction matches
    at least one gold pair (strict or relaxed rules for aspect).
    """
    pe, pa = _lower_pair(pred_pair)
    for ge, ga in gold_pairs:
        # entity exact?
        if _same_strict(pe, ge):
            return True
        # aspect matches?
        if relaxed:
            if _same_relaxed_aspect(pa, ga):
                return True
        else:
            if _same_strict(pa, ga):
                return True
    return False

def _row_label(pred_pairs, gold_pairs, relaxed=False):
    """
    Row-level label:
      - 'TP' if every predicted pair matches (by the rule above)
      - 'FP' if at least one prediction matches, but not all
      - 'FN' if none match
    """
    if not pred_pairs:
        return "FN" if gold_pairs else "TP"
    hits = [ _pred_match_vs_gold(p, gold_pairs, relaxed=relaxed) for p in pred_pairs ]
    if all(hits):
        return "TP"
    elif any(hits):
        return "FP"
    else:
        return "FN"

def _eval_block(df, pred_col_pairs, mode_name="STRICT", relaxed=False):
    # Per-pred counters (micro)
    TP = FP = FN = 0
    rows = len(df)
    any_tp_rows = 0
    per_row = []

    for _, r in df.iterrows():
        gold = _parse_gold(r.get("category"))
        preds = r.get(pred_col_pairs) or []
        # count per-pred
        for p in preds:
            if _pred_match_vs_gold(p, gold, relaxed=relaxed):
                TP += 1
            else:
                FP += 1
        # per-gold items that had no matching prediction on either part
        matched_gold = [False]*len(gold)
        for i, g in enumerate(gold):
            # a gold is considered "recalled" if ANY prediction matches it on either part
            if any(_same_strict(_lower_pair(p)[0], g[0]) or
                   (_same_relaxed_aspect(_lower_pair(p)[1], g[1]) if relaxed else _same_strict(_lower_pair(p)[1], g[1]))
                   for p in preds):
                matched_gold[i] = True
        FN += matched_gold.count(False)

        # row label
        row_lab = _row_label(preds, gold, relaxed=relaxed)
        per_row.append(row_lab)
        if any(_pred_match_vs_gold(p, gold, relaxed=relaxed) for p in preds):
            any_tp_rows += 1

    # metrics
    num_preds = TP + FP
    prec = TP / num_preds if num_preds else 0.0
    acc_pred = TP / rows if rows else 0.0
    f1 = (2*prec*(TP/(TP+FN) if (TP+FN) else 0.0)) / (prec + (TP/(TP+FN) if (TP+FN) else 0.0)) if (prec and (TP+FN)) else 0.0

    # Micro recall over gold items:
    # total gold = TP + FN by construction (each recalled gold contributes a TP; non-recalled contributes FN)
    micro_recall = TP / (TP + FN) if (TP + FN) else 0.0

    # Macro recall (rows): average of per-row hit-or-miss (any TP in row)
    macro_recall_rows = any_tp_rows / rows if rows else 0.0

    print(f"=== {mode_name} ===")
    print(f"                TP: {TP}")
    print(f"                FP: {FP}")
    print(f"                FN: {FN}")
    print(f"            #preds: {num_preds}")
    print(f"              Rows: {rows}")
    print(f"        Any-TP@row: {any_tp_rows/rows if rows else 0.0:.4f}")
    print(f"         Precision: {prec:.4f}")
    print(f"         Acc(pred): {acc_pred:.4f}")
    print(f"          F1(pred): {f1:.4f}")
    print(f"Micro Recall(gold): {micro_recall:.4f}")
    print(f"Macro Recall(rows): {macro_recall_rows:.4f}")

    return per_row

# ---------- 1) Lift aspect predictions to ENTITY#ASPECT ----------
test = test.copy()
test["kg_final_pairs"] = test.apply(
    lambda r: _lift_to_pairs(r.get("kg_final_preds") or [], _norm_domain(r.get("domain")), kg), axis=1
)

# ---------- 2) Evaluate (STRICT & RELAXED) and tag rows ----------
test["row_result_strict"]  = _eval_block(test, "kg_final_pairs", mode_name="STRICT",  relaxed=False)
test["row_result_relaxed"] = _eval_block(test, "kg_final_pairs", mode_name="RELAXED", relaxed=True)

# ---------- 3) (Optional) Quick error analysis table ----------
def explain_row(idx):
    r = test.loc[idx]
    return {
        "sentence": r.get("sentence"),
        "domain": r.get("domain"),
        "pred_pairs": r.get("kg_final_pairs"),
        "gold_pairs": _parse_gold(r.get("category")),
        "strict_row_result": r.get("row_result_strict"),
        "relaxed_row_result": r.get("row_result_relaxed"),
    }

# Example: show a few FP rows for inspection
# fp_rows = test.index[test["row_result_strict"]=="FP"][:5].tolist()
# [explain_row(i) for i in fp_rows]
# test
```

```python
# === Aspect ‚Üí ENTITY#ASPECT post-processing + strict/relaxed evaluation ===
# - DOES NOT change your KG inference; it only lifts aspect predictions to ENTITY#ASPECT via the KG edges.
# - Normalization: lowercase + simple English singularization (entities & aspects).
# - Matching is ONE-TO-ONE (each prediction can match at most one gold, and vice versa).
# - Strict: a pred matches a gold if EITHER entity matches OR aspect matches (after normalization).
# - Relaxed: same, but aspect match also allows head-noun equality or token-set inclusion.
# - Adds:
#     test["kg_final_pairs"]     -> ["FOOD#TASTE", ...]
#     test["row_result_strict"]  -> {"TP","FP","FN"}
#     test["row_result_relaxed"] -> {"TP","FP","FN"}
#   and prints micro/macro metrics for both modes.

import ast
import re
import pandas as pd
from collections import defaultdict

# ---------- helpers ----------
DOMAIN_MAP = {"restaurants":"restaurant","restaurant":"restaurant",
              "hotels":"hotel","hotel":"hotel",
              "laptops":"laptop","laptop":"laptop"}

def _norm_domain(d):
    if d is None: return None
    d = str(d).strip().lower()
    return DOMAIN_MAP.get(d, d)

# ---------------- Normalization: lowercase + singularization ----------------
# minimal irregulars + common ABSA nouns
_IRREGULAR_SINGULAR = {
    "rooms": "room",
    "men": "man", "women": "woman", "children": "child", "people": "person",
    "facilities": "facility", "amenities": "amenity", "services": "service",
    "batteries": "battery", "dishes": "dish", "buses": "bus", "watches": "watch",
    "indices": "index", "matrices": "matrix",
}

def _singularize_token(tok: str) -> str:
    t = tok.lower().strip()
    if t in _IRREGULAR_SINGULAR:
        return _IRREGULAR_SINGULAR[t]
    # simple rules (order matters)
    if re.search(r'[^aeiou]ies$', t):          # batteries -> battery
        return re.sub(r'ies$', 'y', t)
    if re.search(r'(sses|zzes|ches|shes|xes)$', t):  # classes -> class, watches -> watch
        return t[:-2]  # drop 'es'
    if t.endswith('s') and not t.endswith('ss'):
        return t[:-1]
    return t

_token_re = re.compile(r"[a-z0-9]+")
def _tokset(s: str) -> list:
    return [_singularize_token(x) for x in _token_re.findall((s or "").lower())]

def _head(s: str) -> str:
    toks = _tokset(s)
    return toks[-1] if toks else ""

def _norm_field(s: str) -> str:
    # normalize a potentially multi-token field (e.g., "Bed Comfort" -> "bed comfort")
    toks = _tokset(s)
    return " ".join(toks)

# ---------------- Gold parsing (normalized) ----------------
# Accepts strings like "['FOOD#QUALITY','SERVICE#GENERAL']" or single "FOOD#QUALITY"
def _parse_gold(val):
    if val is None:
        return []
    if isinstance(val, list):
        raw = val
    else:
        s = str(val).strip()
        try:
            v = ast.literal_eval(s)
            raw = v if isinstance(v, list) else [s]
        except Exception:
            raw = [s] if s else []
    pairs = []
    for x in raw:
        xs = str(x)
        if "#" in xs:
            e, a = xs.split("#", 1)
            e = _norm_field(e)
            a = _norm_field(a)
            a = "general" if a == "miscellaneous" else a
            pairs.append((e, a))
        elif xs:
            # aspect-only gold with unknown entity
            pairs.append(("*", _norm_field(xs)))
    return pairs

# ---------------- Lifting aspects to ENTITY#ASPECT (unchanged policy) ----------------
ENTITY_PREF = {
    "restaurant": ["food","service","location","ambience","restaurant","drinks","price","cleanliness","amenities","portion"],
    "hotel":      ["rooms","service","location","hotel","cleanliness","value","facilities","internet","parking","breakfast"],
    "laptop":     ["laptop","performance","battery","display","keyboard","trackpad","audio","connectivity","quality","storage","camera"],
}

def _entities_for_aspect(kg, domain: str, aspect: str):
    ents = []
    if hasattr(kg, "get_entities"):
        try:
            for e in kg.get_entities(aspect) or []:
                ok = True
                if hasattr(kg, "get_domains"):
                    ds = [d.lower() for d in kg.get_domains(e)]
                    ok = (_norm_domain(domain) in ds) if ds else True
                if ok:
                    ents.append(str(e).strip())
        except Exception:
            pass
    # dedup keep order
    seen = set(); out = []
    for e in ents:
        en = e.lower().strip()
        if en not in seen:
            seen.add(en); out.append(e)
    return out

def _choose_entity(kg, domain: str, aspect: str):
    ents = _entities_for_aspect(kg, domain, aspect)
    if ents:
        prefs = [p.lower() for p in ENTITY_PREF.get(_norm_domain(domain), [])]
        ranked = sorted(ents, key=lambda e: (prefs.index(e.lower()) if e.lower() in prefs else 10_000, len(e)))
        return ranked[0]
    prefs = ENTITY_PREF.get(_norm_domain(domain), [])
    return prefs[0] if prefs else str(domain)

def _lift_to_pairs(aspects, domain, kg):
    pairs = []
    for a in (aspects or []):
        ent = _choose_entity(kg, domain, a)
        pairs.append(f"{str(ent).upper()}#{str(a).upper()}")
    return pairs

# ---------------- Pair normalization + predicates ----------------
def _lower_pair(p):
    e, a = p.split("#", 1)
    # normalize both halves (lower + singularize)
    return (_norm_field(e), _norm_field(a))

def _same_strict(a: str, b: str) -> bool:
    return _norm_field(a) == _norm_field(b)

def _same_relaxed_aspect(a: str, b: str) -> bool:
    # exact OR head-noun equality OR token-set inclusion ‚Äî all after singularization
    a = _norm_field(a); b = _norm_field(b)
    if a == b:
        return True
    if _head(a) and _head(a) == _head(b):
        return True
    sa, sb = set(a.split()), set(b.split())
    if sa and (sa.issubset(sb) or sb.issubset(sa)):
        return True
    return False

# def _compatible(pred_pair_norm, gold_pair, relaxed=False):
#     """
#     Compatibility rule for a single prediction vs a single gold:
#     - entity OR aspect match is sufficient (entities/aspects normalized to singular lowercase)
#     - entity '*' in gold is treated as unknown ‚Üí cannot be used for entity equality
#     """
#     pe, pa = pred_pair_norm
#     ge, ga = gold_pair
#     if ge != "*" and pe == ge:
#         return True
#     return _same_relaxed_aspect(pa, ga) if relaxed else (pa == ga)

def _compatible(pred_pair_norm, gold_pair, relaxed=False):
    pe, pa = pred_pair_norm
    ge, ga = gold_pair

    # 1) aspect-driven match
    if relaxed:
        if _same_relaxed_aspect(pa, ga):
            return True
    else:
        if pa == ga:
            return True

    # 2) special-case: allow entity-only match when either side is 'general'
    if ge != "*" and pe == ge and ("general" in (pa, ga)):
        return True

    return False
# ---------------- ONE-TO-ONE evaluation block ----------------
def _eval_block(df, pred_col_pairs, mode_name="STRICT", relaxed=False):
    rows = len(df)
    TP = FP = FN = 0
    any_tp_rows = 0
    per_row_labels = []

    def _build_adj(preds_norm, gold_pairs):
        # adjacency: left = preds indices, right = gold indices
        adj = [[] for _ in range(len(preds_norm))]
        for i, p in enumerate(preds_norm):
            for j, g in enumerate(gold_pairs):
                if _compatible(p, g, relaxed=relaxed):
                    adj[i].append(j)
        return adj

    def _max_bipartite_matching(adj, n_left, n_right):
        # Kuhn's algorithm (DFS-based augmenting paths). Fine for small row sizes.
        matchR = [-1] * n_right  # right->left
        def dfs(u, seen):
            for v in adj[u]:
                if not seen[v]:
                    seen[v] = True
                    if matchR[v] == -1 or dfs(matchR[v], seen):
                        matchR[v] = u
                        return True
            return False
        for u in range(n_left):
            seen = [False] * n_right
            dfs(u, seen)
        # Build matched pairs list from matchR
        pairs = []
        for v, u in enumerate(matchR):
            if u != -1:
                pairs.append((u, v))  # (pred_idx, gold_idx)
        return pairs

    for _, r in df.iterrows():
        gold_pairs = _parse_gold(r.get("category"))
        # de-duplicate identical gold pairs within the row
        seen = set()
        dedup_gold = []
        for g in gold_pairs:
            if g not in seen:
                seen.add(g)
                dedup_gold.append(g)
        gold_pairs = dedup_gold

        pred_pairs = r.get(pred_col_pairs) or []

        preds_norm = [_lower_pair(p) for p in pred_pairs]
        adj = _build_adj(preds_norm, gold_pairs)
        pairs = _max_bipartite_matching(adj, len(preds_norm), len(gold_pairs))
        matched_count = len(pairs)

        # Update counters
        TP += matched_count
        FP += max(0, len(preds_norm) - matched_count)
        FN += max(0, len(gold_pairs) - matched_count)

        row_has_tp = matched_count > 0
        # Row label:
        if not pred_pairs:
            row_lab = "FN" if gold_pairs else "TP"
        else:
            if matched_count == len(preds_norm):
                row_lab = "TP"
            elif row_has_tp:
                row_lab = "FP"
            else:
                row_lab = "FN"
        per_row_labels.append(row_lab)
        if row_has_tp:
            any_tp_rows += 1

    num_preds = TP + FP
    prec = TP / num_preds if num_preds else 0.0
    rec  = TP / (TP + FN) if (TP + FN) else 0.0
    f1   = (2 * prec * rec) / (prec + rec) if (prec and rec) else 0.0
    acc_pred = TP / rows if rows else 0.0
    macro_recall_rows = any_tp_rows / rows if rows else 0.0

    print(f"=== {mode_name} ===")
    print(f"                TP: {TP}")
    print(f"                FP: {FP}")
    print(f"                FN: {FN}")
    print(f"            #preds: {num_preds}")
    print(f"              Rows: {rows}")
    print(f"        Any-TP@row: {macro_recall_rows:.4f}")
    print(f"         Precision: {prec:.4f}")
    print(f"         Acc(pred): {acc_pred:.4f}")
    print(f"          F1(pred): {f1:.4f}")
    print(f"Micro Recall(gold): {rec:.4f}")
    print(f"Macro Recall(rows): {macro_recall_rows:.4f}")

    return per_row_labels


# ---------- 1) Lift aspect predictions to ENTITY#ASPECT ----------
test = test.copy()
test["kg_final_pairs"] = test.apply(
    lambda r: _lift_to_pairs(r.get("kg_final_preds") or [], _norm_domain(r.get("domain")), kg), axis=1
)

# ---------- 2) Evaluate (STRICT & RELAXED) and tag rows ----------
test["row_result_strict"]  = _eval_block(test, "kg_final_pairs", mode_name="STRICT",  relaxed=False)
test["row_result_relaxed"] = _eval_block(test, "kg_final_pairs", mode_name="RELAXED", relaxed=True)

# ---------- 3) (Optional) Quick error analysis helper ----------
def explain_row(idx):
    r = test.loc[idx]
    return {
        "sentence": r.get("sentence"),
        "domain": r.get("domain"),
        "pred_pairs": r.get("kg_final_pairs"),
        "gold_pairs": _parse_gold(r.get("category")),
        "strict_row_result": r.get("row_result_strict"),
        "relaxed_row_result": r.get("row_result_relaxed"),
    }

# Example for manual inspection:
# fp_rows = test.index[test["row_result_strict"]=="FP"][:5].tolist()
# [explain_row(i) for i in fp_rows]
```

```python
# ===================== Clue ‚Üí Aspect Confusion Analysis (robust) =====================
# Finds cases that are TP under "Entity OR Aspect" but NOT under "Aspect-Driven"
# (i.e., lenient policy credits, strict policy does not), then summarizes by clue word(s).
#
# Changes vs earlier version:
#  - Keeps rows even when clues are missing (uses "<none>")
#  - Fallbacks for clues: (1) exact aspect in all_hits ‚Üí (2) any all_hits terms ‚Üí (3) light lexical harvest
#  - Writes combined STRICT+RELAXED CSV & LaTeX for the paper

import re
import pandas as pd
from collections import defaultdict
from pathlib import Path

# ---------------- Policy definitions (pair-level, normalized fields) ----------------
def _compat_entity_or_aspect(pred_pair_norm, gold_pair, relaxed=False):
    pe, pa = pred_pair_norm
    ge, ga = gold_pair
    if ge != "*" and pe == ge:
        return True
    return _same_relaxed_aspect(pa, ga) if relaxed else (pa == ga)

def _compat_aspect_driven(pred_pair_norm, gold_pair, relaxed=False, allow_general_entity=True):
    pe, pa = pred_pair_norm
    ge, ga = gold_pair
    # aspect must match
    if relaxed:
        if _same_relaxed_aspect(pa, ga):
            return True
    else:
        if pa == ga:
            return True
    # special-case: allow entity-only when either aspect is 'general'
    if allow_general_entity and ge != "*" and pe == ge and ("general" in (pa, ga)):
        return True
    return False

# ---------------- Clue extraction helpers ----------------
# lightweight sentiment/ABSA lexicon for fallback harvest (extend as needed)
_FALLBACK_ADJ = {
    # taste / quality
    "delicious","tasty","yummy","savory","flavorful","fresh","stale","bland","soggy","crispy","juicy",
    "amazing","great","good","bad","awful","terrible","meh","disappointing","average","mediocre","fantastic",
    # style/menu
    "creative","inventive","variety","options","fusion","traditional","authentic",
    # price/value
    "overpriced","expensive","cheap","inexpensive","worth","value","affordable",
    # service/ambience/general
    "friendly","rude","slow","fast","cozy","noisy","quiet","clean","dirty","recommended","highly","perfect"
}

_ws = re.compile(r"[a-z0-9']+")
def _simple_tokens(text: str):
    return [t.lower() for t in _ws.findall(str(text or ""))]

def _harvest_clues_from_sentence(sentence: str, cap: int = 5):
    toks = _simple_tokens(sentence)
    hits = []
    seen = set()
    for t in toks:
        if t in _FALLBACK_ADJ and t not in seen:
            seen.add(t); hits.append(t)
        if len(hits) >= cap:
            break
    return hits

def _clues_for_pred_aspect(row_idx, pred_aspect_norm, KG_debug, sentence):
    """
    Try to get human-readable clue terms that supported the predicted aspect.
    Priority:
      1) KG_debug['all_hits'][pred_aspect_norm]
      2) Any terms from KG_debug['all_hits'] (best-effort)
      3) Light lexical harvest from sentence (_FALLBACK_ADJ)
    Returns (clues, source_tag)
    """
    all_hits = None
    try:
        all_hits = KG_debug.loc[row_idx, "all_hits"]
    except Exception:
        pass

    # 1) exact aspect hits
    if isinstance(all_hits, dict):
        exact = []
        for a_key, terms in all_hits.items():
            if _norm_field(str(a_key)) == pred_aspect_norm:
                for t in (terms or []):
                    tt = str(t).strip()
                    if tt:
                        exact.append(tt.lower())
        if exact:
            # dedup, keep top few
            uniq = []
            seen = set()
            for x in exact:
                if x not in seen:
                    seen.add(x); uniq.append(x)
            return uniq[:5], "all_hits:pred_aspect"

        # 2) any hits at all (fallback)
        any_terms = []
        for terms in all_hits.values():
            for t in (terms or []):
                tt = str(t).strip().lower()
                if tt:
                    any_terms.append(tt)
        if any_terms:
            uniq = []
            seen = set()
            for x in any_terms:
                if x not in seen:
                    seen.add(x); uniq.append(x)
            return uniq[:5], "all_hits:any"

    # 3) lexical harvest
    harvest = _harvest_clues_from_sentence(sentence, cap=5)
    if harvest:
        return harvest, "sentence:lexical"

    # nothing found
    return ["<none>"], "none"

# ---------------- Scan dataset and collect "policy gap" confusions ----------------
def collect_policy_gap_confusions(
    df_test: pd.DataFrame,
    KG_debug: pd.DataFrame,
    relaxed=False,
    allow_general_entity=True,
):
    records = []
    for idx, r in df_test.iterrows():
        sentence = r.get("sentence")
        pred_pairs = r.get("kg_final_pairs") or []
        gold_pairs = _parse_gold(r.get("category"))

        preds_norm = [_lower_pair(p) for p in pred_pairs]
        gold_norm  = [(_norm_field(e), _norm_field(a)) for (e, a) in gold_pairs]
        if not preds_norm or not gold_norm:
            continue

        for p_e, p_a in preds_norm:
            eo_matches = [g for g in gold_norm if _compat_entity_or_aspect((p_e, p_a), g, relaxed=relaxed)]
            if not eo_matches:
                continue
            ad_matches = [g for g in gold_norm if _compat_aspect_driven((p_e, p_a), g, relaxed=relaxed, allow_general_entity=allow_general_entity)]
            if not ad_matches:
                clues, src = _clues_for_pred_aspect(idx, p_a, KG_debug, sentence)
                g_e, g_a = eo_matches[0]  # representative gold it matched under lenient policy
                records.append({
                    "row_index": idx,
                    "sentence": sentence,
                    "pred_entity": p_e,
                    "pred_aspect": p_a,
                    "gold_entity_sample": g_e,
                    "gold_aspect_sample": g_a,
                    "clues": clues,
                    "clue_source": src,
                    "num_gold_matches_EoA": len(eo_matches),
                    "num_gold_matches_AD":  len(ad_matches),
                    "policy": "EoA_only",
                })
    return pd.DataFrame.from_records(records)

# ---------------- Build aggregated clue‚Üíaspect tables ----------------
def build_clue_confusion_tables(confusions: pd.DataFrame, top_k=30, keep_none=True):
    """
    Outputs:
      - by_tuple: (clue, pred_aspect, gold_aspect) with counts
      - by_clue:  (clue) total counts
      - samples:  one example sentence per tuple
    If keep_none=True, keeps "<none>" clues so tables are non-empty even without trace terms.
    """
    if confusions.empty:
        return (
            pd.DataFrame(columns=["clue","pred_aspect","gold_aspect","count"]),
            pd.DataFrame(columns=["clue","count"]),
            pd.DataFrame(columns=["clue","pred_aspect","gold_aspect","example_sentence"])
        )

    cf = confusions.copy().explode("clues")
    cf["clue"] = cf["clues"].fillna("").astype(str).str.lower().str.strip()
    if not keep_none:
        cf = cf[cf["clue"] != ""]
        cf = cf[cf["clue"] != "<none>"]
    # aggregated by (clue, pred_aspect, gold_aspect)
    by_tuple = (
        cf.groupby(["clue","pred_aspect","gold_aspect_sample"])
          .size().reset_index(name="count")
          .sort_values(["count","clue"], ascending=[False, True])
          .head(top_k)
          .rename(columns={"gold_aspect_sample": "gold_aspect"})
    )
    # totals per clue
    by_clue = (
        cf.groupby("clue")
          .size().reset_index(name="count")
          .sort_values("count", ascending=False)
          .head(top_k)
    )
    # sample sentences per (clue, pred_aspect, gold_aspect)
    samples = (
        cf.groupby(["clue","pred_aspect","gold_aspect_sample"])
          .agg(example_sentence=("sentence", "first"))
          .reset_index()
          .rename(columns={"gold_aspect_sample": "gold_aspect"})
    )
    return by_tuple, by_clue, samples

# ---------------- Run the analysis & emit artifacts ----------------
ALLOW_GENERAL_ENTITY = True
TOP_K = 40
KEEP_NONE = True  # keep "<none>" so tables never come back empty

confusions_strict = collect_policy_gap_confusions(test, KG_debug, relaxed=False, allow_general_entity=ALLOW_GENERAL_ENTITY)
confusions_relaxed = collect_policy_gap_confusions(test, KG_debug, relaxed=True,  allow_general_entity=ALLOW_GENERAL_ENTITY)

by_tuple_s, by_clue_s, samples_s = build_clue_confusion_tables(confusions_strict, top_k=TOP_K, keep_none=KEEP_NONE)
by_tuple_r, by_clue_r, samples_r = build_clue_confusion_tables(confusions_relaxed, top_k=TOP_K, keep_none=KEEP_NONE)

print("=== Policy-gap confusions (STRICT) ===")
print(f"Rows impacted: {confusions_strict['row_index'].nunique() if not confusions_strict.empty else 0}")
print(confusions_strict.head(12))
print("\nTop (clue, pred_aspect, gold_aspect) tuples:")
print(by_tuple_s.head(20))

print("\n=== Policy-gap confusions (RELAXED) ===")
print(f"Rows impacted: {confusions_relaxed['row_index'].nunique() if not confusions_relaxed.empty else 0}")
print(confusions_relaxed.head(12))
print("\nTop (clue, pred_aspect, gold_aspect) tuples:")
print(by_tuple_r.head(20))

# Save per-policy CSVs
by_tuple_s.to_csv("clue_confusions_strict_by_tuple.csv", index=False)
by_clue_s.to_csv("clue_confusions_strict_by_clue.csv", index=False)
samples_s.to_csv("clue_confusions_strict_samples.csv", index=False)

by_tuple_r.to_csv("clue_confusions_relaxed_by_tuple.csv", index=False)
by_clue_r.to_csv("clue_confusions_relaxed_by_clue.csv", index=False)
samples_r.to_csv("clue_confusions_relaxed_samples.csv", index=False)

# --------- Build combined STRICT+RELAXED table + LaTeX ---------
def combine_sr(strict_df, relaxed_df, topN=None):
    s = strict_df.rename(columns={"count":"count_strict"}).copy()
    r = relaxed_df.rename(columns={"count":"count_relaxed"}).copy()
    merged = pd.merge(s, r, how="outer", on=["clue","pred_aspect","gold_aspect"]).fillna(0)
    merged["count_strict"]  = merged["count_strict"].astype(int)
    merged["count_relaxed"] = merged["count_relaxed"].astype(int)
    merged["delta"] = merged["count_relaxed"] - merged["count_strict"]
    merged["total"] = merged["count_strict"] + merged["count_relaxed"]
    merged = merged.sort_values(["total","count_relaxed","count_strict","clue"],
                                ascending=[False, False, False, True])
    if topN is not None:
        merged = merged.head(topN)
    return merged

combined_full = combine_sr(by_tuple_s, by_tuple_r, topN=None)
combined_top  = combine_sr(by_tuple_s, by_tuple_r, topN=25)

# LaTeX
def to_latex_table(df: pd.DataFrame) -> str:
    if df.empty:
        return "% No data"
    cols = ["clue","pred_aspect","gold_aspect","count_strict","count_relaxed","delta","total"]
    df2 = df[cols].copy().rename(columns={
        "clue":"Clue","pred_aspect":"Pred.~Aspect","gold_aspect":"Gold~Aspect",
        "count_strict":"\\#(Strict)","count_relaxed":"\\#(Relaxed)","delta":"$\\Delta$","total":"Total"
    })
    def esc(s): return str(s).replace("_","\\_")
    for c in ["Clue","Pred.~Aspect","Gold~Aspect"]:
        df2[c] = df2[c].apply(esc)
    return df2.to_latex(index=False, escape=False, column_format="lllr r r r".replace(" ",""))

latex_snippet = (
    "\\begin{table}[t]\n\\centering\n\\small\n"
    "\\caption{Top clue$\\rightarrow$aspect confusions credited under \\textit{Entity-or-Aspect} but not under \\textit{Aspect-Driven}. "
    "Counts shown under Strict and Relaxed aspect matching, their difference $\\Delta$, and the total.}\n"
    "\\label{tab:clue-confusions}\n" + to_latex_table(combined_top) +
    "\n\\vspace{-0.5em}\n\\end{table}\n"
)

# Write files
Path("clue_confusion_combined_full.csv").write_text(combined_full.to_csv(index=False), encoding="utf-8")
Path("clue_confusion_combined.tex").write_text(latex_snippet, encoding="utf-8")

# If you're in Colab, uncomment to download:
from google.colab import files
for p in ["clue_confusions_strict_by_tuple.csv","clue_confusions_strict_by_clue.csv","clue_confusions_strict_samples.csv",
          "clue_confusions_relaxed_by_tuple.csv","clue_confusions_relaxed_by_clue.csv","clue_confusions_relaxed_samples.csv",
          "clue_confusion_combined_full.csv"]:
    files.download(p)
```

```python
import pandas as pd
from tqdm import tqdm
import time, gc
import torch

# ================================================================
# STEP 1: Iterative Residual Decoding
# ================================================================
from tqdm import tqdm
tqdm.pandas()

"""
Run the full inference pipeline on a test DataFrame.
Saves intermediate results to CSV after each component.

Parameters
----------
test : pd.DataFrame
    Must contain columns ['sentence', 'aspect', 'paraphrases']
kg : KnowledgeGraph
    Knowledge graph instance
domain_embs : dict
    Precomputed domain embeddings
save_path : str
    File path to save CSV after each stage
max_rounds : int
    Max rounds for iterative residual decoding
top_k : int
    Keep top-k aspects in paraphrase step (None = keep all)
sim_threshold : float
    Similarity threshold for paraphrase filtering (None = keep all)
domain_conf_threshold : float
    Confidence threshold for domain inference in KG

Returns
-------
pd.DataFrame
    Test DataFrame with new prediction columns

Columns produced:
- explicit_iterative
- explicit_paraphrases
- implicit_kg
- combined_aspects
"""

# ================================================================
# STEP 1: Finetuned model only Extraction
# ================================================================

def run_baseline(test, save_path="test_baseline.csv"):
    print("\n[STEP 0] Baseline...")
    t0 = time.time()

    test["baseline"] = test["sentence"].progress_apply(
        lambda s: extract_aspect(s)
    )

    elapsed = time.time() - t0
    test.to_csv(save_path, index=False, encoding="utf-8")
    print(f"[SAVED] Baseline results -> {save_path}")
    print(f"[TIME] Baseline  inference took {elapsed:.2f} sec")
    # --- Memory cleanup block ---
    del t0
    gc.collect()
    torch.cuda.empty_cache()
    # ----------------------------
    return test



# ================================================================
# STEP 2: Iterative Generation Extraction
# ================================================================

def run_iterative_only(test, save_path="test_iterative.csv", max_rounds=3):
    print("\n[STEP 1] Iterative Residual Decoding...")
    t0 = time.time()

    test["explicit_iterative"] = test["sentence"].progress_apply(
        lambda s: iterative_residual_decoding(s, max_rounds=max_rounds)
    )

    elapsed = time.time() - t0
    test.to_csv(save_path, index=False, encoding="utf-8")
    print(f"[SAVED] Iterative results -> {save_path}")
    print(f"[TIME] Iterative decoding took {elapsed:.2f} sec")

    # --- Memory cleanup block ---
    del t0
    gc.collect()
    torch.cuda.empty_cache()
    # ----------------------------
    return test

def run_iterative_only2(test, save_path="test_iterative.csv", max_rounds=3):
    print("\n[STEP 1] Iterative Residual Decoding...")
    t0 = time.time()

    test["explicit_iterative2"] = test["sentence"].progress_apply(
        lambda s: iterative_residual_decoding2(s, max_rounds=max_rounds)
    )

    elapsed = time.time() - t0
    test.to_csv(save_path, index=False, encoding="utf-8")
    print(f"[SAVED] Iterative results -> {save_path}")
    print(f"[TIME] Iterative decoding took {elapsed:.2f} sec")

    # --- Memory cleanup block ---
    del t0
    gc.collect()
    torch.cuda.empty_cache()
    # ----------------------------
    return test


# ================================================================
# STEP 3: Paraphrase-Augmented Extraction
# ================================================================
# def run_paraphrase_only(test, save_path="test_paraphrase.csv",
#                         paraphrase_top_k=None, paraphrase_sim_threshold=None):
#     print("\n[STEP 2] Paraphrase-Augmented Extraction...")
#     t0 = time.time()

#     test["cleaned_explicit_paraphrases"] = test.progress_apply(
#         lambda r: extract_aspects_with_paraphrases(
#             [r["sentence"]] + (r["paraphrases"] if isinstance(r["paraphrases"], list) else []),
#             top_k=2,
#             similarity_threshold=0.58, #0.65,
#             min_consensus=1,
#             include_original_always=True,
#             verbose=False,
#         ),
#         axis=1
#     )


#     elapsed = time.time() - t0
#     test.to_csv(save_path, index=False, encoding="utf-8")
#     print(f"[SAVED] Paraphrase results -> {save_path}")
#     print(f"[TIME] Paraphrase extraction took {elapsed:.2f} sec")

#     # --- Memory cleanup block ---
#     del t0
#     gc.collect()
#     torch.cuda.empty_cache()
#     # ----------------------------
#     return test

# ================================================================
# STEP 3: Paraphrase-Augmented Extraction (Edited)
# ================================================================


def run_paraphrase_only(
    test,
    input_path="/content/drive/MyDrive/colab_files/test_14_2.csv",
    save_path=None,
):
    """
    Runs paraphrase-aided extraction once per row (orig + 10 paraphrases),
    stores cached predictions (pred_orig, pred_para_list) and the k-sweep outputs,
    and saves to Parquet (recommended for nested lists).

    If save_path is None, it saves next to input_path with suffix '_with_para_cache.parquet'.
    """

    print("\n[STEP 2] Paraphrase-Augmented Extraction (k-sweep cached inference)...")

    # ---- Decide save path in same folder as input ----
    if save_path is None:
        base_dir = os.path.dirname(input_path)
        base_name = os.path.splitext(os.path.basename(input_path))[0]
        save_path = os.path.join(base_dir, f"{base_name}_English_with_para_cache.parquet")

    t0 = time.time()

    def _apply_row(r):
        paras = r["paraphrases"] if isinstance(r.get("paraphrases", None), list) else []
        sentences = [r["sentence"]] + paras  # expected 11 (1 + 10)

        df_out = extract_aspects_with_paraphrases(
            sentences,
            top_k=2,
            similarity_threshold=0.58,
            include_original_always=True,
            jaccard_explicitness=0.50,
            verbose=False,
        )

        # df_out is 1-row DataFrame
        return df_out.iloc[0].to_dict()

    results = test.progress_apply(_apply_row, axis=1)

    # Expand dict results into columns
    results_df = pd.DataFrame(list(results))

    # These are the columns we expect from extract_aspects_with_paraphrases()
    cols_to_add = [
        "pred_orig", "pred_para_list",
        "k_0Preds", "k_2Preds", "k_4Preds", "k_6Preds", "k_8Preds", "k_10Preds"
    ]

    for col in cols_to_add:
        if col in results_df.columns:
            test[col] = results_df[col].values
        else:
            # if missing for any reason, create empty column to avoid crash
            test[col] = [[] for _ in range(len(test))]

    elapsed = time.time() - t0

    # Save as Parquet (preserves nested lists)
    test.to_parquet(save_path, index=False)
    print(f"[SAVED] Paraphrase cached results -> {save_path}")
    print(f"[TIME] Paraphrase extraction took {elapsed:.2f} sec")

    # --- Memory cleanup block ---
    del t0, results, results_df
    gc.collect()
    torch.cuda.empty_cache()
    # ----------------------------

    return test


# ================================================================
# STEP 4: Knowledge Graph Inference
# ================================================================
def run_kg_only(test, kg, domain_embs, save_path="test_kg.csv", domain_conf_threshold=0.15):
    print("\n[STEP 3] Knowledge Graph Inference...")
    t0 = time.time()

    test["implicit_kg"] = test["sentence"].progress_apply(
        lambda s: infer_implicit_from_kg_best(
            s, kg, domain_embs, domain_conf_threshold=domain_conf_threshold
        )
    )



    elapsed = time.time() - t0
    test.to_csv(save_path, index=False, encoding="utf-8")
    print(f"[SAVED] KG results -> {save_path}")
    print(f"[TIME] KG inference took {elapsed:.2f} sec")

    # --- Memory cleanup block ---
    del t0
    gc.collect()
    torch.cuda.empty_cache()
    # ----------------------------
    return test


# ================================================================
# STEP 4: Knowledge Graph Inference
# ================================================================
def run_kg_only2(test, kg, domain_embs, save_path="test_kg.csv", domain_conf_threshold=0.15):
    print("\n[STEP 3] Knowledge Graph Inference...")
    t0 = time.time()

    test["implicit_kg2"] = test["sentence"].progress_apply(
        lambda s: infer_implicit_with_category(
            s, kg, domain_embs, domain_categories, domain_conf_threshold=domain_conf_threshold
        )
    )




    elapsed = time.time() - t0
    test.to_csv(save_path, index=False, encoding="utf-8")
    print(f"[SAVED] KG results -> {save_path}")
    print(f"[TIME] KG inference took {elapsed:.2f} sec")

    # --- Memory cleanup block ---
    del t0
    gc.collect()
    torch.cuda.empty_cache()
    # ----------------------------
    return test


# ================================================================
# STEP 4: Combine Explicit + Implicit
# ================================================================
def run_combined_only(test, save_path="test_combined.csv"):
    print("\n[STEP 4] Combine Explicit + Implicit...")
    t0 = time.time()

    def combine_aspects(row):
        seen, combined = set(), []
        for source in [row.get("explicit_iterative", []),
                       row.get("cleaned_explicit_paraphrases", [])]:#,

            if isinstance(source, list):
                for asp in source:
                    if asp and asp.lower() not in seen:
                        seen.add(asp.lower())
                        combined.append(asp)
        return combined
    def combine_aspects2(row):
        seen, combined = set(), []
        for source in [#row.get("cleaned_explicit_iterative", []),
                       row.get("cleaned_explicit_paraphrases", []),
                       row.get("implicit_kg", [])]:
            if isinstance(source, list):
                for asp in source:
                    if asp and asp.lower() not in seen:
                        seen.add(asp.lower())
                        combined.append(asp)
        return combined


    test["combined_aspects"] = test.apply(combine_aspects, axis=1)
    test["combined_aspects2"] = test.apply(combine_aspects2, axis=1)


    elapsed = time.time() - t0
    test.to_csv(save_path, index=False, encoding="utf-8")
    print(f"[SAVED] Combined results -> {save_path}")
    print(f"[TIME] Combining took {elapsed:.2f} sec")

    # --- Memory cleanup block ---
    del t0
    gc.collect()
    torch.cuda.empty_cache()
    # ----------------------------
    return test
```

```python
import re
from difflib import SequenceMatcher
import spacy
nlp = spacy.load("en_core_web_sm")

NUMERIC_RE = re.compile(r"^[\d\.\,]+%?$")

import re

# strip all unwanted special characters anywhere in the text
PUNCTUATION_STRIP_RE = re.compile(r"[\-\_\!\@\#\$\%\^\&\*\(\)\+\=\?\.\,<>\"';:\[\]\{\}]")
MULTISPACE_RE = re.compile(r"\s+")

def clean_text(text: str) -> str:
    # remove special characters
    cleaned = PUNCTUATION_STRIP_RE.sub("", text)
    # collapse multiple spaces to one
    cleaned = MULTISPACE_RE.sub(" ", cleaned)
    # trim leading/trailing spaces
    return cleaned.strip()

def is_nouny_text(text):
    doc = nlp(text)
    return any(t.pos_ in ("NOUN","PROPN") for t in doc)

def canonicalize(aspect, lexicon=None):
    doc = nlp(aspect.lower().strip())
    heads = [t.lemma_ for t in doc if t.pos_ in ("NOUN","PROPN")]
    base = heads[-1] if heads else aspect.lower().strip()
    if lexicon and base in lexicon:
        return lexicon[base]
    return base

def refine_aspects(aspects, original_sentence=None, enforce_verbatim=False,
                   sim_dedupe_threshold=0.80, lexicon=None):
    if not aspects or not isinstance(aspects, list):
        return []

    # 1) clean + POS
    pool = []

    for a in aspects:
        a = str(a).strip()
        if not a or NUMERIC_RE.match(a):
            continue
        a = clean_text(a)
        if not is_nouny_text(a):
            continue
        pool.append(a)

    # 2) enforce explicitness vs original (optional, recommended)
    if enforce_verbatim and original_sentence:
        keep = []
        s = original_sentence.lower()
        for a in pool:
            low = a.lower()
            if low in s:
                keep.append(a)
                continue
            # head-noun fallback
            doc = nlp(a)
            head = next((t.text for t in doc if t.pos_ in ("NOUN","PROPN")), doc[-1].text)
            if head.lower() in s:
                keep.append(a)
        pool = keep

    if not pool:
        return []

    # 3) canonicalize
    canon_pairs = [(a, canonicalize(a, lexicon)) for a in pool]

    # 4) dedup by canonical form
    seen, refined = set(), []
    for raw, can in canon_pairs:
        if can not in seen:
            seen.add(can)
            refined.append(raw)

    # 5) optional fuzzy collapse of near-duplicates (token-set-ish)
    final = []
    for cand in refined:
        add = True
        for ex in final:
            sim = SequenceMatcher(None, cand.lower(), ex.lower()).ratio()
            if sim >= sim_dedupe_threshold:
                # prefer shorter canonical (tend to be more general)
                if len(canonicalize(cand)) < len(canonicalize(ex)):
                    final.remove(ex)
                    final.append(cand)
                add = False
                break
        if add:
            final.append(cand)

    return final
```

```python
import ast

def clean_list_values(x):
    """
    Clean a cell that may be a list, stringified list, or a single value.
    - If list: remove NaN-like items.
    - If stringified list: parse first.
    - If scalar: return [] if NaN-like, else wrap in list.
    """
    # Try to parse stringified lists
    if isinstance(x, str):
        try:
            parsed = ast.literal_eval(x)
            if isinstance(parsed, list):
                x = parsed
        except Exception:
            pass  # keep as string if not parsable

    # Case 1: list
    if isinstance(x, list):
        cleaned = []
        for v in x:
            if v is None or pd.isna(v) or str(v).strip().lower() == "nan":
                continue
            cleaned.append(str(v).strip())
        return cleaned

    # Case 2: scalar
    if x is None or pd.isna(x) or str(x).strip().lower() == "nan":
        return []
    return [str(x).strip()]
```

#### New Evaluation function:

Evaluation pipeline:

1. post-process test['preidction'] (parse ‚Üí remove adjectives ‚Üí normalize ‚Üí dedup)

2. parse gold aspects

3. compute precision/recall/F1 per row and overall (micro & macro)

```python
def evaluate_dataframe(
    df: pd.DataFrame,
    pred_col: str = "preidction",
    gold_col: str = "target",
    use_camel: bool = False,
    jaccard_threshold: float = 0.6
):
    df = df.copy()
    df["pred_list_clean"] = df[pred_col].apply(lambda s: clean_predicted_aspects(s, use_camel=use_camel))

    def parse_gold(x):
        if isinstance(x, list):
            return x
        return parse_aspect_string(x)
    df["gold_list"] = df[gold_col].apply(parse_gold)

    per = df.apply(
        lambda r: prf1_for_example(r["gold_list"], r["pred_list_clean"], thresh=jaccard_threshold), axis=1
    )
    df["precision"] = per.apply(lambda m: m["precision"])
    df["recall"]    = per.apply(lambda m: m["recall"])
    df["f1"]        = per.apply(lambda m: m["f1"])
    df["tp"]        = per.apply(lambda m: m["tp"])
    df["fp"]        = per.apply(lambda m: m["fp"])
    df["fn"]        = per.apply(lambda m: m["fn"])
    # number of gold items in each example
    df["support"]   = df["gold_list"].apply(len)

    # -------- micro --------
    tot_tp, tot_fp, tot_fn = df["tp"].sum(), df["fp"].sum(), df["fn"].sum()
    micro_p = tot_tp / (tot_tp + tot_fp) if (tot_tp + tot_fp) else 0.0
    micro_r = tot_tp / (tot_tp + tot_fn) if (tot_tp + tot_fn) else 0.0
    micro_f = 2*micro_p*micro_r/(micro_p+micro_r) if (micro_p+micro_r) else 0.0

    # -------- macro --------
    macro_p = df["precision"].mean() if len(df) else 0.0
    macro_r = df["recall"].mean()    if len(df) else 0.0
    macro_f = df["f1"].mean()        if len(df) else 0.0

    # -------- weighted macro --------
    total_support = df["support"].sum()
    if total_support > 0:
        weighted_p = (df["precision"] * df["support"]).sum() / total_support
        weighted_r = (df["recall"]    * df["support"]).sum() / total_support
        weighted_f = (df["f1"]        * df["support"]).sum() / total_support
    else:
        weighted_p = weighted_r = weighted_f = 0.0


#     summary = {
#     "micro": {
#         "precision": round(float(micro_p), 4),
#         "recall": round(float(micro_r), 4),
#         "f1": round(float(micro_f), 4),
#         "tp": int(tot_tp),
#         "fp": int(tot_fp),
#         "fn": int(tot_fn),
#     },
#     "macro": {
#         "precision": round(float(macro_p), 4),
#         "recall": round(float(macro_r), 4),
#         "f1": round(float(macro_f), 4),
#     },
#     "weighted_macro": {
#         "precision": round(float(weighted_p), 4),
#         "recall": round(float(weighted_r), 4),
#         "f1": round(float(weighted_f), 4),
#     },
#     "df": df,
# }
    summary = {
        "micro": {
            "precision": round(float(micro_p), 4),
            "recall": round(float(micro_r), 4),
            "f1": round(float(micro_f), 4),
            "accuracy": round(float(tot_tp / (tot_tp + tot_fp + tot_fn)), 4),
            "tp": int(tot_tp),
            "fp": int(tot_fp),
            "fn": int(tot_fn),
        },
        "macro": {
            "precision": round(float(macro_p), 4),
            "recall": round(float(macro_r), 4),
            "f1": round(float(macro_f), 4),
            # optional: add macro accuracy if you compute per-class accuracies
        },
        "weighted_macro": {
            "precision": round(float(weighted_p), 4),
            "recall": round(float(weighted_r), 4),
            "f1": round(float(weighted_f), 4),
            # optional: weighted accuracy (same logic)
        },
        "df": df,
    }


    return summary
```

```python
def _explicitness_flags(aspect: str, original_sentence: str):
    """
    Compute explicitness components:
      - is_verbatim: aspect text (normalized) appears as a substring in sentence
      - head_in_sent: noun head lemma appears in sentence
      - jacc: token-level Jaccard between aspect and sentence
    These are used later with a tunable jaccard_explicitness threshold.
    """
    sent_norm = _normalize_for_verbatim(original_sentence)
    sent_tokens = set(sent_norm.split())

    low = _normalize_for_verbatim(aspect)
    if not low:
        return False, False, 0.0

    # verbatim substring
    is_verbatim = low in sent_norm

    # head noun
    head = _noun_head_lemma(aspect)
    head_in_sent = bool(head and head in sent_norm)

    # token-level Jaccard vs sentence
    a_tokens = set(low.split())
    inter = len(a_tokens & sent_tokens)
    union = len(a_tokens | sent_tokens) or 1
    jacc = inter / union

    return is_verbatim, head_in_sent, jacc
```

```python
def build_paraphrase_cache(sentences, verbose=False):
    """
    Run the expensive parts ONCE:
      - iterative residual decoding on original + paraphrases
      - collect original aspects
      - collect ALL paraphrase aspects (union)
      - compute:
          * canonical form (noun-head lemma)
          * support: how many paraphrases propose this canon
          * cosine similarity with original sentence (SBERT)
          * explicitness flags (verbatim, head_in_sent, jaccard)
    No filtering is applied here (no thresholds, no top_k).
    This cache can later be reused to apply different thresholds cheaply.
    """
    if not sentences or not isinstance(sentences, list):
        return {
            "original_sentence": None,
            "pinned_originals": [],
            "candidates": [],
        }

    original_sentence = sentences[0]

    # 1) iterative decoding on each input (original + paraphrases)
    per_input = []
    for s in sentences:
        per_input.append(iterative_residual_decoding(s, max_rounds=2))

    orig_aspects_raw = per_input[0]
    para_lists = per_input[1:]  # paraphrases only

    # 2) PIN original aspects (these are always preserved if include_original_always=True)
    pinned_originals = _dedup_preserve(orig_aspects_raw)

    # 3) collect ALL paraphrase aspects (union) ‚Äî no filtering yet
    if para_lists:
        all_para_aspects = _dedup_preserve(
            [a for lst in para_lists for a in lst]
        )
    else:
        all_para_aspects = []

    candidate_infos = []

    if all_para_aspects:
        # compute support = how many paraphrases contain the same canonical form
        # first build a map canon -> count
        canon_counts = Counter()
        per_para_canon_sets = []
        for lst in para_lists:
            canon_set = set(_noun_head_lemma(a) for a in lst)
            per_para_canon_sets.append(canon_set)
            for c in canon_set:
                canon_counts[c] += 1

        # SBERT embeddings (once per sentence)
        orig_emb = _sbert.encode(original_sentence, convert_to_tensor=True)
        asp_embs = _sbert.encode(all_para_aspects, convert_to_tensor=True)
        sims = util.cos_sim(asp_embs, orig_emb).cpu().numpy().flatten().tolist()

        # explicitness flags per candidate
        for a, sim_score in zip(all_para_aspects, sims):
            canon = _noun_head_lemma(a)
            support = canon_counts.get(canon, 0)
            is_verbatim, head_in_sent, jacc = _explicitness_flags(a, original_sentence)

            candidate_infos.append({
                "aspect": a,
                "canon": canon,
                "support": support,         # how many paraphrases voted this canon
                "sim": float(sim_score),    # cosine similarity
                "is_verbatim": is_verbatim,
                "head_in_sent": head_in_sent,
                "explicit_jacc": float(jacc),
            })

    if verbose:
        print(f"[CACHE] orig={len(pinned_originals)}, "
              f"para_raw={sum(len(lst) for lst in para_lists)}, "
              f"unique_para_cands={len(candidate_infos)}")

    return {
        "original_sentence": original_sentence,
        "pinned_originals": pinned_originals,
        "para_lists": para_lists,       # keep for debugging if needed
        "candidates": candidate_infos,  # list[dict]
    }
```

```python
def apply_paraphrase_filters_from_cache(
    cache_entry,
    top_k=2,
    similarity_threshold=0.58,
    min_consensus=1,
    jaccard_explicitness=0.50,
    include_original_always=True,
):
    """
    Rebuild final aspect list from a single cache_entry using:
      - min_consensus: minimum support (votes) over paraphrases
      - similarity_threshold: minimum cosine similarity
      - top_k: top-k by similarity
      - jaccard_explicitness: threshold over explicit_jacc
      - include_original_always: whether to keep original aspects regardless
    This is CHEAP: no SBERT, no decoding ‚Äî just filtering the cached candidates.
    """

    if cache_entry is None or cache_entry.get("original_sentence") is None:
        return []

    pinned_originals = cache_entry.get("pinned_originals", [])
    candidates = cache_entry.get("candidates", [])

    # 1) start from all candidates
    pool = candidates

    # 2) apply consensus via support (like _consensus_on_canonical)
    if min_consensus > 1:
        pool = [c for c in pool if c["support"] >= min_consensus]

        # fallback: if nothing survives but there were candidates, revert to all
        if not pool and candidates:
            pool = candidates

    # 3) similarity filter and sort
    pool = [c for c in pool if c["sim"] >= similarity_threshold]
    pool = sorted(pool, key=lambda c: c["sim"], reverse=True)
    if top_k is not None:
        pool = pool[:top_k]

    # 4) explicitness filter (same logic as _explicitness_filter)
    filtered = []
    for c in pool:
        if c["is_verbatim"] or c["head_in_sent"] or c["explicit_jacc"] >= jaccard_explicitness:
            filtered.append(c)

    kept_from_para = [c["aspect"] for c in filtered]

    # 5) merge with originals, dedup
    final = []
    if include_original_always:
        final.extend(pinned_originals)
    final.extend(kept_from_para)

    final = _dedup_preserve(final)
    return final
```

```python
import re

def normalize_aspect(aspect: str) -> str:
    asp = aspect.lower().strip()
    asp = re.sub(r"^\W+|\W+$", "", asp)   # remove punctuation at edges
    if asp.endswith("s") and len(asp) > 3:  # crude singularization
        asp = asp[:-1]
    return asp

def filter_to_sentence(aspects, original_sentence):
    sent = original_sentence.lower()
    return [a for a in aspects if a.lower() in sent]


def iterative_residual_decoding_with_history(sentence, max_rounds=5):
    """
    Same logic as iterative_residual_decoding, but:
      - runs up to max_rounds
      - after each round, stores the *current* final_aspects
    Returns:
      per_round_outputs: list of lists
        per_round_outputs[r-1] = aspects after r rounds (1-based)
    """
    original_sentence = sentence
    remaining_sentence = sentence
    all_aspects = []
    seen_norm = set()
    per_round_outputs = []

    for round_id in range(max_rounds):
        aspects = extract_aspect(remaining_sentence)  # returns list
        if not aspects:
            break

        new_aspects = [a.strip() for a in aspects if a.strip()]
        if not new_aspects or all(a.lower() == "none" for a in new_aspects):
            break

        new_aspects = filter_to_sentence(new_aspects, original_sentence)

        for asp in new_aspects:
            norm = normalize_aspect(asp)
            if norm not in seen_norm:
                seen_norm.add(norm)
                all_aspects.append(asp)  # keep original form

            # remove exact form from sentence
            pattern = r"\b{}\b".format(re.escape(asp))
            remaining_sentence = re.sub(pattern, "", remaining_sentence, flags=re.IGNORECASE).strip()

        if not remaining_sentence:
            # sentence fully consumed
            # still record current snapshot then break
            final_aspects = [a for a in all_aspects
                             if a.lower() != "none" or len(all_aspects) == 1]
            per_round_outputs.append(final_aspects.copy())
            break

        # snapshot after this round
        final_aspects = [a for a in all_aspects
                         if a.lower() != "none" or len(all_aspects) == 1]
        per_round_outputs.append(final_aspects.copy())

    return per_round_outputs
```

```python
from tqdm import tqdm

# Prepare containers for each target round
round_cols = {2: [], 3: [], 4: [], 5: []}

for s in tqdm(test["sentence"], desc="Running IG up to 5 rounds"):
    hist = iterative_residual_decoding_with_history(s, max_rounds=5)

    # If IG stopped early (e.g. only 1 or 2 rounds), we reuse the last snapshot.
    if not hist:
        # no aspects at all
        last = []
    else:
        last = hist[-1]

    for r in [2, 3, 4, 5]:
        if len(hist) >= r:
            round_cols[r].append(hist[r-1])   # aspects after r rounds
        else:
            round_cols[r].append(last)        # fallback to last available snapshot

# Assign to DataFrame
for r in [2, 3, 4, 5]:
    test[f"ig_rounds_{r}"] = round_cols[r]
```

```python
from sklearn.metrics import classification_report
import numpy as np

def macro_metrics(df, gold_col='target_en', pred_col='prediction', sep_token='[sep]', match_type='partial'):
    def normalize(text):
        return text.lower().strip()

    def is_match(pred, gold_terms, exact=False):
        pred = normalize(pred)
        for gold in gold_terms:
            gold = normalize(gold)
            if exact:
                if pred == gold:
                    return True
            else:
                if pred in gold or gold in pred:
                    return True
        return False

    exact = (match_type == "exact")

    # Collect all unique normalized terms from gold and predictions
    all_labels = set()

    y_true = []
    y_pred = []

    for _, row in df.iterrows():
        gold = row.get(gold_col, [])
        pred_str = row.get(pred_col, "")

        if isinstance(pred_str, list):
            pred_str = pred_str[0]

        if not isinstance(gold, list):
            continue

        gold_terms = [normalize(t) for t in gold]
        pred_terms = [normalize(t) for t in pred_str.split(sep_token) if t.strip()] if isinstance(pred_str, str) else []

        all_labels.update(gold_terms)
        all_labels.update(pred_terms)

        y_true.append(gold_terms)
        y_pred.append(pred_terms)

    all_labels = sorted(all_labels)
    label_to_id = {label: i for i, label in enumerate(all_labels)}
    n_labels = len(all_labels)

    def to_binary_vec(example_terms, reference_terms):
        vec = np.zeros(n_labels)
        for label in all_labels:
            if is_match(label, example_terms, exact=exact):
                vec[label_to_id[label]] = 1
        return vec

    y_true_bin = np.array([to_binary_vec(gold, gold) for gold in y_true])
    y_pred_bin = np.array([to_binary_vec(pred, gold) for pred, gold in zip(y_pred, y_true)])

    # Generate classification report
    report = classification_report(
        y_true_bin, y_pred_bin,
        target_names=all_labels,
        output_dict=True,
        zero_division=0
    )

    # Summary
    print(f"\nMacro Precision: {report['macro avg']['precision']:.2f}")
    print(f"Macro Recall:    {report['macro avg']['recall']:.2f}")
    print(f"Macro F1:        {report['macro avg']['f1-score']:.2f}")
    print(f"Weighted F1:     {report['weighted avg']['f1-score']:.2f}")

    return report
```

```python
## Measure BLEU Score ( Precision-oriented measure)
from nltk.translate.bleu_score import SmoothingFunction, corpus_bleu, sentence_bleu


def bleu(ref, gen):
    '''
    calculate pair wise bleu score. uses nltk implementation
    Args:
        references : a list of reference sentences
        candidates : a list of candidate(generated) sentences
    Returns:
        bleu score(float)
    '''
    ref_bleu = []
    gen_bleu = []
    for l in gen:
        gen_bleu.append(l.split())
    for i,l in enumerate(ref):
        ref_bleu.append([l.split()])
    cc = SmoothingFunction()
    score_bleu = corpus_bleu(ref_bleu, gen_bleu, weights=(0, 1, 0, 0), smoothing_function=cc.method4)
    return score_bleu
```

```python
#rouge scores for a reference/generated sentence pair
#source google seq2seq source code.

import itertools

#supporting function
def _split_into_words(sentences):
  """Splits multiple sentences into words and flattens the result"""
  return list(itertools.chain(*[_.split(" ") for _ in sentences]))

#supporting function
def _get_word_ngrams(n, sentences):
  """Calculates word n-grams for multiple sentences.
  """
  assert len(sentences) > 0
  assert n > 0

  words = _split_into_words(sentences)
  return _get_ngrams(n, words)

#supporting function
def _get_ngrams(n, text):
  """Calcualtes n-grams.
  Args:
    n: which n-grams to calculate
    text: An array of tokens
  Returns:
    A set of n-grams
  """
  ngram_set = set()
  text_length = len(text)
  max_index_ngram_start = text_length - n
  for i in range(max_index_ngram_start + 1):
    ngram_set.add(tuple(text[i:i + n]))
  return ngram_set

def rouge_n(reference_sentences, evaluated_sentences, n=2):
  """
  Computes ROUGE-N of two text collections of sentences.
  Source: http://research.microsoft.com/en-us/um/people/cyl/download/
  papers/rouge-working-note-v1.3.1.pdf
  Args:
    evaluated_sentences: The sentences that have been picked by the summarizer
    reference_sentences: The sentences from the referene set
    n: Size of ngram.  Defaults to 2.
  Returns:
    recall rouge score(float)
  Raises:
    ValueError: raises exception if a param has len <= 0
  """
  if len(evaluated_sentences) <= 0 or len(reference_sentences) <= 0:
    raise ValueError("Collections must contain at least 1 sentence.")

  evaluated_ngrams = _get_word_ngrams(n, evaluated_sentences)
  reference_ngrams = _get_word_ngrams(n, reference_sentences)
  reference_count = len(reference_ngrams)
  evaluated_count = len(evaluated_ngrams)

  # Gets the overlapping ngrams between evaluated and reference
  overlapping_ngrams = evaluated_ngrams.intersection(reference_ngrams)
  overlapping_count = len(overlapping_ngrams)

  # Handle edge case. This isn't mathematically correct, but it's good enough
  if evaluated_count == 0:
    precision = 0.0
  else:
    precision = overlapping_count / evaluated_count

  if reference_count == 0:
    recall = 0.0
  else:
    recall = overlapping_count / reference_count

  f1_score = 2.0 * ((precision * recall) / (precision + recall + 1e-8))

  #just returning recall count in rouge, useful for our purpose
  return recall
```

#### Caching block (k = 2):

```python
from collections import Counter
from tqdm import tqdm
import pandas as pd

tqdm.pandas()

# -----------------------------
# Helpers used by cache builder
# (assumes these already exist in your notebook):
#   _normalize_for_verbatim, _noun_head_lemma, _dedup_preserve
#   _sbert (SentenceTransformer) and util.cos_sim imported
# -----------------------------

def _explicitness_flags(aspect: str, original_sentence: str):
    """Return (is_verbatim, head_in_sent, token_jaccard) vs original sentence."""
    sent_norm = _normalize_for_verbatim(original_sentence)
    sent_tokens = set(sent_norm.split())

    low = _normalize_for_verbatim(aspect)
    if not low:
        return False, False, 0.0

    is_verbatim = low in sent_norm
    head = _noun_head_lemma(aspect)
    head_in_sent = bool(head and head in sent_norm)

    a_tokens = set(low.split())
    inter = len(a_tokens & sent_tokens)
    union = len(a_tokens | sent_tokens) or 1
    jacc = inter / union
    return is_verbatim, head_in_sent, float(jacc)


def build_para_cache_from_cached_preds(
    original_sentence: str,
    pred_orig: list,
    pred_para_list: list,   # list[list[str]]
    max_paras: int = 2,     # <-- k fixed here: use only first 2 paraphrases
):
    """
    Build a cache object (no LLM calls):
      - Uses cached predictions: pred_orig, pred_para_list
      - Computes per-candidate:
          support (votes across paraphrases), SBERT sim vs original,
          explicitness flags (verbatim/head/jaccard)
    """
    if not original_sentence:
        return {"original_sentence": None, "pinned_originals": [], "candidates": []}

    pred_orig = pred_orig if isinstance(pred_orig, list) else []
    pred_para_list = pred_para_list if isinstance(pred_para_list, list) else []
    pred_para_list = pred_para_list[:max_paras]  # enforce k

    pinned_originals = _dedup_preserve(pred_orig)

    # union of paraphrase aspects (dedup)
    all_para_aspects = _dedup_preserve([a for lst in pred_para_list for a in (lst or [])])

    candidate_infos = []
    if all_para_aspects and pred_para_list:
        # support: how many paraphrases propose this canonical form
        canon_counts = Counter()
        for lst in pred_para_list:
            canon_set = set(_noun_head_lemma(a) for a in (lst or []))
            for c in canon_set:
                canon_counts[c] += 1

        # SBERT sims vs original (once per row)
        orig_emb = _sbert.encode(original_sentence, convert_to_tensor=True)
        asp_embs = _sbert.encode(all_para_aspects, convert_to_tensor=True)
        sims = util.cos_sim(asp_embs, orig_emb).cpu().numpy().flatten().tolist()

        for a, sim_score in zip(all_para_aspects, sims):
            canon = _noun_head_lemma(a)
            support = int(canon_counts.get(canon, 0))
            is_verbatim, head_in_sent, jacc = _explicitness_flags(a, original_sentence)
            candidate_infos.append({
                "aspect": a,
                "canon": canon,
                "support": support,
                "sim": float(sim_score),
                "is_verbatim": bool(is_verbatim),
                "head_in_sent": bool(head_in_sent),
                "explicit_jacc": float(jacc),
            })

    return {
        "original_sentence": original_sentence,
        "pinned_originals": pinned_originals,
        "candidates": candidate_infos,
    }


# -----------------------------
# Build para_cache for the whole DF (NO LLM inference)
# Requires columns: sentence, pred_orig, pred_para_list
# -----------------------------
MAX_PARAS_FOR_CACHE = 2  # <-- FIXED k=2 here

test["para_cache"] = test.progress_apply(
    lambda r: build_para_cache_from_cached_preds(
        original_sentence=r["sentence"],
        pred_orig=r.get("cleaned_pred_orig", []),
        pred_para_list=r.get("pred_para_list_clean", []),
        max_paras=MAX_PARAS_FOR_CACHE,
    ),
    axis=1
)

print(f"[OK] Built para_cache with k={MAX_PARAS_FOR_CACHE} paraphrases (original + {MAX_PARAS_FOR_CACHE}).")
```

####Sensitivity tests block (English):

```python
import numpy as np
import pandas as pd

def apply_paraphrase_filters_from_cache(
    cache_entry,
    top_k=2,
    similarity_threshold=0.58,
    min_consensus=1,
    jaccard_explicitness=0.50,
    include_original_always=True,
):
    """Rebuild final aspect list from cached candidates (cheap; no LLM)."""
    if cache_entry is None or cache_entry.get("original_sentence") is None:
        return []

    pinned_originals = cache_entry.get("pinned_originals", [])
    candidates = cache_entry.get("candidates", [])

    pool = candidates

    # 1) consensus via support
    if min_consensus > 1:
        pool = [c for c in pool if c["support"] >= min_consensus]
        if not pool and candidates:
            pool = candidates  # fallback

    # 2) similarity filter + sort
    pool = [c for c in pool if c["sim"] >= similarity_threshold]
    pool = sorted(pool, key=lambda c: c["sim"], reverse=True)
    if top_k is not None:
        pool = pool[:top_k]

    # 3) explicitness filter
    filtered = []
    for c in pool:
        if c["is_verbatim"] or c["head_in_sent"] or c["explicit_jacc"] >= jaccard_explicitness:
            filtered.append(c)

    kept_from_para = [c["aspect"] for c in filtered]

    # 4) merge with originals, dedup
    final = []
    if include_original_always:
        final.extend(pinned_originals)
    final.extend(kept_from_para)

    return _dedup_preserve(final)


# -----------------------------
# Grids (edit as needed)
# -----------------------------
grid_similarity = [0.50, 0.55, 0.58, 0.60, 0.62, 0.65, 0.70, 0.75, 0.80]
grid_top_k      = [1, 2, 3, 4, 5]
grid_min_cons   = [1, 2, 3]
grid_jacc_exp   = [0.00, 0.25, 0.50, 0.75, 1.00]

GOLD_COL = "aspect"   # your English gold column


# Utility: run evaluation for a produced prediction column
def _eval_col(colname: str):
    return evaluate_dataframe(
        test,
        pred_col=colname,
        gold_col=GOLD_COL,
        use_camel=False,
        jaccard_threshold=0.6
    )


# -----------------------------
# 1) similarity sweep (fix others)
# -----------------------------
TOP_K = 2
MIN_CONS = 1
JACC_EXP = 0.50

sim_results = []
for sim in grid_similarity:
    col = f"cache_sim{sim}_k{TOP_K}_cons{MIN_CONS}_j{JACC_EXP}"
    test[col] = test["para_cache"].apply(
        lambda cache: apply_paraphrase_filters_from_cache(
            cache,
            top_k=TOP_K,
            similarity_threshold=sim,
            min_consensus=MIN_CONS,
            jaccard_explicitness=JACC_EXP,
            include_original_always=True,
        )
    )
    res = _eval_col(col)
    sim_results.append({
        "similarity_threshold": sim,
        "precision": res["micro"]["precision"],
        "recall": res["micro"]["recall"],
        "micro_F1": res["micro"]["f1"],
    })

df_sim = pd.DataFrame(sim_results)
print(df_sim)


# -----------------------------
# 2) top_k sweep (pick best sim from df_sim)
# -----------------------------
best_sim = float(df_sim.loc[df_sim["micro_F1"].idxmax(), "similarity_threshold"])

topk_results = []
for k in grid_top_k:
    col = f"cache_sim{best_sim}_k{k}_cons{MIN_CONS}_j{JACC_EXP}"
    test[col] = test["para_cache"].apply(
        lambda cache: apply_paraphrase_filters_from_cache(
            cache,
            top_k=k,
            similarity_threshold=best_sim,
            min_consensus=MIN_CONS,
            jaccard_explicitness=JACC_EXP,
            include_original_always=True,
        )
    )
    res = _eval_col(col)
    topk_results.append({
        "top_k": k,
        "precision": res["micro"]["precision"],
        "recall": res["micro"]["recall"],
        "micro_F1": res["micro"]["f1"],
    })

df_topk = pd.DataFrame(topk_results)
print(df_topk)


# -----------------------------
# 3) min_consensus sweep (fix sim & top_k at their best)
# -----------------------------
best_topk = int(df_topk.loc[df_topk["micro_F1"].idxmax(), "top_k"])

cons_results = []
for mc in grid_min_cons:
    col = f"cache_sim{best_sim}_k{best_topk}_cons{mc}_j{JACC_EXP}"
    test[col] = test["para_cache"].apply(
        lambda cache: apply_paraphrase_filters_from_cache(
            cache,
            top_k=best_topk,
            similarity_threshold=best_sim,
            min_consensus=mc,
            jaccard_explicitness=JACC_EXP,
            include_original_always=True,
        )
    )
    res = _eval_col(col)
    cons_results.append({
        "min_consensus": mc,
        "precision": res["micro"]["precision"],
        "recall": res["micro"]["recall"],
        "micro_F1": res["micro"]["f1"],
    })

df_cons = pd.DataFrame(cons_results)
print(df_cons)


# -----------------------------
# 4) jaccard_explicitness sweep (fix sim, top_k, min_consensus at their best)
# -----------------------------
best_mc = int(df_cons.loc[df_cons["micro_F1"].idxmax(), "min_consensus"])

jacc_results = []
for j in grid_jacc_exp:
    col = f"cache_sim{best_sim}_k{best_topk}_cons{best_mc}_j{j}"
    test[col] = test["para_cache"].apply(
        lambda cache: apply_paraphrase_filters_from_cache(
            cache,
            top_k=best_topk,
            similarity_threshold=best_sim,
            min_consensus=best_mc,
            jaccard_explicitness=j,
            include_original_always=True,
        )
    )
    res = _eval_col(col)
    jacc_results.append({
        "jaccard_explicitness": j,
        "precision": res["micro"]["precision"],
        "recall": res["micro"]["recall"],
        "micro_F1": res["micro"]["f1"],
    })

df_jacc = pd.DataFrame(jacc_results)
print(df_jacc)
```
