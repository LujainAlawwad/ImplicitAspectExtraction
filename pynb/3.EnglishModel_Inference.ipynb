```python
import ast

def fast_parse_list(x):
    """Parse string/list values, and split by newline if needed."""
    if isinstance(x, str):
        x_strip = x.strip()
        # Try parse if it looks like a Python list
        if x_strip.startswith("[") and x_strip.endswith("]"):
            try:
                parsed = ast.literal_eval(x_strip)
                return fast_parse_list(parsed)  # handle nested
            except Exception:
                return [s.strip() for s in x_strip.split("\n") if s.strip()]
        # Otherwise, split by newline directly
        return [s.strip() for s in x_strip.split("\n") if s.strip()]

    elif isinstance(x, list):
        # flatten, split by newline if needed
        out = []
        for i in x:
            if isinstance(i, str):
                out.extend([s.strip() for s in i.split("\n") if s.strip()])
        return out

    else:
        return [str(x)]

def process_data2(df):
    df = df.copy()
    df["aspect"] = df["aspect"].apply(fast_parse_list)
    df["paraphrases"] = df["paraphrases"].apply(fast_parse_list)
    df.reset_index(drop=True, inplace=True)
    return df
```

```python
import re
from typing import List, Tuple, Dict, Iterable

# -----------------------------
# 1) Parse model string -> list
# -----------------------------
_SEP_PAT = re.compile(
    r"\s*(?:\[sep\]|\[SEP\]|،|,|;|؛|\||\n|\t|\/|\\|:|\s{2,})\s*", re.IGNORECASE
)

def parse_aspect_string(s) -> List[str]:
    """
    Robustly parse a model's string into a list of aspect terms.
    Handles Arabic/English commas, ; ؛ | [sep], brackets, quotes, whitespace.
    If input is already a list, returns a cleaned version.
    """
    if s is None:
        return []
    if isinstance(s, list):
        items = s
    else:
        # strip surrounding brackets/quotes if present
        s = str(s).strip()
        s = s.strip("[](){}")
        s = s.replace("'", "").replace('"', "")
        # split on many possible separators
        items = _SEP_PAT.split(s)

    # clean tokens: remove empties/stray punctuation
    cleaned = []
    for t in items:
        t = t.strip()
        if not t:
            continue
        # remove leading/trailing punctuation
        t = re.sub(r"^[\s\-\–\—\·\•\*\.]+|[\s\-\–\—\·\•\*\.]+$", "", t)
        if t:
            cleaned.append(t)
    # de-duplicate while preserving order
    seen = set()
    dedup = []
    for w in cleaned:
        if w not in seen:
            seen.add(w)
            dedup.append(w)
    return dedup

# ----------------------------------------
# 2) Light Arabic normalization for eval
# ----------------------------------------
_DIACRITICS = re.compile(r'[\u0617-\u061A\u064B-\u0652]')
# common clitic prefixes and pronominal suffixes
_PREFIXES = re.compile(r'^(?:ال|و|ف|ب|ل|ك)+')
_SUFFIXES = re.compile(r'(?:هما|كما|ها|هم|نا|كم|هن|كن|ه|ي|ة|ات)$')

def normalize_token(tok: str) -> str:
    t = str(tok)
    # unify common letters
    t = t.replace("أ", "ا").replace("إ", "ا").replace("آ", "ا")
    t = t.replace("ى", "ي").replace("ة", "ه").replace("ؤ", "ء").replace("ئ", "ء")
    # remove diacritics & tatweel
    t = _DIACRITICS.sub("", t)
    t = t.replace("ـ", "")
    # trim clitics (light)
    t = _PREFIXES.sub("", t)
    t = _SUFFIXES.sub("", t)
    return t.strip()

def normalize_phrase(phrase: str) -> str:
    # token-wise normalization, preserve spaces between meaningful tokens
    toks = [w for w in re.split(r"\s+", phrase.strip()) if w]
    return " ".join(normalize_token(w) for w in toks)

# ----------------------------------------------------
# 3) Flexible matching: exact, substring, Jaccard(IoU)
# ----------------------------------------------------
def jaccard(a: str, b: str) -> float:
    A = set(a.split())
    B = set(b.split())
    if not A and not B:
        return 1.0
    if not A or not B:
        return 0.0
    return len(A & B) / len(A | B)

def similarity(gold: str, pred: str) -> float:
    """Return similarity score in [0,1] after normalization.
    1.0 if exact or substring; else token Jaccard."""
    g = normalize_phrase(gold)
    p = normalize_phrase(pred)
    if not g or not p:
        return 0.0
    if g == p:
        return 1.0
    if g in p or p in g:
        return 1.0
    return jaccard(g, p)

# ---------------------------------------------------
# 4) Greedy bipartite matching above a threshold
# ---------------------------------------------------
def best_alignment(gold: List[str], pred: List[str], thresh: float = 0.6) -> Tuple[int, List[Tuple[int,int,float]]]:
    """
    Greedy maximum-weight matching:
    - Build all pairs with sim >= thresh
    - Sort by similarity desc
    - Pick non-conflicting pairs
    Returns TP count and list of matches (gi, pj, sim)
    """
    pairs = []
    for i, g in enumerate(gold):
        for j, p in enumerate(pred):
            s = similarity(g, p)
            if s >= thresh:
                pairs.append((i, j, s))
    pairs.sort(key=lambda x: x[2], reverse=True)

    matched_g = set()
    matched_p = set()
    matches = []
    for gi, pj, s in pairs:
        if gi not in matched_g and pj not in matched_p:
            matched_g.add(gi)
            matched_p.add(pj)
            matches.append((gi, pj, s))
    return len(matches), matches

# ----------------------------------------------------------------
# 5) Precision/Recall/F1 for a single example and aggregated stats
# ----------------------------------------------------------------
# def prf1_for_example(gold_terms: Iterable[str], pred_terms: Iterable[str], thresh: float = 0.6) -> Dict[str, float]:
#     gold = [t for t in map(str, gold_terms) if t and t.strip()]
#     pred = [t for t in map(str, pred_terms) if t and t.strip()]

#     tp, _ = best_alignment(gold, pred, thresh=thresh)
#     fp = max(0, len(pred) - tp)
#     fn = max(0, len(gold) - tp)

#     prec = tp / (tp + fp) if (tp + fp) > 0 else 0.0
#     rec  = tp / (tp + fn) if (tp + fn) > 0 else 0.0
#     f1   = 2 * prec * rec / (prec + rec) if (prec + rec) > 0 else 0.0

#     return {"tp": tp, "fp": fp, "fn": fn, "precision": prec, "recall": rec, "f1": f1}

def prf1_for_example(gold_terms: Iterable[str], pred_terms: Iterable[str], thresh: float = 0.6) -> Dict[str, float]:
    gold = [t for t in map(str, gold_terms) if t and t.strip()]
    pred = [t for t in map(str, pred_terms) if t and t.strip()]

    # Special case: both empty → perfect prediction
    if not gold and not pred:
        return {"tp": 0, "fp": 0, "fn": 0,
                "precision": 1.0, "recall": 1.0, "f1": 1.0}

    tp, _ = best_alignment(gold, pred, thresh=thresh)
    fp = max(0, len(pred) - tp)
    fn = max(0, len(gold) - tp)

    prec = tp / (tp + fp) if (tp + fp) > 0 else 0.0
    rec  = tp / (tp + fn) if (tp + fn) > 0 else 0.0
    f1   = 2 * prec * rec / (prec + rec) if (prec + rec) > 0 else 0.0

    return {"tp": tp, "fp": fp, "fn": fn,
            "precision": prec, "recall": rec, "f1": f1}


def evaluate_aspects(
    gold_lists: List[List[str]],
    pred_strings: List[str],
    parse_threshold: float = 0.6
) -> Dict[str, object]:
    """
    Evaluate lists of gold aspects vs model string predictions.
    - gold_lists: list of gold lists, e.g., [['الفندق','موقف'], ...]
    - pred_strings: list of raw model strings for each example
    - parse_threshold: similarity threshold for a match (default 0.6)
    Returns dict with micro/macro P/R/F1 and per-example scores.
    """
    assert len(gold_lists) == len(pred_strings), "gold and pred sizes must match"

    per_example = []
    tot_tp = tot_fp = tot_fn = 0

    for gold, pred_str in zip(gold_lists, pred_strings):
        pred_list = parse_aspect_string(pred_str)
        metrics = prf1_for_example(gold, pred_list, thresh=parse_threshold)
        per_example.append(metrics)
        tot_tp += metrics["tp"]
        tot_fp += metrics["fp"]
        tot_fn += metrics["fn"]

    # micro
    micro_p = tot_tp / (tot_tp + tot_fp) if (tot_tp + tot_fp) > 0 else 0.0
    micro_r = tot_tp / (tot_tp + tot_fn) if (tot_tp + tot_fn) > 0 else 0.0
    micro_f = 2 * micro_p * micro_r / (micro_p + micro_r) if (micro_p + micro_r) > 0 else 0.0

    # macro
    if per_example:
        macro_p = sum(x["precision"] for x in per_example) / len(per_example)
        macro_r = sum(x["recall"] for x in per_example) / len(per_example)
        macro_f = sum(x["f1"] for x in per_example) / len(per_example)
    else:
        macro_p = macro_r = macro_f = 0.0

    return {
        "micro": {"precision": micro_p, "recall": micro_r, "f1": micro_f, "tp": tot_tp, "fp": tot_fp, "fn": tot_fn},
        "macro": {"precision": macro_p, "recall": macro_r, "f1": macro_f},
        "per_example": per_example,
    }
```

```python
import json
from pathlib import Path
import networkx as nx
from typing import Dict, Any, List

class KnowledgeGraph:
    """
    Reusable Knowledge Graph wrapper (NetworkX).
    - Handles both structured aspects (dicts) and string lists.
    - Normalizes IDs but preserves original labels.
    - IMPORTANT: Ignores any JSON edges of the form aspect -> is_relevant_to_domain -> domain.
                 Only uses:
                   adjective -> describes_aspect -> aspect
                   aspect    -> belongs_to_entity -> entity
                   entity    -> is_relevant_to_domain -> domain
    """
    def __init__(self):
        self.G = nx.DiGraph()
        self.meta: Dict[str, Any] = {}
        # Quick indexes for convenience
        self._id2label: Dict[str, str] = {}
        self._label2id: Dict[str, str] = {}

    @staticmethod
    def _norm(s):
        return s.strip().lower() if isinstance(s, str) else s

    @staticmethod
    def _pred_str(d) -> str:
        return (d.get("predicate") or "").lower().replace(" ", "_")

    @classmethod
    def from_json(cls, path: str | Path) -> "KnowledgeGraph":
        kg = cls()
        kg._load(Path(path))
        return kg

    def _add_node(self, nid: str, label: str, ntype: str, **attrs):
        if nid not in self.G:
            self.G.add_node(nid, type=ntype, label=label, **attrs)
            if label and label not in self._label2id:
                self._label2id[label] = nid
            self._id2label[nid] = label
        else:
            # upgrade type if unknown
            if ntype and self.G.nodes[nid].get("type") == "unknown":
                self.G.nodes[nid]["type"] = ntype
            if "label" not in self.G.nodes[nid]:
                self.G.nodes[nid]["label"] = label

    def _load(self, path: Path):
        with open(path, "r", encoding="utf-8") as f:
            data = json.load(f)
        self.meta["source_path"] = str(path)

        # 1) Domains
        for d in data.get("domains", []):
            if isinstance(d, dict) and "name" in d:
                did = self._norm(d["name"])
                self._add_node(did, d["name"], "domain", **{k: v for k, v in d.items() if k != "name"})

        # 2) Entities (+ entity -> domain)
        for e in data.get("entities", []):
            if isinstance(e, dict) and "name" in e:
                eid = self._norm(e["name"])
                self._add_node(eid, e["name"], "entity", **{k: v for k, v in e.items() if k != "name"})
                dom = e.get("domain")
                if isinstance(dom, str) and dom.strip():
                    did = self._norm(dom)
                    if did not in self.G:
                        self._add_node(did, dom, "domain")
                    self.G.add_edge(
                        eid, did,
                        predicate="is_relevant_to_domain",
                        subject=e["name"], object=dom,
                        sub_lang=e.get("language"), obj_lang=e.get("language")
                    )

        # 3) Aspects (+ aspect -> entity)
        aspects = data.get("aspects", [])
        if aspects and isinstance(aspects[0], dict):
            for a in aspects:
                if not isinstance(a, dict) or "name" not in a:
                    continue
                aid = self._norm(a["name"])
                self._add_node(aid, a["name"], "aspect", **{k: v for k, v in a.items() if k != "name"})
                ents = a.get("entity", []) or []
                for ent in ents:
                    eid = self._norm(ent)
                    if eid not in self.G:
                        self._add_node(eid, ent, "entity")
                    self.G.add_edge(
                        aid, eid,
                        predicate="belongs_to_entity",
                        subject=a["name"], object=ent,
                        sub_lang=a.get("language"), obj_lang=a.get("language")
                    )
        else:
            # Aspects as strings
            for a in aspects:
                if isinstance(a, str):
                    aid = self._norm(a)
                    self._add_node(aid, a, "aspect")

        # 4) Adjectives
        for adj in data.get("adjectives", []):
            if isinstance(adj, str):
                aid = self._norm(adj)
                self._add_node(aid, adj, "adjective")

        # 5) Relationships (ignore aspect -> domain)
        for rel in data.get("relationships", []):
            subj = self._norm(rel.get("subject"))
            obj  = self._norm(rel.get("object"))
            pred = self._pred_str(rel)
            if not subj or not obj or not pred:
                continue

            # ensure nodes exist
            if subj not in self.G:
                self._add_node(subj, rel.get("subject", subj), "unknown")
            if obj not in self.G:
                self._add_node(obj, rel.get("object", obj), "unknown")

            # HARD RULE: JSON must NOT define aspect -> is_relevant_to_domain -> domain
            if pred == "is_relevant_to_domain" and self.G.nodes[subj].get("type") == "aspect":
                continue

            # add allowed edge as-is
            self.G.add_edge(subj, obj, **rel)

        # Summary
        self.meta["counts"] = self.counts()

    # ------------------- Queries -------------------
    def counts(self) -> dict:
        types = {}
        for _, d in self.G.nodes(data=True):
            t = d.get("type", "unknown")
            types[t] = types.get(t, 0) + 1
        return {
            "nodes": self.G.number_of_nodes(),
            "edges": self.G.number_of_edges(),
            **{f"n_{k}": v for k, v in types.items()},
        }

    def get_traces_for_adjective(self, adjective: str) -> List[dict]:
        """Return all adjective -> aspect -> entity -> domain traces."""
        adj_id = self._norm(adjective)
        if adj_id not in self.G:
            return []
        traces = []
        for _, aspect_id, d1 in self.G.out_edges(adj_id, data=True):
            if self._pred_str(d1) != "describes_aspect":
                continue
            for _, ent_id, d2 in self.G.out_edges(aspect_id, data=True):
                if self._pred_str(d2) != "belongs_to_entity":
                    continue
                for _, dom_id, d3 in self.G.out_edges(ent_id, data=True):
                    if self._pred_str(d3) != "is_relevant_to_domain":
                        continue
                    traces.append({
                        "adjective": self.G.nodes[adj_id].get("label", adj_id),
                        "aspect":    self.G.nodes[aspect_id].get("label", aspect_id),
                        "entity":    self.G.nodes[ent_id].get("label", ent_id),
                        "domain":    self.G.nodes[dom_id].get("label", dom_id),
                    })
        return traces

    def get_aspects_by_entity(self, entity: str) -> List[str]:
        """Aspects that belong to a given entity."""
        eid = self._norm(entity)
        if eid not in self.G:
            return []
        aspects = [
            u for u, v, d in self.G.in_edges(eid, data=True)
            if self._pred_str(d) == "belongs_to_entity"
        ]
        return sorted({self.G.nodes[a].get("label", a) for a in aspects})

    def get_entities_by_domain(self, domain: str) -> List[str]:
        """Entities that are relevant to a given domain."""
        did = self._norm(domain)
        if did not in self.G:
            return []
        ents = [
            u for u, v, d in self.G.in_edges(did, data=True)
            if self._pred_str(d) == "is_relevant_to_domain"
        ]
        return sorted({self.G.nodes[e].get("label", e) for e in ents})

    def get_adjectives_for_aspect(self, aspect: str) -> List[str]:
        """Adjectives that describe a given aspect."""
        aid = self._norm(aspect)
        if aid not in self.G:
            return []
        adjs = [
            u for u, v, d in self.G.in_edges(aid, data=True)
            if self._pred_str(d) == "describes_aspect"
        ]
        return sorted({self.G.nodes[x].get("label", x) for x in adjs})

    # ---------- NEW convenience helpers (names as requested) ----------
    def get_aspects(self, adjective: str) -> List[str]:
        """Aspects linked from an adjective via 'describes_aspect'."""
        adj_id = self._norm(adjective)
        if adj_id not in self.G:
            return []
        aspects = [
            v for _, v, d in self.G.out_edges(adj_id, data=True)
            if self._pred_str(d) == "describes_aspect"
        ]
        return sorted({self.G.nodes[a].get("label", a) for a in aspects})

    def get_entities(self, aspect: str) -> List[str]:
        """Entities linked from an aspect via 'belongs_to_entity'."""
        aid = self._norm(aspect)
        if aid not in self.G:
            return []
        ents = [
            v for _, v, d in self.G.out_edges(aid, data=True)
            if self._pred_str(d) == "belongs_to_entity"
        ]
        return sorted({self.G.nodes[e].get("label", e) for e in ents})

    def get_domains(self, entity: str) -> List[str]:
        """Domains linked from an entity via 'is_relevant_to_domain'."""
        eid = self._norm(entity)
        if eid not in self.G:
            return []
        doms = [
            v for _, v, d in self.G.out_edges(eid, data=True)
            if self._pred_str(d) == "is_relevant_to_domain"
        ]
        return sorted({self.G.nodes[d].get("label", d) for d in doms})

    # ------------------- Persistence -------------------
    def save_pickle(self, path: str | Path):
        import pickle
        with open(path, "wb") as f:
            pickle.dump({"graph": self.G, "meta": self.meta}, f)

    @classmethod
    def load_pickle(cls, path: str | Path) -> "KnowledgeGraph":
        import pickle
        with open(path, "rb") as f:
            payload = pickle.load(f)
        kg = cls()
        kg.G = payload["graph"]
        kg.meta = payload.get("meta", {})
        # rebuild label caches
        for n, d in kg.G.nodes(data=True):
            label = d.get("label", n)
            kg._id2label[n] = label
            if label not in kg._label2id:
                kg._label2id[label] = n
        return kg
```

```python
def post_process_aspects_list(aspects, original_sentence=None,
                              enforce_verbatim=False,
                              sim_dedupe_threshold=0.80,
                              lexicon=None,
                              empty_to_none=True):
    """
    Post-process a single prediction list: list[str] -> list[str]
    Uses refine_aspects(), then optionally replaces empty with ['none'].
    """
    cleaned = refine_aspects(
        aspects,
        original_sentence=original_sentence,
        enforce_verbatim=enforce_verbatim,
        sim_dedupe_threshold=sim_dedupe_threshold,
        lexicon=lexicon
    )

    if empty_to_none and (not cleaned):
        return ["none"]
    return cleaned


def post_process_pred_para_list(pred_para_list, original_sentence=None,
                                enforce_verbatim=False,
                                sim_dedupe_threshold=0.80,
                                lexicon=None,
                                empty_to_none=True):
    """
    Post-process pred_para_list which is list[list[str]].
    Returns list[list[str]] with cleaning applied to each inner list.
    """
    if pred_para_list is None:
        return []

    # ensure it is a list of lists
    if not isinstance(pred_para_list, list):
        return []

    out = []
    for lst in pred_para_list:
        # each lst should be list[str]
        if lst is None:
            lst = []
        elif not isinstance(lst, list):
            # if somehow a string slipped in, wrap it
            lst = [lst]

        out.append(
            post_process_aspects_list(
                lst,
                original_sentence=original_sentence,
                enforce_verbatim=enforce_verbatim,
                sim_dedupe_threshold=sim_dedupe_threshold,
                lexicon=lexicon,
                empty_to_none=empty_to_none
            )
        )
    return out
```

```python
import re

def normalize_aspect(aspect: str) -> str:
    """
    Basic normalization to reduce duplicates:
    - lowercase
    - remove surrounding punctuation
    - strip plural 's' (English simple rule)
    - strip extra spaces
    """
    asp = aspect.lower().strip()
    asp = re.sub(r"^\W+|\W+$", "", asp)   # remove punctuation at edges
    if asp.endswith("s") and len(asp) > 3:  # crude singularization
        asp = asp[:-1]
    return asp


def filter_to_sentence(aspects, original_sentence):
    """
    Keep only aspects that appear in the original sentence (case-insensitive).
    """
    sent = original_sentence.lower()
    return [a for a in aspects if a.lower() in sent]


# ------------------ NEW HELPERS FOR IG ------------------ #

def _ig_norm_for_overlap(s: str) -> str:
    """
    Normalize aspect strings for overlap comparison.
    Prefer the global _normalize_for_verbatim if available
    (already defined in the paraphrase component), otherwise
    fall back to simple lowercase + whitespace collapse.
    """
    if "_normalize_for_verbatim" in globals():
        return _normalize_for_verbatim(s)
    s = s.lower()
    s = re.sub(r"\s+", " ", s).strip()
    return s


def _ig_is_valid_span(asp: str) -> bool:
    """
    POS-based filter for IG candidates:
      - Reject single-token ADJ / ADV / VERB / AUX (pure adjectives/actions).
      - Require at least one NOUN/PROPN somewhere in the span.
    If spaCy nlp is not available for any reason, we fall back to keeping the span.
    """
    asp = asp.strip()
    if not asp:
        return False

    # If spaCy isn't available, don't block anything here.
    if "nlp" not in globals():
        return True

    doc = nlp(asp)

    # Single-word adjective/verb/etc → drop (e.g. "speedy", "cooked", "works")
    if len(doc) == 1 and doc[0].pos_ in {"ADJ", "ADV", "VERB", "AUX"}:
        return False

    # Require at least one noun/proper noun somewhere in the phrase
    has_noun = any(t.pos_ in {"NOUN", "PROPN"} for t in doc)
    if not has_noun:
        return False

    return True


def _ig_overlap_cleanup(aspects):
    """
    Simple overlap cleanup inside IG predictions:
      - If one normalized aspect is a substring of another, keep only the longer one.
      - Uses _ig_norm_for_overlap for robust comparison.
    This is purely intra-IG (does not look at baseline/KG).
    """
    if not aspects:
        return []

    norms = [_ig_norm_for_overlap(a) for a in aspects]
    keep = [True] * len(aspects)

    for i, ni in enumerate(norms):
        if not keep[i] or not ni:
            continue
        for j, nj in enumerate(norms):
            if i == j or not keep[j] or not nj:
                continue
            # substring relation → drop the shorter surface form
            if ni in nj or nj in ni:
                if len(aspects[i]) < len(aspects[j]):
                    keep[i] = False
                    break
                else:
                    keep[j] = False

    return [asp for asp, k in zip(aspects, keep) if k]


# ------------------ UPDATED IG COMPONENT ------------------ #

def iterative_residual_decoding2(sentence, max_rounds=3):
    """
    Iterative Residual Decoding with:
      - filtering to aspects that appear in the original sentence
      - POS-based removal of adjective/verb-only spans
      - simple overlap cleanup at the end
    """
    original_sentence = sentence
    remaining_sentence = sentence
    all_aspects = []
    seen_norm = set()

    for round_id in range(max_rounds):
        aspects = extract_aspect(remaining_sentence)  # returns list
        if not aspects:
            break

        # basic cleanup
        new_aspects = [a.strip() for a in aspects if a and a.strip()]
        if not new_aspects or all(a.lower() == "none" for a in new_aspects):
            break

        # keep only those that actually occur in the original sentence
        new_aspects = filter_to_sentence(new_aspects, original_sentence)

        # POS-based filtering: drop adjective-only / verb-only spans
        new_aspects = [a for a in new_aspects if _ig_is_valid_span(a)]

        if not new_aspects:
            # nothing useful this round
            break

        for asp in new_aspects:
            norm = normalize_aspect(asp)
            if norm not in seen_norm:
                seen_norm.add(norm)
                all_aspects.append(asp)  # keep original surface form

            # remove exact form from remaining sentence (case-insensitive)
            pattern = r"\b{}\b".format(re.escape(asp))
            remaining_sentence = re.sub(
                pattern, "", remaining_sentence, flags=re.IGNORECASE
            ).strip()

        if not remaining_sentence:
            break

    # drop literal "none" unless it's the only thing we ever saw
    final_aspects = [
        a for a in all_aspects
        if a.lower() != "none" or len(all_aspects) == 1
    ]

    # overlap cleanup inside IG outputs
    final_aspects = _ig_overlap_cleanup(final_aspects)

    return final_aspects
```

```python
import pandas as pd
from collections import Counter
from sentence_transformers import SentenceTransformer, util
import re, spacy
import math

# load once (global)
_sbert = SentenceTransformer("all-MiniLM-L6-v2")
_nlp = spacy.load("en_core_web_sm")

# ------------ helpers (unchanged) ------------
def _noun_head_lemma(text: str) -> str:
    doc = _nlp(text.lower().strip())
    heads = [t.lemma_ for t in doc if t.pos_ in ("NOUN","PROPN")]
    return heads[-1] if heads else text.lower().strip()


def min_consensus_for_k(k: int) -> int:
    return max(math.floor(k / 3), 1)

def _canonicalize_list(lst):
    return [(_noun_head_lemma(a), a) for a in lst]  # (canon, raw)

def _consensus_on_canonical(aspects_by_input, min_votes=2):
    """Vote using canonical (noun-head lemma) forms, across lists of aspects."""
    cnt = Counter()
    for lst in aspects_by_input:
        canon_set = set(_noun_head_lemma(a) for a in lst)
        for c in canon_set:
            cnt[c] += 1
    return {c for c, v in cnt.items() if v >= min_votes}

def _normalize_for_verbatim(s: str) -> str:
    s = s.lower()
    s = re.sub(r"[\"'`]+", "", s)
    s = re.sub(r"[\(\)\[\]\{\},;:]+", " ", s)
    s = re.sub(r"\s+", " ", s).strip()
    return s

def _explicitness_filter(aspects, original_sentence, jaccard_threshold=0.5):
    sent_norm = _normalize_for_verbatim(original_sentence)
    sent_tokens = set(sent_norm.split())

    kept = []
    for a in aspects:
        low = _normalize_for_verbatim(a)
        if not low:
            continue
        if low in sent_norm:
            kept.append(a); continue
        head = _noun_head_lemma(a)
        if head and head in sent_norm:
            kept.append(a); continue
        a_tokens = set(low.split())
        inter = len(a_tokens & sent_tokens)
        union = len(a_tokens | sent_tokens) or 1
        jacc = inter / union
        if jacc >= jaccard_threshold:
            kept.append(a)
    return kept

def _dedup_preserve(seq):
    seen, out = set(), []
    for x in seq:
        k = _normalize_for_verbatim(x)
        if k not in seen and k:
            seen.add(k)
            out.append(x)
    return out


# ------------ main (edited) ------------
# def extract_aspects_with_paraphrases(
#     sentences,
#     top_k=2,
#     similarity_threshold=0.58,
#     min_consensus=1,
#     include_original_always=True,
#     jaccard_explicitness=0.50,
#     verbose=False,
# ):
#     """
#     Edited version:
#     - sentences must be a list of 11 strings: [original, para1, ..., para10]
#     - runs iterative_residual_decoding ONCE per sentence (11 calls total)
#     - then materializes predictions for k in {0,2,4,6,8,10} WITHOUT extra model calls
#     - returns a 1-row DataFrame with 6 columns:
#         k_0Preds, k_2Preds, k_4Preds, k_6Preds, k_8Preds, k_10Preds
#     """
#     # -------- validation --------
#     if not sentences or not isinstance(sentences, list):
#         return pd.DataFrame([{
#             "k_0Preds": [], "k_2Preds": [], "k_4Preds": [],
#             "k_6Preds": [], "k_8Preds": [], "k_10Preds": []
#         }])

#     original_sentence = sentences[0]
#     paraphrases = sentences[1:]  # could be <10, but expected 10
#     # Ensure we only consider up to 10 paraphrases
#     paraphrases = paraphrases[:10]

#     # -------- 1) EXPENSIVE ONCE: iterative decoding for original + each paraphrase --------
#     per_input = []
#     for s in [original_sentence] + paraphrases:
#         per_input.append(iterative_residual_decoding(s, max_rounds=2))

#     pred_orig = per_input[0]         # list[str]
#     pred_para_list = per_input[1:]   # list[list[str]] length up to 10

#     # Prepare pinned originals once
#     pinned_originals = _dedup_preserve(pred_orig) if include_original_always else []

#     # -------- helper: aggregate for a given k (CHEAP) --------
#     def _aggregate_for_k(k: int):

#         # k=0 means strictly original-only, no paraphrase logic
#         if k == 0 or not pred_para_list:
#             return _dedup_preserve(pinned_originals)

#         para_lists_k = pred_para_list[:k]

#         # 3) paraphrase-derived candidates
#         # 3a) consensus on canonical FOR PARAPHRASES ONLY
#         # A. Here, the min_votes changes accoring to the value of k:
#         mc = min_consensus_for_k(k)
#         voted_canon = _consensus_on_canonical(para_lists_k, min_votes=mc)
#         # B.  the min_consensus is fixed as received in parameters:
#         # voted_canon = _consensus_on_canonical(para_lists_k, min_votes=min_consensus)

#         # map canon -> representative raw (prefer longest)
#         canon_to_raw = {}
#         for lst in para_lists_k:
#             for c, raw in _canonicalize_list(lst):
#                 if c in voted_canon:
#                     if c not in canon_to_raw or len(raw) > len(canon_to_raw[c]):
#                         canon_to_raw[c] = raw

#         candidates = list(canon_to_raw.values())

#         # fallback: if no candidates survived voting, use union of paraphrases (subset)
#         if not candidates:
#             candidates = list({a for lst in para_lists_k for a in lst})

#         # 3b) similarity filter & rank vs original
#         kept_from_para = []
#         if candidates:
#             orig_emb = _sbert.encode(original_sentence, convert_to_tensor=True)
#             asp_embs = _sbert.encode(candidates, convert_to_tensor=True)
#             sims = util.cos_sim(asp_embs, orig_emb).cpu().numpy().flatten().tolist()
#             ranked = sorted(zip(candidates, sims), key=lambda x: x[1], reverse=True)
#             ranked = [(a, s) for a, s in ranked if s >= similarity_threshold]
#             if top_k is not None:
#                 ranked = ranked[:top_k]
#             kept_from_para = [a for a, _ in ranked]

#         # 3c) explicitness vs original (soft)
#         kept_from_para = _explicitness_filter(
#             kept_from_para, original_sentence, jaccard_threshold=jaccard_explicitness
#         )

#         # 4) union + dedup
#         final = _dedup_preserve(pinned_originals + kept_from_para)

#         if verbose:
#             print(f"[PARA-k={k}] orig={len(pinned_originals)}, "
#                   f"para_cands={len(candidates)}, kept_para={len(kept_from_para)}, final={len(final)}")
#         return final

#     # -------- 2) Materialize all requested k columns --------
#     out = {
#         "k_0Preds":  _aggregate_for_k(0),
#         "k_2Preds":  _aggregate_for_k(2),
#         "k_4Preds":  _aggregate_for_k(4),
#         "k_6Preds":  _aggregate_for_k(6),
#         "k_8Preds":  _aggregate_for_k(8),
#         "k_10Preds": _aggregate_for_k(10),
#     }

#     return pd.DataFrame([out])
def extract_aspects_with_paraphrases(
    sentences,
    top_k=2,
    similarity_threshold=0.58,
    include_original_always=True,
    jaccard_explicitness=0.50,
    verbose=False,
):
    """
    - sentences: list[str] expected [original, para1, ..., para10]
    - Runs iterative_residual_decoding ONCE per input sentence (original + paraphrases)
    - Caches:
        pred_orig: list[str]
        pred_para_list: list[list[str]] length up to 10
    - Materializes predictions for k in {0,2,4,6,8,10} WITHOUT extra model calls
    - Returns a 1-row DataFrame with:
        pred_orig, pred_para_list,
        k_0Preds, k_2Preds, k_4Preds, k_6Preds, k_8Preds, k_10Preds
    """
    # -------- validation --------
    empty = {
        "pred_orig": [],
        "pred_para_list": [],
        "k_0Preds": [], "k_2Preds": [], "k_4Preds": [],
        "k_6Preds": [], "k_8Preds": [], "k_10Preds": []
    }
    if not sentences or not isinstance(sentences, list):
        return pd.DataFrame([empty])

    original_sentence = sentences[0]
    paraphrases = sentences[1:] if len(sentences) > 1 else []
    paraphrases = paraphrases[:10]  # enforce max 10

    # -------- 1) EXPENSIVE ONCE: iterative decoding for original + each paraphrase --------
    per_input = []
    for s in [original_sentence] + paraphrases:
        per_input.append(iterative_residual_decoding(s, max_rounds=2))

    pred_orig = per_input[0]          # list[str]
    pred_para_list = per_input[1:]    # list[list[str]] length up to 10

    # PIN original aspects (if enabled)
    pinned_originals = _dedup_preserve(pred_orig) if include_original_always else []

    # -------- helper: aggregate for a given k (CHEAP) --------
    def _aggregate_for_k(k: int):
        # k=0 means strictly original-only, no paraphrase logic
        if k == 0 or not pred_para_list:
            return _dedup_preserve(pinned_originals)

        para_lists_k = pred_para_list[:k]

        # A) dynamic min_consensus based on k
        mc = min_consensus_for_k(k)

        # 3a) consensus on canonical FOR PARAPHRASES ONLY
        voted_canon = _consensus_on_canonical(para_lists_k, min_votes=mc)

        # map canon -> representative raw (prefer longest)
        canon_to_raw = {}
        for lst in para_lists_k:
            for c, raw in _canonicalize_list(lst):
                if c in voted_canon:
                    if c not in canon_to_raw or len(raw) > len(canon_to_raw[c]):
                        canon_to_raw[c] = raw

        candidates = list(canon_to_raw.values())

        # fallback: if no candidates survived voting, use union of paraphrases (subset)
        if not candidates:
            candidates = list({a for lst in para_lists_k for a in lst})

        # 3b) similarity filter & rank vs original
        kept_from_para = []
        if candidates:
            orig_emb = _sbert.encode(original_sentence, convert_to_tensor=True)
            asp_embs = _sbert.encode(candidates, convert_to_tensor=True)
            sims = util.cos_sim(asp_embs, orig_emb).cpu().numpy().flatten().tolist()
            ranked = sorted(zip(candidates, sims), key=lambda x: x[1], reverse=True)
            ranked = [(a, s) for a, s in ranked if s >= similarity_threshold]
            if top_k is not None:
                ranked = ranked[:top_k]
            kept_from_para = [a for a, _ in ranked]

        # 3c) explicitness vs original (soft)
        kept_from_para = _explicitness_filter(
            kept_from_para, original_sentence, jaccard_threshold=jaccard_explicitness
        )

        # 4) union + dedup (originals protected)
        final = _dedup_preserve(pinned_originals + kept_from_para)

        if verbose:
            print(f"[PARA-k={k}] mc={mc} orig={len(pinned_originals)}, "
                  f"cands={len(candidates)}, kept_para={len(kept_from_para)}, final={len(final)}")

        return final

    # -------- 2) Materialize all requested k columns --------
    out = {
        "pred_orig": pred_orig,
        "pred_para_list": pred_para_list,

        "k_0Preds":  _aggregate_for_k(0),
        "k_2Preds":  _aggregate_for_k(2),
        "k_4Preds":  _aggregate_for_k(4),
        "k_6Preds":  _aggregate_for_k(6),
        "k_8Preds":  _aggregate_for_k(8),
        "k_10Preds": _aggregate_for_k(10),
    }

    return pd.DataFrame([out])
```

##### Post-processing refinements:

```python
import re
from difflib import SequenceMatcher
import spacy
nlp = spacy.load("en_core_web_sm")

NUMERIC_RE = re.compile(r"^[\d\.\,]+%?$")

import re

# strip all unwanted special characters anywhere in the text
PUNCTUATION_STRIP_RE = re.compile(r"[\-\_\!\@\#\$\%\^\&\*\(\)\+\=\?\.\,<>\"';:\[\]\{\}]")
MULTISPACE_RE = re.compile(r"\s+")

def clean_text(text: str) -> str:
    # remove special characters
    cleaned = PUNCTUATION_STRIP_RE.sub("", text)
    # collapse multiple spaces to one
    cleaned = MULTISPACE_RE.sub(" ", cleaned)
    # trim leading/trailing spaces
    return cleaned.strip()

def is_nouny_text(text):
    doc = nlp(text)
    return any(t.pos_ in ("NOUN","PROPN") for t in doc)

def canonicalize(aspect, lexicon=None):
    doc = nlp(aspect.lower().strip())
    heads = [t.lemma_ for t in doc if t.pos_ in ("NOUN","PROPN")]
    base = heads[-1] if heads else aspect.lower().strip()
    if lexicon and base in lexicon:
        return lexicon[base]
    return base

def refine_aspects(aspects, original_sentence=None, enforce_verbatim=False,
                   sim_dedupe_threshold=0.80, lexicon=None):
    if not aspects or not isinstance(aspects, list):
        return []

    # 1) clean + POS
    pool = []

    for a in aspects:
        a = str(a).strip()
        if not a or NUMERIC_RE.match(a):
            continue
        a = clean_text(a)
        if not is_nouny_text(a):
            continue
        pool.append(a)

    # 2) enforce explicitness vs original (optional, recommended)
    if enforce_verbatim and original_sentence:
        keep = []
        s = original_sentence.lower()
        for a in pool:
            low = a.lower()
            if low in s:
                keep.append(a)
                continue
            # head-noun fallback
            doc = nlp(a)
            head = next((t.text for t in doc if t.pos_ in ("NOUN","PROPN")), doc[-1].text)
            if head.lower() in s:
                keep.append(a)
        pool = keep

    if not pool:
        return []

    # 3) canonicalize
    canon_pairs = [(a, canonicalize(a, lexicon)) for a in pool]

    # 4) dedup by canonical form
    seen, refined = set(), []
    for raw, can in canon_pairs:
        if can not in seen:
            seen.add(can)
            refined.append(raw)

    # 5) optional fuzzy collapse of near-duplicates (token-set-ish)
    final = []
    for cand in refined:
        add = True
        for ex in final:
            sim = SequenceMatcher(None, cand.lower(), ex.lower()).ratio()
            if sim >= sim_dedupe_threshold:
                # prefer shorter canonical (tend to be more general)
                if len(canonicalize(cand)) < len(canonicalize(ex)):
                    final.remove(ex)
                    final.append(cand)
                add = False
                break
        if add:
            final.append(cand)

    return final
```

```python
import pandas as pd


def evaluate(df, gold_col, prediction_col, match_type="partial", sep_token="[sep]"):
    def normalize(text):
        return text.lower().strip()

    def is_match(pred, gold_terms, exact=False):
        pred = normalize(pred)
        for gold in gold_terms:
            gold = normalize(gold)
            if exact:
                if pred == gold:
                    return True
            else:
                if pred in gold or gold in pred:
                    return True
        return False

    exact = (match_type == "exact")

    TP = FP = FN = 0

    for _, row in df.iterrows():
        gold = row.get(gold_col, None)
        pred_str = row.get(prediction_col, "")
        # if pred_str is list, just retrieve the first item:
        # if isinstance(pred_str, list):
        #     pred_str = pred_str[0]
        #     # new (safe):
        if isinstance(pred_str, list):
            pred_str = next((x for x in pred_str if isinstance(x, str) and x.strip()), "none")

        if not isinstance(gold, list):
            continue  # Skip rows with no aspectTerms

        gold_terms = [normalize(t) for t in gold]

        if not isinstance(pred_str, str) or pred_str.strip() == "":
            FN += len(gold_terms)
            continue

        predicted_terms = [normalize(p) for p in pred_str.split(sep_token) if p.strip()]
        matched_gold = set()

        for pred in predicted_terms:
            if is_match(pred, gold_terms, exact=exact):
                TP += 1
                for i, gold in enumerate(gold_terms):
                    if i not in matched_gold and (pred == gold if exact else pred in gold or gold in pred):
                        matched_gold.add(i)
                        break
            else:
                FP += 1

        FN += len(gold_terms) - len(matched_gold)

    precision = TP / (TP + FP) if (TP + FP) > 0 else 0.0
    recall = TP / (TP + FN) if (TP + FN) > 0 else 0.0
    f1 = 2 * precision * recall / (precision + recall) if (precision + recall) > 0 else 0.0
    accuracy = TP / (TP + FP + FN) if (TP + FP + FN) > 0 else 0.0

    return {
        "Match Type": match_type,
        "Precision": round(precision * 100, 2),
        "Recall": round(recall * 100, 2),
        "F1": round(f1 * 100, 2),
        "Accuracy": round(accuracy * 100, 2),
        "TP": TP,
        "FP": FP,
        "FN": FN,
    }

print(f"baseline predictions performance           : { evaluate(test,'aspect','baseline', match_type='partial')}")
print(f"explicit_iterative predictions performance           : { evaluate(test,'aspect','explicit_iterative', match_type='partial')}")
# print(f"explicit_paraphrases predictions performance           : { evaluate(test,'aspect','explicit_paraphrases', match_type='partial')}")
print(f"implicit_kg predictions performance           : { evaluate(test,'aspect','implicit_kg', match_type='partial')}")
print(f"Explicit and implicit predictions performance           : { evaluate(test,'aspect','combined_aspects', match_type='partial')}")
print(f"cleaned_baseline predictions performance           : { evaluate(test,'aspect','cleaned_baseline', match_type='partial')}")
print(f"cleaned_explicit_iterative predictions performance           : { evaluate(test,'aspect','cleaned_explicit_iterative', match_type='partial')}")
print(f"cleaned_explicit_paraphrases predictions performance           : { evaluate(test,'aspect','cleaned_explicit_paraphrases', match_type='partial')}")
print(f"cleaned_implicit_kg predictions performance           : { evaluate(test,'aspect','cleaned_implicit_kg', match_type='partial')}")
print(f"cleaned_Explicit and implicit predictions performance           : { evaluate(test,'aspect','cleaned_combined_aspects', match_type='partial')}")

print(f"Explicit predictions performance           : { evaluate(test_explicit,'aspect','prediction', match_type='partial')}")
print(f"Implicit predictions performance           : { evaluate(test_implicit,'aspect','prediction', match_type='partial')}")
# print(f"Objective predictions performance           : { evaluate(test_objective,'aspect','prediction', match_type='partial')}")

# print(f"iterative_predictions performance : { evaluate(test,'target_en','iterative_predictions')}")
# print(f"iterative_predictions performance : { evaluate(test,'target_en','predicted_paraphrased_aspects')}")
```

2. parse gold aspects

3. compute precision/recall/F1 per row and overall (micro & macro)

```python
def evaluate_dataframe(
    df: pd.DataFrame,
    pred_col: str = "preidction",
    gold_col: str = "target",
    use_camel: bool = False,
    jaccard_threshold: float = 0.6
):
    df = df.copy()
    df["pred_list_clean"] = df[pred_col].apply(lambda s: clean_predicted_aspects(s, use_camel=use_camel))

    def parse_gold(x):
        if isinstance(x, list):
            return x
        return parse_aspect_string(x)
    df["gold_list"] = df[gold_col].apply(parse_gold)

    per = df.apply(
        lambda r: prf1_for_example(r["gold_list"], r["pred_list_clean"], thresh=jaccard_threshold), axis=1
    )
    df["precision"] = per.apply(lambda m: m["precision"])
    df["recall"]    = per.apply(lambda m: m["recall"])
    df["f1"]        = per.apply(lambda m: m["f1"])
    df["tp"]        = per.apply(lambda m: m["tp"])
    df["fp"]        = per.apply(lambda m: m["fp"])
    df["fn"]        = per.apply(lambda m: m["fn"])
    # number of gold items in each example
    df["support"]   = df["gold_list"].apply(len)

    # -------- micro --------
    tot_tp, tot_fp, tot_fn = df["tp"].sum(), df["fp"].sum(), df["fn"].sum()
    micro_p = tot_tp / (tot_tp + tot_fp) if (tot_tp + tot_fp) else 0.0
    micro_r = tot_tp / (tot_tp + tot_fn) if (tot_tp + tot_fn) else 0.0
    micro_f = 2*micro_p*micro_r/(micro_p+micro_r) if (micro_p+micro_r) else 0.0

    # -------- macro --------
    macro_p = df["precision"].mean() if len(df) else 0.0
    macro_r = df["recall"].mean()    if len(df) else 0.0
    macro_f = df["f1"].mean()        if len(df) else 0.0

    # -------- weighted macro --------
    total_support = df["support"].sum()
    if total_support > 0:
        weighted_p = (df["precision"] * df["support"]).sum() / total_support
        weighted_r = (df["recall"]    * df["support"]).sum() / total_support
        weighted_f = (df["f1"]        * df["support"]).sum() / total_support
    else:
        weighted_p = weighted_r = weighted_f = 0.0

    summary = {
        "micro": {
            "precision": round(float(micro_p), 4),
            "recall": round(float(micro_r), 4),
            "f1": round(float(micro_f), 4),
            "accuracy": round(float(tot_tp / (tot_tp + tot_fp + tot_fn)), 4),
            "tp": int(tot_tp),
            "fp": int(tot_fp),
            "fn": int(tot_fn),
        },
        "macro": {
            "precision": round(float(macro_p), 4),
            "recall": round(float(macro_r), 4),
            "f1": round(float(macro_f), 4),
            # optional: add macro accuracy if you compute per-class accuracies
        },
        "weighted_macro": {
            "precision": round(float(weighted_p), 4),
            "recall": round(float(weighted_r), 4),
            "f1": round(float(weighted_f), 4),
            # optional: weighted accuracy (same logic)
        },
        "df": df,
    }


    return summary
```
