```python
# ================================================================
# English KG Lexical Inference — MULTI-DATASET VERSION
# ================================================================

# ── 0. Installs ──────────────────────────────────────────────────
# !pip install -q sympy==1.12
# !pip install -q spacy sentence-transformers pandas tqdm
# !python -m spacy download en_core_web_sm

# ================================================================
# 1. Imports
# ================================================================
import json
import re
import ast
import spacy
import numpy as np
import pandas as pd
from collections import defaultdict
from typing import Dict, List, Set, Tuple, Optional
from tqdm import tqdm
from sentence_transformers import SentenceTransformer
from google.colab import drive, files

tqdm.pandas()
drive.mount('/content/drive')
print("✅ Imports done.")

# ================================================================
# 2. Paths
# ================================================================
KG_PATH = "/content/drive/MyDrive/colab_files/Knowledge_graph_mabsaEnriched.json"

DATASET_PATHS = {
#     "SemEval15":       "/content/drive/MyDrive/colab_files/test_15.csv",
#     "SemEval16":       "/content/drive/MyDrive/colab_files/test_16.csv",
#     "ACOS_restaurant": "/content/drive/MyDrive/colab_files/test_acos.csv",
#     "ACOS_laptop":     "/content/drive/MyDrive/colab_files/test_acos.csv",
    # "MABSA_EN":        "/content/drive/MyDrive/colab_files/m-absa_English_test.csv",
      "SemEval15":       "/content/full_predictions_se15.csv",
      "SemEval16":       "/content/full_predictions_se16.csv",
      # "ACOS_restaurant": "/content/drive/MyDrive/colab_files/test_acos.csv",
      "ACOS_laptop":     "/content/full_predictions_acos.csv",
      "ACOS_restaurant": "/content/full_predictions_acos.csv",
      "MABSA_EN":        "/content/drive/MyDrive/colab_files/m-absa_English_test.csv",

}

# ================================================================
# 3. Load models
# ================================================================
print("\nLoading spaCy English pipeline...")
_en_nlp = spacy.load("en_core_web_sm")
print("✅ spaCy ready.")

print("Loading sentence encoder...")
encoder = SentenceTransformer(
    "sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2"
)
print("✅ Encoder ready.")

# ================================================================
# 4. Normalisation helper
# ================================================================
def _norm(s: str) -> str:
    if not isinstance(s, str):
        return ""
    return re.sub(r"\s+", " ", s.strip().lower())

# ================================================================
# 5. Load and index the KG
#    - describes_aspect only (indicates_implicit_aspect disabled)
#    - all domain names normalised to lowercase at index time
# ================================================================
def load_english_kg(kg_path: str) -> Dict:
    with open(kg_path, "r", encoding="utf-8") as f:
        raw = json.load(f)

    clue_to_aspects: Dict[str, List[Tuple[str, str]]] = defaultdict(list)
    for r in raw["relationships"]:
        if r["predicate"] == "describes_aspect":
            clue = _norm(r["subject"])
            asp  = _norm(r["object"])
            if clue and asp:
                clue_to_aspects[clue].append((asp, r["predicate"]))

    aspect_to_entities: Dict[str, Set[str]] = defaultdict(set)
    for a in raw["aspects"]:
        asp = _norm(a["name"])
        for e in a.get("entity", []):
            aspect_to_entities[asp].add(_norm(e))

    entity_to_domain: Dict[str, str] = {}
    for e in raw["entities"]:
        entity_to_domain[_norm(e["name"])] = e["domain"].strip().lower()

    domain_to_entities: Dict[str, Set[str]] = defaultdict(set)
    for ent, dom in entity_to_domain.items():
        domain_to_entities[dom].add(ent)

    aspect_to_domains: Dict[str, Set[str]] = defaultdict(set)
    for asp, ents in aspect_to_entities.items():
        for ent in ents:
            dom = entity_to_domain.get(ent)
            if dom:
                aspect_to_domains[asp].add(dom)

    print(f"✅ English KG loaded (describes_aspect only):")
    print(f"   Clues   : {len(clue_to_aspects)}")
    print(f"   Aspects : {len(aspect_to_entities)}")
    print(f"   Entities: {len(entity_to_domain)}")
    print(f"   Domains : {sorted(domain_to_entities.keys())}")

    return {
        "raw":                raw,
        "clue_to_aspects":    dict(clue_to_aspects),
        "aspect_to_entities": dict(aspect_to_entities),
        "entity_to_domain":   entity_to_domain,
        "domain_to_entities": dict(domain_to_entities),
        "aspect_to_domains":  dict(aspect_to_domains),
    }

kg_index = load_english_kg(KG_PATH)

# ================================================================
# 6. Domain configurations
# ================================================================

DOMAIN_VOCAB_RESTRICTED = {
    "restaurant": [
        "food", "meal", "restaurant", "menu", "waiter", "service",
        "dish", "delicious", "portion", "taste", "chef", "cuisine",
        "dining", "reservation", "appetizer", "dessert", "drink",
        "fresh", "flavor", "ambience", "table", "order", "cook",
    ],
    "food": [
        "food", "ingredient", "recipe", "grocery", "product", "flavor",
        "taste", "package", "nutrition", "diet", "snack", "meal",
        "fresh", "organic", "delivery", "brand", "quality", "price",
        "portion", "yummy", "delicious", "consume", "eat",
    ],
    "hotel": [
        "hotel", "room", "check-in", "checkout", "staff", "reception",
        "bed", "bathroom", "pool", "lobby", "breakfast", "housekeeping",
        "accommodation", "stay", "booking", "amenity", "facility",
        "clean", "comfortable", "view", "floor", "elevator", "towel",
    ],
    "laptop": [
        "laptop", "computer", "keyboard", "screen", "battery", "processor",
        "RAM", "storage", "display", "charger", "performance", "speed",
        "software", "hardware", "port", "wifi", "boot", "mouse",
        "fan", "temperature", "weight", "design",
    ],
}

DOMAIN_ALIASES_RESTRICTED: Dict[str, Set[str]] = {
    "restaurant": {"restaurant", "food"},
    "food":       {"food", "restaurant"},
    "hotel":      {"hotel"},
    "laptop":     {"laptop"},
}

DOMAIN_VOCAB_FULL = {
    **DOMAIN_VOCAB_RESTRICTED,
    "phone": [
        "phone", "smartphone", "camera", "screen", "battery", "charging",
        "app", "signal", "call", "speaker", "microphone", "display",
        "RAM", "storage", "fingerprint", "selfie", "network", "sim",
        "bluetooth", "processor", "performance", "case", "price",
    ],
    "books": [
        "book", "author", "chapter", "story", "plot", "character",
        "writing", "read", "novel", "narrative", "genre", "publish",
        "edition", "review", "content", "cover", "series", "language",
    ],
}

DOMAIN_ALIASES_FULL: Dict[str, Set[str]] = {
    **DOMAIN_ALIASES_RESTRICTED,
    "phone": {"phone"},
    "books": {"books"},
}

DATASET_DOMAIN_CONFIG: Dict[str, Tuple[Dict, Dict]] = {
    "SemEval15":       (DOMAIN_VOCAB_RESTRICTED, DOMAIN_ALIASES_RESTRICTED),
    "SemEval16":       (DOMAIN_VOCAB_RESTRICTED, DOMAIN_ALIASES_RESTRICTED),
    "ACOS_restaurant": (DOMAIN_VOCAB_RESTRICTED, DOMAIN_ALIASES_RESTRICTED),
    "ACOS_laptop":     (DOMAIN_VOCAB_RESTRICTED, DOMAIN_ALIASES_RESTRICTED),
    "MABSA_EN":        (DOMAIN_VOCAB_FULL,        DOMAIN_ALIASES_FULL),
}

_GOLD_DOMAIN_MAP: Dict[str, str] = {
    "restaurants": "restaurant",
    "restaurant":  "restaurant",
    "laptops":     "laptop",
    "laptop":      "laptop",
    "hotels":      "hotel",
    "hotel":       "hotel",
    "phone":       "phone",
    "phones":      "phone",
    "books":       "books",
    "book":        "books",
}

_ACOS_DOMAIN_FILTER: Dict[str, str] = {
    "ACOS_restaurant": "restaurants",
    "ACOS_laptop":     "laptops",
}

# ================================================================
# 7. Domain inference — vocabulary centroid method
# ================================================================
_domain_embeddings: Dict[str, np.ndarray] = {}

def build_domain_index(encoder, domain_vocab: Dict) -> None:
    global _domain_embeddings
    _domain_embeddings = {}
    for domain, vocab in domain_vocab.items():
        word_embs = encoder.encode(
            vocab, normalize_embeddings=True,
            convert_to_numpy=True, show_progress_bar=False,
        )
        centroid = word_embs.mean(axis=0)
        centroid = centroid / np.linalg.norm(centroid)
        _domain_embeddings[domain] = centroid
    print(f"✅ Domain index built: {list(_domain_embeddings.keys())}")

def infer_domain_from_sentence(sentence: str, encoder) -> str:
    if not sentence or not _domain_embeddings:
        return "unknown"
    sent_emb = encoder.encode(
        sentence, normalize_embeddings=True, convert_to_numpy=True
    )
    best_domain, best_sim = "unknown", -1.0
    for domain, centroid in _domain_embeddings.items():
        sim = float(np.dot(sent_emb, centroid))
        if sim > best_sim:
            best_sim    = sim
            best_domain = domain
    return best_domain

# ================================================================
# 8. Lonely adjective detection
#
# Pattern 1 fix — PRON nsubj relaxation:
#   OLD rule: block acomp if ANY nsubj exists (noun OR pronoun)
#   NEW rule: block acomp if nsubj is NOUN/PROPN (named entity = explicit)
#             ALLOW acomp if nsubj is PRON AND pronoun is anaphoric
#             (it/they/this/these/them = product reference = implicit)
#             BLOCK acomp if nsubj is first-person PRON (I/we/he/she)
#             because "I am happy" refers to the speaker, not the product
# ================================================================

GENERIC_ABSTRACT_NOUNS: Set[str] = {
    "thing", "reason", "point", "fact", "issue", "problem",
    "aspect", "feature", "part", "side", "element", "detail",
    "turnoff", "downside", "highlight", "drawback", "plus",
    "minus", "pro", "con", "disappointment", "surprise", "bonus",
    "place", "spot", "venue", "area", "neighborhood", "district",
    "find", "gem", "choice", "option", "pick", "destination",
    "way", "manner", "approach", "job", "work",
    "everything", "nothing", "something", "anything",
    "experience", "visit", "trip", "stay", "time", "moment",
}

GENERIC_SENT_EN: Set[str] = {
    "good", "great", "excellent", "amazing", "awesome", "fantastic",
    "wonderful", "superb", "outstanding", "terrific", "brilliant",
    "perfect", "decent", "solid", "fine", "pleasant",
    "bad", "terrible", "awful", "horrible", "dreadful", "horrid",
    "poor", "mediocre", "disappointing", "lousy",
    "favorite", "favourite", "best", "worst", "better", "worse",
    "top", "ultimate", "absolute",
    "nice", "lovely", "okay", "ok", "alright",
}

NON_OPINION_DEP_EN: Set[str] = {
    "nummod", "det", "advmod", "compound", "fixed", "flat",
}

ADJ_SURFACE_BLOCKLIST: Set[str] = {
    "back", "front", "left", "right", "top", "bottom", "upper", "lower",
    "inner", "outer", "middle", "central", "side", "near", "far",
    "current", "recent", "previous", "next", "last", "first", "second",
    "third", "early", "late", "former", "latter",
    "only", "main", "major", "minor", "total", "full", "whole", "entire",
    "single", "double", "multiple", "various", "certain", "general",
    "japanese", "italian", "french", "chinese", "mexican", "indian",
    "thai", "greek", "american", "spanish", "korean", "vietnamese",
}

# ── Pattern 1: Anaphoric pronouns — refer to the product/entity ──
# When these are the nsubj of a copula, the sentence is IMPLICIT
# because the referent is not named in the sentence.
# e.g. "it is expensive" / "they are great" / "this is horrible"
ANAPHORIC_PRONOUNS: Set[str] = {
    "it", "its", "they", "them", "their",
    "this", "these", "that", "those",
    "everything", "nothing", "something",
}

# ── First-person pronouns — refer to the speaker, NOT the product ─
# When these are the nsubj, the sentence is about the speaker's
# experience, not an implicit product aspect.
# e.g. "I am satisfied" → speaker is satisfied, not the product
# These continue to BLOCK acomp (old behaviour preserved)
FIRST_PERSON_PRONOUNS: Set[str] = {
    "i", "we", "he", "she", "me", "us", "my", "our",
}


def find_lonely_adjectives_en(sentence: str) -> List[Dict]:
    if not sentence or not isinstance(sentence, str):
        return []

    doc    = _en_nlp(sentence)
    lonely = []
    seen   = set()

    for tok in doc:
        if tok.pos_ != "ADJ":
            continue
        if tok.dep_ in NON_OPINION_DEP_EN:
            continue
        if tok.lemma_.lower() in ADJ_SURFACE_BLOCKLIST:
            continue

        head = tok.head

        # ── acomp — Pattern 1 fix ────────────────────────────────
        # Climb up if head is ADJ (conj chain) to find the predicate
        if tok.dep_ == "acomp":
            governing_verb = head
            if governing_verb.pos_ == "ADJ":
                governing_verb = governing_verb.head

            if governing_verb.pos_ in ("AUX", "VERB"):
                nsubj_tokens = [
                    c for c in governing_verb.children
                    if c.dep_ in ("nsubj", "nsubjpass")
                ]
                for subj in nsubj_tokens:
                    subj_text = subj.text.lower()
                    if subj.pos_ in ("NOUN", "PROPN"):
                        # Named noun subject → explicit sentence → block
                        # e.g. "The pasta IS incredible" → block
                        break  # will continue outer loop after block
                    elif subj.pos_ == "PRON":
                        if subj_text in FIRST_PERSON_PRONOUNS:
                            # Speaker reference → block
                            # e.g. "I am satisfied" → block
                            break
                        elif subj_text in ANAPHORIC_PRONOUNS:
                            # Product anaphora → IMPLICIT → ALLOW
                            # e.g. "it is expensive" / "they are great"
                            # Do NOT break — let the adjective through
                            subj = None  # signal: allow
                            break
                        else:
                            # Unknown pronoun → conservative: block
                            break
                else:
                    subj = None  # no nsubj found → allow

                # If subj is not None at this point → block
                if subj is not None:
                    continue

        # ── conj — block if conjoined adj is acomp of verb ───────
        # with a NOUN/PROPN subject (explicit sentence)
        # If the conjoined acomp has only PRON subject → allow
        elif tok.dep_ == "conj":
            conj_head = head
            if conj_head.pos_ == "ADJ" and conj_head.dep_ == "acomp":
                pred = conj_head.head
                if pred.pos_ in ("AUX", "VERB"):
                    nsubj_tokens = [
                        c for c in pred.children
                        if c.dep_ in ("nsubj", "nsubjpass")
                    ]
                    block = False
                    for subj in nsubj_tokens:
                        if subj.pos_ in ("NOUN", "PROPN"):
                            block = True
                            break
                        elif subj.pos_ == "PRON":
                            if subj.text.lower() in FIRST_PERSON_PRONOUNS:
                                block = True
                                break
                            # Anaphoric PRON → allow (don't block)
                    if block:
                        continue

        # ── amod — allow ONLY if head noun is generic abstract ───
        elif tok.dep_ == "amod":
            if head and head.pos_ in ("NOUN", "PROPN"):
                head_lemma = (head.lemma_ or head.text).lower().strip()
                if head_lemma not in GENERIC_ABSTRACT_NOUNS:
                    continue
            else:
                continue

        # ── Other dep_rels — reject if direct NOUN/PROPN/NUM head ─
        elif head and head.i != tok.i:
            if head.pos_ in ("NOUN", "PROPN", "NUM"):
                continue

        # No NOUN/PROPN children
        if any(c.pos_ in ("NOUN", "PROPN") for c in tok.children):
            continue

        lemma = tok.lemma_.lower().strip()
        if not lemma or lemma in seen:
            continue
        seen.add(lemma)

        lonely.append({
            "surface":    tok.text,
            "lemma":      lemma,
            "dep_rel":    tok.dep_,
            "head_text":  head.text if head else "ROOT",
            "head_upos":  head.pos_ if head else "ROOT",
            "is_generic": lemma in GENERIC_SENT_EN,
        })

    return lonely

# ================================================================
# 8b. Verb clue surface scan
#
# Pattern 2 fix — action verbs added to Tier 1:
#   "will order again", "bought again", "never buy again"
#   all express strong overall sentiment about the product
# ================================================================
VERB_CLUES_TIER1: Set[str] = {
    # Strong sentiment verbs (existing)
    "love", "loved", "loving",
    "recommend", "recommended", "recommending",
    "return", "returned", "returning",
    "comeback", "impressed", "impress",
    "satisfied", "pleased", "delighted", "thrilled",
    "obsessed", "hooked", "addicted",
    "disappoint", "disappointed", "disappointing",
    "dislike", "disliked", "regret", "regretted",
    "hate", "hated", "avoid", "avoided", "avoiding",
    "ignored", "ignore", "neglected", "neglect",
    "worth",
    "miss", "missed", "missing",
    # ── Pattern 2: action verbs expressing overall satisfaction ──
    # Positive repurchase intent → strong GENERAL signal
    "reorder", "reordered", "reordering",
    "repurchase", "repurchased",
    # "order" / "buy" / "purchase" alone are weak — only added
    # in conjunction with "again" or "more" (handled in scan below)
}

VERB_CLUES_TIER2: Set[str] = {
    "like", "liked",
    "enjoy", "enjoyed", "enjoying",
    "suggest", "suggested",
    "tried", "try",
    "waited", "wait", "waiting",
    "went", "go",
    "visit", "visited",
    # Action verbs (weaker — need "again"/"more"/"never" to fire)
    "order", "ordered", "ordering",
    "buy", "bought", "purchase", "purchased",
    "use", "used", "using",
}

NEGATION_WORDS: Set[str] = {
    "not", "never", "don't", "dont", "won't", "wont",
    "wouldn't", "wouldnt", "can't", "cant", "no",
}

# ── Words that strengthen action verbs into GENERAL signals ──────
# "will order AGAIN" / "buy MORE" → positive intent
# "NEVER buy again" → negative intent (still GENERAL)
ACTION_AMPLIFIERS: Set[str] = {
    "again", "more", "another", "always", "definitely",
    "certainly", "absolutely", "forever", "regularly",
}


def _scan_verb_clues(sentence: str) -> int:
    """
    Scans sentence for sentiment verb clues.
    Returns:
        1 — Tier 1 verb found (strong GENERAL signal)
        2 — Tier 2 verb only (weaker, context-dependent)
        0 — no sentiment verb found

    Pattern 2 fix:
        Action verbs (order/buy/purchase/use) in Tier 2 are promoted
        to Tier 1 strength when combined with an amplifier or negation,
        because:
          "will order again"  → strong positive GENERAL
          "never buy again"   → strong negative GENERAL
          "always use this"   → strong positive GENERAL
          "ordered three"     → too weak — no signal (stays Tier 2)
    """
    words = set(re.findall(r"\b\w+\b", sentence.lower()))

    # Tier 1 exact match (includes reorder/repurchase)
    if words & VERB_CLUES_TIER1:
        return 1

    # Pattern 2: action verbs promoted to Tier 1 with amplifier/negation
    action_hits = words & {"order", "ordered", "ordering",
                           "buy", "bought", "purchase", "purchased",
                           "use", "used", "using"}
    if action_hits:
        has_amplifier = bool(words & ACTION_AMPLIFIERS)
        has_negation  = bool(words & NEGATION_WORDS)
        if has_amplifier or has_negation:
            return 1   # promoted: "will order again" / "never buy"

    # Remaining Tier 2
    tier2_hits = words & VERB_CLUES_TIER2
    if tier2_hits:
        strong_tier2 = tier2_hits - {"go", "like", "liked", "tried", "try",
                                      "order", "ordered", "ordering",
                                      "buy", "bought", "purchase", "purchased",
                                      "use", "used", "using"}
        has_negation = bool(words & NEGATION_WORDS)
        has_exclaim  = "!" in sentence
        if strong_tier2:
            return 2
        if (has_negation or has_exclaim) and tier2_hits:
            return 2

    return 0

# ================================================================
# 9. Semantic index — ENTITY#ASPECT compound embeddings
#    Built ONCE — shared across all datasets
# ================================================================
def build_semantic_index(kg_index: Dict, encoder) -> Dict:
    compound_to_aspect: Dict[str, str] = {}
    compounds: List[str] = []

    for asp, ents in kg_index["aspect_to_entities"].items():
        if not ents:
            compounds.append(asp)
            compound_to_aspect[asp] = asp
        else:
            for ent in ents:
                comp = f"{ent}#{asp}"
                compounds.append(comp)
                compound_to_aspect[comp] = asp

    entities = list(kg_index["entity_to_domain"].keys())

    print("Encoding ENTITY#ASPECT compounds...")
    comp_embs = encoder.encode(
        compounds, normalize_embeddings=True,
        convert_to_numpy=True, show_progress_bar=True,
    )
    print("Encoding entities...")
    ent_embs = encoder.encode(
        entities, normalize_embeddings=True,
        convert_to_numpy=True, show_progress_bar=False,
    )

    print(f"✅ Semantic index built:")
    print(f"   Compounds : {len(compounds)}")
    print(f"   Entities  : {len(entities)}")

    return {
        "compounds":          compounds,
        "compound_to_aspect": compound_to_aspect,
        "comp_embs":          comp_embs,
        "entities":           entities,
        "ent_embs":           ent_embs,
        "ent_idx":            {e: i for i, e in enumerate(entities)},
    }

sem_index = build_semantic_index(kg_index, encoder)

# ================================================================
# 10. Aspect ranking — cosine sim using domain-filtered compounds
# ================================================================
def rank_aspects_by_similarity(
    sentence: str,
    candidate_aspects: List[str],
    allowed_doms: Set[str],
    kg_index: Dict,
    sem_index: Dict,
    encoder,
) -> List[Tuple[str, float]]:
    if not candidate_aspects:
        return []

    sent_emb = encoder.encode(
        sentence, normalize_embeddings=True, convert_to_numpy=True
    )

    compounds          = sem_index["compounds"]
    comp_embs          = sem_index["comp_embs"]
    compound_to_aspect = sem_index["compound_to_aspect"]

    asp_to_indices: Dict[str, List[int]] = defaultdict(list)
    for i, comp in enumerate(compounds):
        asp = compound_to_aspect[comp]
        if asp not in candidate_aspects:
            continue
        if '#' in comp:
            ent     = comp.split('#')[0]
            ent_dom = kg_index["entity_to_domain"].get(ent, '')
            if ent_dom in allowed_doms:
                asp_to_indices[asp].append(i)
        else:
            asp_to_indices[asp].append(i)

    for asp in candidate_aspects:
        if asp not in asp_to_indices:
            for i, comp in enumerate(compounds):
                if compound_to_aspect[comp] == asp:
                    asp_to_indices[asp].append(i)

    scored = []
    for asp in candidate_aspects:
        indices = asp_to_indices.get(asp, [])
        if not indices:
            scored.append((asp, 0.0))
            continue
        sims     = comp_embs[indices] @ sent_emb
        best_sim = float(sims.max())
        scored.append((asp, round(best_sim, 4)))

    scored.sort(key=lambda x: x[1], reverse=True)
    return scored

# ================================================================
# 11. Entity selection — cosine sim sentence vs entity name
# ================================================================
def pick_entity_for_aspect(
    sentence: str,
    aspect: str,
    domain: str,
    domain_aliases: Dict[str, Set[str]],
    kg_index: Dict,
    sem_index: Dict,
    encoder,
) -> str:
    allowed_doms = domain_aliases.get(domain, {domain})
    asp_entities = kg_index["aspect_to_entities"].get(aspect, set())
    domain_ents  = [
        e for e in asp_entities
        if kg_index["entity_to_domain"].get(e) in allowed_doms
    ]

    if not domain_ents:
        return domain.upper()
    if len(domain_ents) == 1:
        return domain_ents[0].upper()

    sent_emb = encoder.encode(
        sentence, normalize_embeddings=True, convert_to_numpy=True
    )
    ent_idx  = sem_index["ent_idx"]
    ent_embs = sem_index["ent_embs"]

    best_ent, best_sim = domain_ents[0], -1.0
    for ent in domain_ents:
        if ent in ent_idx:
            sim = float(np.dot(sent_emb, ent_embs[ent_idx[ent]]))
        else:
            emb = encoder.encode(
                ent, normalize_embeddings=True, convert_to_numpy=True
            )
            sim = float(np.dot(sent_emb, emb))
        if sim > best_sim:
            best_sim = sim
            best_ent = ent

    return best_ent.upper()

# ================================================================
# 12. Per-clue inference: one clue → one aspect
# ================================================================
def infer_aspect_for_clue(
    clue: str,
    sentence: str,
    domain: str,
    domain_aliases: Dict[str, Set[str]],
    kg_index: Dict,
    sem_index: Dict,
    encoder,
    is_generic: bool = False,
) -> Dict:
    clue_norm    = _norm(clue)
    allowed_doms = domain_aliases.get(domain, {domain})

    kg_entries = kg_index["clue_to_aspects"].get(clue_norm, [])
    if not kg_entries:
        return {
            "clue": clue, "in_kg": False, "is_generic": is_generic,
            "candidates": [], "scored": [],
            "predicted_aspect": None, "reason": "not_in_kg",
        }

    raw_aspects: Set[str] = set()
    for asp, pred in kg_entries:
        asp_domains  = kg_index["aspect_to_domains"].get(asp, set())
        asp_entities = kg_index["aspect_to_entities"].get(asp, set())
        if not asp_domains:
            raw_aspects.add(asp)
            continue
        matching_ents = {
            e for e in asp_entities
            if kg_index["entity_to_domain"].get(e) in allowed_doms
        }
        if matching_ents or (asp_domains & allowed_doms):
            raw_aspects.add(asp)

    if not raw_aspects:
        return {
            "clue": clue, "in_kg": True, "is_generic": is_generic,
            "candidates": [], "scored": [],
            "predicted_aspect": None, "reason": "domain_filter_removed_all",
        }

    if is_generic:
        if "general" in raw_aspects:
            return {
                "clue": clue, "in_kg": True, "is_generic": True,
                "candidates": ["general"], "scored": [("general", 1.0)],
                "predicted_aspect": "general",
                "reason": "generic_clue_forced_general",
            }
        return {
            "clue": clue, "in_kg": True, "is_generic": True,
            "candidates": [], "scored": [],
            "predicted_aspect": None,
            "reason": "generic_clue_no_general_in_domain",
        }

    candidates = list(raw_aspects)
    scored     = rank_aspects_by_similarity(
        sentence, candidates, allowed_doms, kg_index, sem_index, encoder
    )

    specifics = [(a, s) for a, s in scored if a != "general"]
    if specifics:
        predicted = specifics[0][0]
        reason    = "top_specific_by_cosine_sim"
    elif scored:
        predicted = scored[0][0]
        reason    = "only_general_available"
    else:
        predicted = None
        reason    = "no_candidates_after_scoring"

    return {
        "clue": clue, "in_kg": True, "is_generic": False,
        "candidates": candidates, "scored": scored,
        "predicted_aspect": predicted, "reason": reason,
    }

# ================================================================
# 13. Per-row inference — gold domain + verb clue + PRON relaxation
# ================================================================
def infer_row_en(
    row: pd.Series,
    kg_index: Dict,
    sem_index: Dict,
    encoder,
    domain_aliases: Dict[str, Set[str]],
) -> Dict:
    sentence = str(
        row.get('sentence', '') or row.get('text', '') or ''
    ).strip()

    empty = {
        'kg_clues': [], 'kg_clue_aspect_scores': {},
        'kg_preds': [], 'kg_final_pairs': [],
        'inferred_domain': 'unknown', 'debug_step': '',
    }

    if not sentence:
        return {**empty, 'debug_step': 'STOPPED_empty_sentence'}

    # ── Resolve domain: gold column first, infer as fallback ─────
    gold_domain_raw = str(row.get('domain', '') or '').strip().lower()
    inferred_domain = (
        _GOLD_DOMAIN_MAP.get(gold_domain_raw)
        or infer_domain_from_sentence(sentence, encoder)
    )
    allowed_doms = domain_aliases.get(inferred_domain, {inferred_domain})

    # ── Gate 1: lonely adjective detection ───────────────────────
    lonely_adjs = find_lonely_adjectives_en(sentence)

    # ── Gate 2: verb clue scan ───────────────────────────────────
    verb_tier = _scan_verb_clues(sentence)

    # ── No lonely adjs → try verb clue fallback ──────────────────
    if not lonely_adjs:
        if verb_tier > 0:
            gen_ent = pick_entity_for_aspect(
                sentence, "general", inferred_domain,
                domain_aliases, kg_index, sem_index, encoder
            )
            step = 'VERB_TIER1_general' if verb_tier == 1 else 'VERB_TIER2_general'
            return {
                'kg_clues':              [],
                'kg_clue_aspect_scores': {},
                'kg_preds':              ['general'],
                'kg_final_pairs':        [f"{gen_ent}#GENERAL"],
                'inferred_domain':       inferred_domain,
                'debug_step':            step,
            }
        return {**empty, 'debug_step': 'STOPPED_no_lonely_adj'}

    # ── Lonely adjs found → run full KG pipeline ─────────────────
    any_in_kg = any(
        _norm(adj['lemma']) in kg_index["clue_to_aspects"]
        for adj in lonely_adjs
    )

    if not any_in_kg:
        gen_ent = pick_entity_for_aspect(
            sentence, "general", inferred_domain,
            domain_aliases, kg_index, sem_index, encoder
        )
        return {
            'kg_clues':              [a['surface'] for a in lonely_adjs],
            'kg_clue_aspect_scores': {},
            'kg_preds':              ['general'],
            'kg_final_pairs':        [f"{gen_ent}#GENERAL"],
            'inferred_domain':       inferred_domain,
            'debug_step':            'FALLBACK_general',
        }

    clue_results   = {}
    kg_preds       = []
    kg_final_pairs = []
    seen_aspects   = set()

    for adj in lonely_adjs:
        result = infer_aspect_for_clue(
            clue           = adj['lemma'],
            sentence       = sentence,
            domain         = inferred_domain,
            domain_aliases = domain_aliases,
            kg_index       = kg_index,
            sem_index      = sem_index,
            encoder        = encoder,
            is_generic     = adj['is_generic'],
        )
        clue_results[adj['surface']] = result

        asp = result['predicted_aspect']
        if asp and asp not in seen_aspects:
            seen_aspects.add(asp)
            kg_preds.append(asp)
            ent = pick_entity_for_aspect(
                sentence, asp, inferred_domain,
                domain_aliases, kg_index, sem_index, encoder
            )
            kg_final_pairs.append(f"{ent.upper()}#{asp.upper()}")

    if not kg_preds:
        gen_ent = pick_entity_for_aspect(
            sentence, "general", inferred_domain,
            domain_aliases, kg_index, sem_index, encoder
        )
        return {
            'kg_clues':              [a['surface'] for a in lonely_adjs],
            'kg_clue_aspect_scores': {s: r['scored'] for s, r in clue_results.items()},
            'kg_preds':              ['general'],
            'kg_final_pairs':        [f"{gen_ent}#GENERAL"],
            'inferred_domain':       inferred_domain,
            'debug_step':            'FALLBACK_general',
        }

    if verb_tier == 1 and "general" not in seen_aspects:
        gen_ent = pick_entity_for_aspect(
            sentence, "general", inferred_domain,
            domain_aliases, kg_index, sem_index, encoder
        )
        kg_preds.append("general")
        kg_final_pairs.append(f"{gen_ent}#GENERAL")

    return {
        'kg_clues':              [a['surface'] for a in lonely_adjs],
        'kg_clue_aspect_scores': {s: r['scored'] for s, r in clue_results.items()},
        'kg_preds':              kg_preds,
        'kg_final_pairs':        kg_final_pairs,
        'inferred_domain':       inferred_domain,
        'debug_step':            'OK',
    }

# ================================================================
# 14. Run inference for a single dataset
# ================================================================
def run_inference(
    dataset_name: str,
    dataset_path: str,
    kg_index: Dict,
    sem_index: Dict,
    encoder,
) -> pd.DataFrame:
    domain_vocab, domain_aliases = DATASET_DOMAIN_CONFIG[dataset_name]
    build_domain_index(encoder, domain_vocab=domain_vocab)

    df = pd.read_csv(dataset_path)

    if dataset_name in _ACOS_DOMAIN_FILTER:
        domain_val = _ACOS_DOMAIN_FILTER[dataset_name]
        df = df[df['domain'] == domain_val].copy().reset_index(drop=True)
        print(f"   Filtered to domain='{domain_val}': {len(df)} rows")

    print(f"\n📊 {dataset_name}: {len(df)} rows")
    if 'type' in df.columns:
        print(f"   Type breakdown:\n{df['type'].value_counts().to_string()}")

    results = df.progress_apply(
        lambda row: infer_row_en(
            row,
            kg_index       = kg_index,
            sem_index      = sem_index,
            encoder        = encoder,
            domain_aliases = domain_aliases,
        ),
        axis=1
    )

    df['kg_clues']              = results.apply(lambda d: d['kg_clues'])
    df['kg_clue_aspect_scores'] = results.apply(lambda d: d['kg_clue_aspect_scores'])
    df['kg_preds']              = results.apply(lambda d: d['kg_preds'])
    df['kg_final_pairs']        = results.apply(lambda d: d['kg_final_pairs'])
    df['kg_domain']             = results.apply(lambda d: d['inferred_domain'])
    df['kg_step']               = results.apply(lambda d: d['debug_step'])

    print(f"\n── Step distribution ──")
    print(df['kg_step'].value_counts())
    if 'type' in df.columns:
        print(f"\n── Step × Type ──")
        print(pd.crosstab(df['type'], df['kg_step']))

    return df

# ================================================================
# 15. Multi-dataset loop
# ================================================================
all_results: Dict[str, pd.DataFrame] = {}

for dataset_name, dataset_path in DATASET_PATHS.items():
    print(f"\n{'#'*60}")
    print(f"# {dataset_name}")
    print(f"{'#'*60}")

    df_out = run_inference(
        dataset_name = dataset_name,
        dataset_path = dataset_path,
        kg_index     = kg_index,
        sem_index    = sem_index,
        encoder      = encoder,
    )

    all_results[dataset_name] = df_out

    out_path = f"KG_{dataset_name}_predictions.csv"
    df_out.to_csv(out_path, index=False, encoding='utf-8-sig')
    print(f"✅ Saved: {out_path}")
    files.download(out_path)

print("\n✅ All datasets processed.")
```

#### Evaluation:

```python
# ================================================================
# Evaluation Block — Strict & Relaxed
# Paste this AFTER the multi-dataset inference loop above
#
# Strict:  predicted ASPECT must exactly match gold ASPECT
# Relaxed: predicted ASPECT matches gold ASPECT by head-noun
#          or token-set inclusion (e.g. "swimming pool" matches "pool")
# Matching: ONE-TO-ONE via bipartite matching (Kuhn's algorithm)
# Scope:   Evaluated on IMPLICIT rows only
# ================================================================

import re, ast
import pandas as pd
from collections import defaultdict
from typing import List, Tuple, Dict

# ================================================================
# A. Normalisation helpers
# ================================================================
GENERAL_SYNONYMS = {
    "misc", "miscellaneous", "anecdotes/miscellaneous",
    "miscellany", "other", "general"
}

_IRREGULAR_SINGULAR = {
    "rooms": "room", "men": "man", "women": "woman",
    "children": "child", "people": "person",
    "facilities": "facility", "amenities": "amenity",
    "services": "service", "batteries": "battery",
    "dishes": "dish", "buses": "bus", "watches": "watch",
    "indices": "index", "matrices": "matrix",
    "prices": "price", "drinks": "drink",
}

def _singularize_token(tok: str) -> str:
    t = tok.lower().strip()
    if t in _IRREGULAR_SINGULAR:
        return _IRREGULAR_SINGULAR[t]
    if re.search(r'[^aeiou]ies$', t):
        return re.sub(r'ies$', 'y', t)
    if re.search(r'(sses|zzes|ches|shes|xes)$', t):
        return t[:-2]
    if t.endswith('s') and not t.endswith('ss') and len(t) > 3:
        return t[:-1]
    return t

_token_re = re.compile(r"[a-z0-9]+")

def _tokset(s: str) -> List[str]:
    """Tokenise + singularize."""
    return [_singularize_token(x) for x in _token_re.findall((s or "").lower())]

def _head(s: str) -> str:
    """Last meaningful token after singularization."""
    toks = _tokset(s)
    return toks[-1] if toks else ""

def _norm_field(s: str) -> str:
    """Normalise a field: lowercase, singularize, join."""
    return " ".join(_tokset(s))

def _norm_aspect(s: str) -> str:
    """Normalise aspect string + canonicalise general synonyms."""
    n = _norm_field(s)
    return "general" if n in GENERAL_SYNONYMS else n

# ================================================================
# B. Gold parsing
#    Accepts: "FOOD#QUALITY" / "['FOOD#QUALITY','SERVICE#GENERAL']"
#    Returns: [(entity_norm, aspect_norm), ...]
# ================================================================
def _parse_gold(val) -> List[Tuple[str, str]]:
    if val is None:
        return []
    if isinstance(val, list):
        raw = val
    else:
        s = str(val).strip()
        try:
            v = ast.literal_eval(s)
            raw = v if isinstance(v, list) else [s]
        except Exception:
            raw = [s] if s else []

    pairs = []
    for x in raw:
        xs = str(x).strip()
        if not xs or xs.lower() in ('nan', 'none', ''):
            continue
        if "#" in xs:
            e, a = xs.split("#", 1)
            e = _norm_field(e)
            a = _norm_aspect(a)
        else:
            e = ""
            a = _norm_aspect(xs)
        if e or a:
            pairs.append((e, a))
    return pairs

# ================================================================
# C. Prediction parsing
#    kg_final_pairs: ["FOOD#QUALITY", "RESTAURANT#GENERAL"]
# ================================================================
def _parse_preds(val) -> List[Tuple[str, str]]:
    if val is None:
        return []
    if isinstance(val, list):
        raw = val
    else:
        s = str(val).strip()
        try:
            v = ast.literal_eval(s)
            raw = v if isinstance(v, list) else [s]
        except Exception:
            raw = [s] if s else []

    pairs = []
    for x in raw:
        xs = str(x).strip()
        if not xs or xs.lower() in ('nan', 'none', ''):
            continue
        if "#" in xs:
            e, a = xs.split("#", 1)
            e = _norm_field(e)
            a = _norm_aspect(a)
            pairs.append((e, a))
    return pairs

# ================================================================
# D. Compatibility predicates
#
# Strict:  predicted aspect == gold aspect (exact after norm)
# Relaxed: exact OR head-noun equality OR token-set inclusion
#          e.g. "swimming pool" matches "pool" ✅
#               "food quality"  matches "quality" ✅
#               "style options" matches "style" ✅
#
# Entity match: used ONLY as fallback when aspect is 'general'
#   on both sides (avoids entity noise polluting aspect matching)
# ================================================================
def _same_strict(a: str, b: str) -> bool:
    return _norm_field(a) == _norm_field(b)

def _same_relaxed(a: str, b: str) -> bool:
    """Aspect relaxed match: exact / head-noun / token-set inclusion."""
    a = _norm_field(a)
    b = _norm_field(b)
    if a == b:
        return True
    # head-noun equality (last token after singularization)
    ha, hb = _head(a), _head(b)
    if ha and ha == hb:
        return True
    # token-set inclusion (partial match)
    sa, sb = set(a.split()), set(b.split())
    if sa and sb and (sa.issubset(sb) or sb.issubset(sa)):
        return True
    return False

def _compatible(
    pred_pair: Tuple[str, str],
    gold_pair: Tuple[str, str],
    relaxed: bool = False,
) -> bool:
    """
    A prediction is compatible with a gold pair when:
      - Aspect matches (strict or relaxed)
      - OR both aspects are 'general' and entities match
        (handles RESTAURANT#GENERAL vs FOOD#GENERAL edge case)
    """
    pe, pa = pred_pair
    ge, ga = gold_pair

    # Primary: aspect-driven match
    if relaxed:
        if _same_relaxed(pa, ga):
            return True
    else:
        if _same_strict(pa, ga):
            return True

    # Secondary: entity match only when both aspects are general
    if pa == "general" and ga == "general":
        if ge and pe == ge:
            return True

    return False

# ================================================================
# E. Bipartite matching (Kuhn's algorithm)
#    ONE-TO-ONE: each prediction matches at most one gold,
#                each gold matched at most once
# ================================================================
def _max_bipartite_matching(
    adj: List[List[int]],
    n_left: int,
    n_right: int,
) -> List[Tuple[int, int]]:
    matchR = [-1] * n_right

    def dfs(u, seen):
        for v in adj[u]:
            if not seen[v]:
                seen[v] = True
                if matchR[v] == -1 or dfs(matchR[v], seen):
                    matchR[v] = u
                    return True
        return False

    for u in range(n_left):
        dfs(u, [False] * n_right)

    return [(u, v) for v, u in enumerate(matchR) if u != -1]

# ================================================================
# F. Single-dataset evaluation block
# ================================================================
def _eval_block(
    df: pd.DataFrame,
    pred_col: str = "kg_final_pairs",
    mode_name: str = "STRICT",
    relaxed: bool = False,
    implicit_only: bool = True,
) -> Tuple[float, float, float, List[str]]:
    """
    Runs evaluation on df (filtered to implicit rows if implicit_only=True).
    Returns (precision, recall, f1, per_row_labels).
    Prints full metrics report.
    """
    if implicit_only and 'type' in df.columns:
        eval_df = df[df['type'] == 'implicit'].copy()
    else:
        eval_df = df.copy()

    TP = FP = FN = 0
    any_tp_rows  = 0
    rows         = len(eval_df)
    per_row_labels = []

    for _, r in eval_df.iterrows():
        # Parse predictions
        pred_pairs = _parse_preds(r.get(pred_col))

        # Parse gold — deduplicate identical pairs within row
        gold_raw = _parse_gold(r.get("category"))
        seen_g = set()
        gold_pairs = []
        for g in gold_raw:
            if g not in seen_g:
                seen_g.add(g)
                gold_pairs.append(g)

        # Skip rows with no gold (e.g. objective rows without annotation)
        if not gold_pairs:
            per_row_labels.append("SKIP")
            continue

        if not pred_pairs:
            # No prediction → all gold are FN
            FN += len(gold_pairs)
            per_row_labels.append("FN")
            continue

        # Build adjacency matrix
        adj = [
            [
                j for j, g in enumerate(gold_pairs)
                if _compatible(p, g, relaxed=relaxed)
            ]
            for p in pred_pairs
        ]

        matched_pairs = _max_bipartite_matching(adj, len(pred_pairs), len(gold_pairs))
        matched_count = len(matched_pairs)

        TP += matched_count
        FP += len(pred_pairs) - matched_count
        FN += len(gold_pairs) - matched_count

        row_has_tp = matched_count > 0
        if row_has_tp:
            any_tp_rows += 1

        if matched_count == len(pred_pairs) and matched_count == len(gold_pairs):
            row_lab = "TP"
        elif row_has_tp:
            row_lab = "FP"
        else:
            row_lab = "FN"
        per_row_labels.append(row_lab)

    num_preds = TP + FP
    prec = TP / num_preds      if num_preds      else 0.0
    rec  = TP / (TP + FN)      if (TP + FN)      else 0.0
    f1   = 2*prec*rec/(prec+rec) if (prec + rec) else 0.0
    hit  = any_tp_rows / rows  if rows           else 0.0

    print(f"\n  === {mode_name} ({'implicit only' if implicit_only else 'all rows'}) ===")
    print(f"    Rows evaluated : {rows}")
    print(f"    TP             : {TP}")
    print(f"    FP             : {FP}")
    print(f"    FN             : {FN}")
    print(f"    Precision      : {prec:.4f}")
    print(f"    Recall         : {rec:.4f}")
    print(f"    F1             : {f1:.4f}")
    print(f"    Any-TP@row     : {hit:.4f}  ({any_tp_rows}/{rows} rows with ≥1 correct pred)")

    return prec, rec, f1, per_row_labels

# ================================================================
# G. Run evaluation across all datasets
# ================================================================
print("\n" + "="*60)
print("EVALUATION — English KG Inference")
print("="*60)

# Accumulate summary rows for final table
summary_rows = []

for dataset_name, df_out in all_results.items():
    print(f"\n{'#'*60}")
    print(f"# DATASET: {dataset_name}")
    print(f"{'#'*60}")

    # ── Step distribution (implicit rows) ───────────────────────
    if 'type' in df_out.columns:
        impl = df_out[df_out['type'] == 'implicit']
    else:
        impl = df_out

    print(f"\n  Implicit rows : {len(impl)}")
    print(f"  Step breakdown:")
    print(impl['kg_step'].value_counts().to_string())

    # ── Strict evaluation ────────────────────────────────────────
    p_s, r_s, f_s, labels_s = _eval_block(
        df_out,
        pred_col      = "kg_final_pairs",
        mode_name     = "STRICT",
        relaxed       = False,
        implicit_only = True,
    )

    # ── Relaxed evaluation ───────────────────────────────────────
    p_r, r_r, f_r, labels_r = _eval_block(
        df_out,
        pred_col      = "kg_final_pairs",
        mode_name     = "RELAXED",
        relaxed       = True,
        implicit_only = True,
    )

    # ── Attach row labels back to dataframe ─────────────────────
    if 'type' in df_out.columns:
        impl_idx = df_out[df_out['type'] == 'implicit'].index
    else:
        impl_idx = df_out.index

    df_out.loc[impl_idx, 'eval_strict']  = labels_s
    df_out.loc[impl_idx, 'eval_relaxed'] = labels_r

    # ── Accumulate for summary table ────────────────────────────
    summary_rows.append({
        'Dataset':          dataset_name,
        'Implicit rows':    len(impl),
        'KG fired (OK+FB)': int(impl['kg_step'].isin(['OK','FALLBACK_general']).sum()),
        'P (strict)':       round(p_s, 4),
        'R (strict)':       round(r_s, 4),
        'F1 (strict)':      round(f_s, 4),
        'P (relaxed)':      round(p_r, 4),
        'R (relaxed)':      round(r_r, 4),
        'F1 (relaxed)':     round(f_r, 4),
    })

    # ── Save updated dataframe with eval labels ──────────────────
    out_path = f"KG_{dataset_name}_eval.csv"
    df_out.to_csv(out_path, index=False, encoding='utf-8-sig')
    print(f"\n  ✅ Saved: {out_path}")
    files.download(out_path)

# ================================================================
# H. Summary table across all datasets
# ================================================================
print("\n" + "="*60)
print("SUMMARY TABLE")
print("="*60)

summary_df = pd.DataFrame(summary_rows)
print(summary_df.to_string(index=False))

summary_df.to_csv("KG_English_evaluation_summary.csv", index=False, encoding='utf-8-sig')
files.download("KG_English_evaluation_summary.csv")
print("\n✅ Summary saved.")
```

```python
# ================================================================
# COMPLETE EVALUATION BLOCK — Self-contained
# Uses original helper functions from the pipeline notebook
# to ensure consistent normalisation with established results.
# ================================================================

import pandas as pd
import ast
import re
import numpy as np
from typing import List, Dict, Tuple, Iterable, Optional

# ================================================================
# 1. Original helper functions (from pipeline notebook)
#    These ensure evaluate_dataframe produces identical results
#    to the original established scores.
# ================================================================

# ── Adjective removal ────────────────────────────────────────────
_ADJ_LEXICON = set("""
جيد جيدة ممتاز ممتازة سيئ سيئة رائع رائعة جميل جميلة كبير كبيرة صغير صغيرة
طويل طويلة قصير قصيرة سريع سريعة بطيء بطيئة مجاني مجانية جديد جديدة قديم
قديمة طبيعي طبيعية واضح واضحة قوي قوية ضعيف ضعيفة
""".split())

def _looks_like_adj(word: str) -> bool:
    w = word.strip()
    if not w:
        return False
    if w.endswith(("ي", "ية", "يات", "ان", "ين")):
        return True
    if w in _ADJ_LEXICON:
        return True
    return False

class CamelAdjRemover:
    def __init__(self):
        from camel_tools.tokenizers.word import simple_word_tokenize
        from camel_tools.disambig.mle import MLEDisambiguator
        self.tok = simple_word_tokenize
        self.mle = MLEDisambiguator.pretrained()

    def strip_adjectives(self, phrase: str) -> str:
        tokens = self.tok(phrase)
        disamb = self.mle.disambiguate(tokens)
        kept = []
        for t, d in zip(tokens, disamb):
            pos = None
            if d.analyses:
                a = d.analyses[0]
                pos = (getattr(a, "pos", None) or
                       getattr(a, "pos_tag", "") or "").lower()
            if not (pos.startswith("adj") or pos == "a"):
                kept.append(t)
        return " ".join(kept).strip()

def strip_adjectives_heuristic(phrase: str) -> str:
    toks = [t for t in re.split(r"\s+", phrase.strip()) if t]
    kept = [t for t in toks if not _looks_like_adj(t)]
    return " ".join(kept).strip()

# ── Normalisation ─────────────────────────────────────────────────
_DIACRITICS = re.compile(r'[\u0617-\u061A\u064B-\u0652]')
_PREFIXES   = re.compile(r'^(?:ال|و|ف|ب|ل|ك)+')
_SUFFIXES   = re.compile(r'(?:هما|كما|ها|هم|نا|كم|هن|كن|ه|ي|ة|ات)$')

def normalize_token(tok: str) -> str:
    t = str(tok)
    t = t.replace("أ","ا").replace("إ","ا").replace("آ","ا")
    t = t.replace("ى","ي").replace("ة","ه").replace("ؤ","ء").replace("ئ","ء")
    t = _DIACRITICS.sub("", t).replace("ـ", "")
    t = _PREFIXES.sub("", t)
    t = _SUFFIXES.sub("", t)
    return t.strip()

def normalize_phrase(phrase: str) -> str:
    toks = [w for w in re.split(r"\s+", phrase.strip()) if w]
    return " ".join(normalize_token(w) for w in toks)

# ── Parser ────────────────────────────────────────────────────────
_SEP_PAT = re.compile(
    r"\s*(?:\[sep\]|\[SEP\]|،|,|;|؛|\||\n|\t|\/|\\|:|\s{2,})\s*",
    re.IGNORECASE
)

def parse_aspect_string(s) -> List[str]:
    if s is None:
        return []
    if isinstance(s, list):
        items = s
    else:
        s = str(s).strip().strip("[](){}").replace("'", "").replace('"', "")
        items = _SEP_PAT.split(s)
    cleaned = []
    for t in items:
        t = re.sub(r"^[\s\-\–\—\·\•\*\.]+|[\s\-\–\—\·\•\*\.]+$", "", t.strip())
        if t:
            cleaned.append(t)
    seen_norm = set()
    out = []
    for w in cleaned:
        key = normalize_phrase(w)
        if key and key not in seen_norm:
            seen_norm.add(key)
            out.append(w)
    return out

# ================================================================
# GOLD DEDUPLICATION — MABSA-specific
# Applied to both 'aspect' and 'category' columns before evaluation.
# Removes exact duplicate items that appear due to annotation
# ================================================================

def dedup_gold_field(val) -> list:
    """
    Parse a gold field (aspect or category) and deduplicate
    while preserving order of first occurrence.
    Works on both string representations and actual lists.

    Examples:
      "['product', 'product']"         → ['product']
      "['food#general', 'food#general']" → ['food#general']
      "['food#quality', 'food#general']" → ['food#quality', 'food#general']  (unchanged)
    """
    # Parse to list
    if isinstance(val, list):
        items = [str(v).strip() for v in val]
    else:
        s = str(val).strip()
        if s.startswith('['):
            try:
                parsed = ast.literal_eval(s)
                items = [str(v).strip() for v in parsed]
            except Exception:
                items = [s]
        else:
            items = [s]

    # Deduplicate preserving order — use lower() key for comparison
    seen = set()
    deduped = []
    for item in items:
        key = item.strip().lower()
        if key and key not in seen:
            seen.add(key)
            deduped.append(item)

    return deduped if deduped else ['none']

# ── Prediction cleaner ────────────────────────────────────────────
def clean_predicted_aspects(raw_pred,
                             use_camel: bool = True,
                             camel_obj: Optional[CamelAdjRemover] = None
                             ) -> List[str]:
    if raw_pred is None:
        aspects = []
    elif isinstance(raw_pred, list):
        aspects = raw_pred
    else:
        aspects = parse_aspect_string(raw_pred)

    if use_camel:
        try:
            camel = camel_obj or CamelAdjRemover()
            stripped = [camel.strip_adjectives(a) for a in aspects]
        except Exception:
            stripped = [strip_adjectives_heuristic(a) for a in aspects]
    else:
        stripped = [strip_adjectives_heuristic(a) for a in aspects]

    seen = set()
    final = []
    for a in stripped:
        a = str(a).strip()
        if not a:
            continue
        key = normalize_phrase(a)
        if key and key not in seen:
            seen.add(key)
            final.append(a)
    return final

# ── Matching ──────────────────────────────────────────────────────
def jaccard(a: str, b: str) -> float:
    A = set(a.split())
    B = set(b.split())
    if not A and not B: return 1.0
    if not A or not B:  return 0.0
    return len(A & B) / len(A | B)

def similarity(gold: str, pred: str) -> float:
    g = normalize_phrase(gold)
    p = normalize_phrase(pred)
    if not g or not p: return 0.0
    if g == p:         return 1.0
    if g in p or p in g: return 1.0
    return jaccard(g, p)

def best_alignment(gold: List[str], pred: List[str],
                   thresh: float = 0.6) -> Tuple[int, List]:
    pairs = []
    for i, g in enumerate(gold):
        for j, p in enumerate(pred):
            s = similarity(g, p)
            if s >= thresh:
                pairs.append((i, j, s))
    pairs.sort(key=lambda x: x[2], reverse=True)
    matched_g, matched_p, matches = set(), set(), []
    for gi, pj, s in pairs:
        if gi not in matched_g and pj not in matched_p:
            matched_g.add(gi); matched_p.add(pj)
            matches.append((gi, pj, s))
    return len(matches), matches

def prf1_for_example(gold_terms: Iterable[str],
                     pred_terms: Iterable[str],
                     thresh: float = 0.6) -> Dict:
    gold = [t for t in map(str, gold_terms) if t and t.strip()]
    pred = [t for t in map(str, pred_terms) if t and t.strip()]
    if not gold and not pred:
        return {"tp":0,"fp":0,"fn":0,"precision":1.0,"recall":1.0,"f1":1.0}
    tp, _ = best_alignment(gold, pred, thresh=thresh)
    fp = max(0, len(pred) - tp)
    fn = max(0, len(gold) - tp)
    prec = tp/(tp+fp) if (tp+fp) else 0.0
    rec  = tp/(tp+fn) if (tp+fn) else 0.0
    f1   = 2*prec*rec/(prec+rec) if (prec+rec) else 0.0
    return {"tp":tp,"fp":fp,"fn":fn,"precision":prec,"recall":rec,"f1":f1}

# ── evaluate_dataframe (original, unchanged) ──────────────────────
def evaluate_dataframe(df: pd.DataFrame,
                       pred_col: str = "prediction",
                       gold_col: str = "target",
                       use_camel: bool = False,
                       jaccard_threshold: float = 0.6):
    df = df.copy()
    df["pred_list_clean"] = df[pred_col].apply(
        lambda s: clean_predicted_aspects(s, use_camel=use_camel)
    )

    def parse_gold(x):
        if isinstance(x, list): return x
        return parse_aspect_string(x)
    df["gold_list"] = df[gold_col].apply(parse_gold)

    per = df.apply(
        lambda r: prf1_for_example(
            r["gold_list"], r["pred_list_clean"], thresh=jaccard_threshold
        ), axis=1
    )
    df["precision"] = per.apply(lambda m: m["precision"])
    df["recall"]    = per.apply(lambda m: m["recall"])
    df["f1"]        = per.apply(lambda m: m["f1"])
    df["tp"]        = per.apply(lambda m: m["tp"])
    df["fp"]        = per.apply(lambda m: m["fp"])
    df["fn"]        = per.apply(lambda m: m["fn"])
    df["support"]   = df["gold_list"].apply(len)

    tot_tp = df["tp"].sum(); tot_fp = df["fp"].sum(); tot_fn = df["fn"].sum()
    micro_p = tot_tp/(tot_tp+tot_fp) if (tot_tp+tot_fp) else 0.0
    micro_r = tot_tp/(tot_tp+tot_fn) if (tot_tp+tot_fn) else 0.0
    micro_f = 2*micro_p*micro_r/(micro_p+micro_r) if (micro_p+micro_r) else 0.0

    macro_p = df["precision"].mean() if len(df) else 0.0
    macro_r = df["recall"].mean()    if len(df) else 0.0
    macro_f = df["f1"].mean()        if len(df) else 0.0

    total_support = df["support"].sum()
    if total_support > 0:
        weighted_p = (df["precision"]*df["support"]).sum()/total_support
        weighted_r = (df["recall"]   *df["support"]).sum()/total_support
        weighted_f = (df["f1"]       *df["support"]).sum()/total_support
    else:
        weighted_p = weighted_r = weighted_f = 0.0

    return {
        "micro": {
            "precision": round(float(micro_p),4),
            "recall":    round(float(micro_r),4),
            "f1":        round(float(micro_f),4),
            "accuracy":  round(float(tot_tp/(tot_tp+tot_fp+tot_fn)),4)
                         if (tot_tp+tot_fp+tot_fn) else 0.0,
            "tp": int(tot_tp), "fp": int(tot_fp), "fn": int(tot_fn),
        },
        "macro": {
            "precision": round(float(macro_p),4),
            "recall":    round(float(macro_r),4),
            "f1":        round(float(macro_f),4),
        },
        "weighted_macro": {
            "precision": round(float(weighted_p),4),
            "recall":    round(float(weighted_r),4),
            "f1":        round(float(weighted_f),4),
        },
        "df": df,
    }

# ── evaluate_aspect: thin wrapper around evaluate_dataframe ───────
def evaluate_aspect(df: pd.DataFrame,
                    pred_col: str,
                    gold_col: str = 'aspect',
                    jaccard_threshold: float = 0.6) -> Dict:
    """
    Wrapper around the original evaluate_dataframe so that explicit
    extraction scores match the established pipeline results exactly.
    """
    result = evaluate_dataframe(df, pred_col=pred_col, gold_col=gold_col,
                                use_camel=False,
                                jaccard_threshold=jaccard_threshold)
    return {
        "micro": {
            "P":  result["micro"]["precision"],
            "R":  result["micro"]["recall"],
            "F1": result["micro"]["f1"],
            "TP": result["micro"]["tp"],
            "FP": result["micro"]["fp"],
            "FN": result["micro"]["fn"],
        },
        "macro": {
            "P":  result["macro"]["precision"],
            "R":  result["macro"]["recall"],
            "F1": result["macro"]["f1"],
        },
    }

# ================================================================
# 2. Normalisation helpers for KG evaluation
# ================================================================
_NULL_VALUES = {
    'none','null','nan','n/a','','[]',"['none']",
    "['null']",'[none]','[null]',"['']",
}

def _is_null_value(v: str) -> bool:
    return str(v).strip().lower() in _NULL_VALUES

def normalise_field(val) -> List[str]:
    if isinstance(val, list):
        items = [str(v).strip().lower() for v in val]
    else:
        s = str(val).strip()
        if s.startswith('['):
            try:
                parsed = ast.literal_eval(s)
                items = [str(v).strip().lower() for v in parsed]
            except Exception:
                items = [s.lower()]
        else:
            items = [s.lower()]
    items = [v for v in items if v != '']
    if not items or all(_is_null_value(v) for v in items):
        return ['none']
    items = ['none' if _is_null_value(v) else v for v in items]
    return items

PRED_COLS = [
    'cleaned_baseline','cleaned_explicit_iterative',
    'cleaned_explicit_paraphrases','cleaned_combined_aspects',
    'kg_preds','kg_final_pairs',
]
GOLD_COLS = ['aspect','category']

def normalise_dataframe(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()
    for col in PRED_COLS + GOLD_COLS:
        if col in df.columns:
            df[col] = df[col].apply(normalise_field)
    if 'kg_step' in df.columns:
        df['kg_step'] = df['kg_step'].fillna('STOPPED_no_lonely_adj').astype(str)
    return df

KEEP_COLS = [
    'dataset','domain','sentence','aspect','category',
    'type','source_dataset','language','paraphrases',
    'cleaned_baseline','cleaned_explicit_iterative',
    'cleaned_explicit_paraphrases','cleaned_combined_aspects',
    'kg_preds','kg_final_pairs','kg_step','kg_domain','kg_clues',
]
# In load_and_normalise(), after loading the dataframe,
# apply the original gold preprocessing to the 'aspect' column:

def preprocess_gold_aspect(df: pd.DataFrame) -> pd.DataFrame:
    """
    Replicates the original process_data() gold preprocessing:
    safe_literal_eval → clean_list_values
    This ensures gold item counts match the original evaluation exactly.
    """
    import ast

    def safe_literal_eval(x):
        if isinstance(x, list): return x
        if pd.isna(x): return []
        try:
            return ast.literal_eval(x)
        except (ValueError, SyntaxError):
            return [x]

    def clean_list_values(x):
        if isinstance(x, list):
            cleaned = []
            for v in x:
                if v is None or pd.isna(v) or str(v).strip().lower() == "nan":
                    continue
                cleaned.append(str(v).strip())
            return cleaned
        if x is None or pd.isna(x) or str(x).strip().lower() == "nan":
            return []
        return [str(x).strip()]

    df = df.copy()
    if 'aspect' in df.columns:
        df['aspect'] = df['aspect'].apply(safe_literal_eval)
        df['aspect'] = df['aspect'].apply(
            lambda x: ast.literal_eval(x) if isinstance(x, str) else x
        )
        df['aspect'] = df['aspect'].apply(clean_list_values)
    return df

def load_and_normalise(df: pd.DataFrame, name: str) -> pd.DataFrame:
    df = df.copy()
    if 'dataset' not in df.columns:
        df['dataset'] = name
    cols = [c for c in KEEP_COLS if c in df.columns]
    df = df[cols]

    # ── Deduplicate gold fields (fixes M-ABSA annotation redundancy) ──
    # Applied to ALL datasets harmlessly — dedup is a no-op when
    # no duplicates exist, so SE14/SE15/SE16/ACOS are unaffected.
    if 'aspect' in df.columns:
        df['aspect'] = df['aspect'].apply(dedup_gold_field)
    if 'category' in df.columns:
        df['category'] = df['category'].apply(dedup_gold_field)

    # ── Normalise KG/prediction columns ──────────────────────────────
    df = normalise_dataframe(df)
    # # ── Apply original gold preprocessing first ───────────────────
    # df = preprocess_gold_aspect(df)
    # ── Then apply KG/prediction normalisation ────────────────────
    df = normalise_dataframe(df)
    print(f"✅ {name}: {len(df)} rows")
    if 'type'    in df.columns: print(f"   Types  : {dict(df['type'].value_counts())}")
    if 'domain'  in df.columns: print(f"   Domains: {dict(df['domain'].value_counts())}")
    if 'kg_step' in df.columns: print(f"   KG steps: {dict(df['kg_step'].value_counts())}")
    print()
    return df


# ================================================================
# 3. Load and merge datasets
# ================================================================
KG_MERGE_COLS = ['sentence','kg_preds','kg_final_pairs',
                 'kg_step','kg_domain','kg_clues']
KG_FILL = {
    'kg_preds':'[]','kg_final_pairs':'[]',
    'kg_step':'STOPPED_no_lonely_adj',
    'kg_domain':'unknown','kg_clues':'[]',
}

def merge_kg(pipeline_df: pd.DataFrame,
             kg_df: pd.DataFrame) -> pd.DataFrame:
    kg_cols = [c for c in KG_MERGE_COLS if c in kg_df.columns]
    merged  = pipeline_df.merge(kg_df[kg_cols], on='sentence', how='left')
    for col, default in KG_FILL.items():
        if col in merged.columns:
            merged[col] = merged[col].fillna(default)
        else:
            merged[col] = default
    return merged

# Pipeline files
pipe_14   = pd.read_csv("/content/full_predictions_se14.csv")
pipe_15   = pd.read_csv("/content/full_predictions_se15.csv")
pipe_16   = pd.read_csv("/content/full_predictions_se16.csv")
pipe_acos = pd.read_csv("/content/full_predictions_acos.csv")
pipe_mabsa = pd.read_csv("/content/full_predictions_mabsa.csv")

# KG inference files
kg_14        = pd.read_csv("/content/KG_SemEval15_predictions.csv") # SemEval14 does not support KG inference because it has no column "category"
kg_15        = pd.read_csv("/content/KG_SemEval15_predictions.csv")
kg_16        = pd.read_csv("/content/KG_SemEval16_predictions.csv")
kg_acos_rest = pd.read_csv("/content/KG_ACOS_restaurant_predictions.csv")
kg_acos_lap  = pd.read_csv("/content/KG_ACOS_laptop_predictions.csv")
kg_mabsa   = pd.read_csv("/content/KG_MABSA_EN_predictions.csv")

# SemEval14
test_14 = merge_kg(pipe_14, kg_14)
test_14 = load_and_normalise(test_14, "SemEval14")

# SemEval15
test_15 = merge_kg(pipe_15, kg_15)
test_15 = load_and_normalise(test_15, "SemEval15")

# SemEval16
test_16 = merge_kg(pipe_16, kg_16)
test_16 = load_and_normalise(test_16, "SemEval16")

# ACOS restaurant
pipe_acos_rest = pipe_acos[
    pipe_acos['domain'].str.lower().isin(['restaurants','restaurant'])
].copy()
test_acos_rest = merge_kg(pipe_acos_rest, kg_acos_rest)
test_acos_rest = load_and_normalise(test_acos_rest, "ACOS_restaurant")

# ACOS laptop
pipe_acos_lap = pipe_acos[
    pipe_acos['domain'].str.lower().isin(['laptops','laptop'])
].copy()
test_acos_lap = merge_kg(pipe_acos_lap, kg_acos_lap)
test_acos_lap = load_and_normalise(test_acos_lap, "ACOS_laptop")

# MABSA (uncomment when ready)
test_mabsa = merge_kg(pipe_mabsa, kg_mabsa)
test_mabsa = load_and_normalise(test_mabsa, "MABSA_EN")

# ================================================================
# 4. KG aggregation — acceptance/rejection rule
# ================================================================
def _norm_kg_asp(asp: str) -> str:
    a = str(asp).strip().lower()
    _MAP = {
        "prices":                "price",
        "style options":         "style_options",
        "design features":       "design_features",
        "operation performance": "operation_performance",
        "food quality":          "quality",
        "miscellaneous":         "general",
    }
    return _MAP.get(a, a)

def aggregate_kg(row: pd.Series) -> List[str]:
    kg_preds = row.get('kg_preds', ['none'])
    if not isinstance(kg_preds, list): kg_preds = ['none']
    kg_preds_norm = [
        _norm_kg_asp(a) for a in kg_preds
        if not _is_null_value(a) and a != 'none'
    ]
    if not kg_preds_norm: return []
    kg_specific     = [a for a in kg_preds_norm if a != 'general']
    kg_only_general = len(kg_specific) == 0
    para_preds = row.get('cleaned_explicit_paraphrases', ['none'])
    if not isinstance(para_preds, list): para_preds = ['none']
    pipeline_has_preds = not all(
        _is_null_value(v) or v == 'none' for v in para_preds
    )
    if kg_only_general and pipeline_has_preds:
        return []
    return kg_preds_norm

def build_combined_kg_column(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()
    def _combine(row):
        base = row.get('cleaned_explicit_paraphrases', ['none'])
        if not isinstance(base, list): base = ['none']
        base_real = [v for v in base if not _is_null_value(v) and v != 'none']
        kg_extra  = aggregate_kg(row)
        kg_new    = [a for a in kg_extra if a not in base_real]
        combined  = base_real + kg_new
        return combined if combined else ['none']
    df['cleaned_combined_aspects_kg'] = df.apply(_combine, axis=1)
    return df
test_14        = build_combined_kg_column(test_14)
test_15        = build_combined_kg_column(test_15)
test_16        = build_combined_kg_column(test_16)
test_acos_rest = build_combined_kg_column(test_acos_rest)
test_acos_lap  = build_combined_kg_column(test_acos_lap)
# test_mabsa   = build_combined_kg_column(test_mabsa)
print("✅ KG combined column built for all datasets.\n")

# ================================================================
# 5. KG-only evaluation helpers (against category gold)
# ================================================================
_GENERAL_SYN = {
    "misc","miscellaneous","anecdotes/miscellaneous",
    "miscellany","other","general"
}
_IRREG = {
    "rooms":"room","prices":"price","drinks":"drink",
    "services":"service","facilities":"facility","amenities":"amenity",
}
_tok_re = re.compile(r"[a-z0-9]+")

def _sing(t):
    t = t.lower()
    if t in _IRREG: return _IRREG[t]
    if re.search(r'[^aeiou]ies$',t): return re.sub(r'ies$','y',t)
    if re.search(r'(sses|ches|shes|xes)$',t): return t[:-2]
    if t.endswith('s') and not t.endswith('ss') and len(t)>3: return t[:-1]
    return t

def _nf(s):
    return " ".join([_sing(x) for x in _tok_re.findall((s or "").lower())])

def _norm_asp_eval(s):
    n = _nf(s)
    return "general" if n in _GENERAL_SYN else n

def _head(s):
    toks = _nf(s).split()
    return toks[-1] if toks else ""

def _same_relaxed(a, b):
    a, b = _norm_asp_eval(a), _norm_asp_eval(b)
    if a == b: return True
    if _head(a) and _head(a) == _head(b): return True
    sa, sb = set(a.split()), set(b.split())
    if sa and sb and (sa.issubset(sb) or sb.issubset(sa)): return True
    return False

def _parse_gold_cat(val):
    if not isinstance(val, list): val = [str(val)]
    pairs = []
    for c in val:
        s = str(c).strip()
        if _is_null_value(s) or s == 'none': continue
        if '#' in s:
            e, a = s.split('#',1)
            pairs.append((_nf(e), _norm_asp_eval(a)))
        else:
            # ACOS format: 'food quality' / 'laptop operation_performance'
            tokens = s.strip().split()
            if len(tokens) >= 2:
                e   = _nf(tokens[0])
                asp = ' '.join(tokens[1:]).replace('_',' ')
                pairs.append((e, _norm_asp_eval(asp)))
            elif len(tokens) == 1:
                pairs.append(("*", _norm_asp_eval(tokens[0])))
    return pairs

def _parse_pred_pairs(val):
    if not isinstance(val, list): val = [str(val)]
    pairs = []
    for x in val:
        s = str(x).strip()
        if '#' in s:
            e, a = s.split('#',1)
            pairs.append((_nf(e), _norm_asp_eval(a)))
    return pairs

def _max_bip(adj, nl, nr):
    matchR = [-1]*nr
    def dfs(u, seen):
        for v in adj[u]:
            if not seen[v]:
                seen[v]=True
                if matchR[v]==-1 or dfs(matchR[v],seen):
                    matchR[v]=u; return True
        return False
    for u in range(nl): dfs(u,[False]*nr)
    return [(u,v) for v,u in enumerate(matchR) if u!=-1]

def evaluate_kg_pairs(df: pd.DataFrame,
                      pred_col: str = 'kg_final_pairs',
                      gold_col: str = 'category',
                      relaxed: bool = False) -> Dict:
    TP=FP=FN=any_tp=rows=0
    for _, row in df.iterrows():
        gold = _parse_gold_cat(row.get(gold_col,[]))
        if not gold: continue
        rows += 1
        pred = _parse_pred_pairs(row.get(pred_col,[]))
        if not pred: FN+=len(gold); continue
        adj = [
            [j for j,g in enumerate(gold)
             if (_same_relaxed(p[1],g[1]) if relaxed else p[1]==g[1])
             or (p[1]=="general" and g[1]=="general"
                 and g[0]!="*" and p[0]==g[0])]
            for p in pred
        ]
        mc = len(_max_bip(adj,len(pred),len(gold)))
        TP+=mc; FP+=len(pred)-mc; FN+=len(gold)-mc
        if mc>0: any_tp+=1
    p = TP/(TP+FP) if (TP+FP) else 0.0
    r = TP/(TP+FN) if (TP+FN) else 0.0
    f = 2*p*r/(p+r) if (p+r) else 0.0
    return {
        "P":round(p,4),"R":round(r,4),"F1":round(f,4),
        "TP":TP,"FP":FP,"FN":FN,"rows":rows,
        "Any-TP@row":round(any_tp/rows,4) if rows else 0.0,
    }

# ================================================================
# 6. Approach 2 helpers (type-aware combined evaluation)
# ================================================================
def _extract_aspect_from_category(cat_list: List[str]) -> List[str]:
    aspects = []
    for c in cat_list:
        s = str(c).strip()
        if _is_null_value(s) or s == 'none': continue
        if '#' in s:
            asp = s.split('#',1)[1].strip().lower().replace('_',' ')
        else:
            tokens = s.strip().split()
            if len(tokens) >= 2:
                asp = ' '.join(tokens[1:]).replace('_',' ')
            elif len(tokens) == 1:
                asp = tokens[0].replace('_',' ')
            else:
                continue
        asp = _norm_asp_eval(asp)
        if asp: aspects.append(asp)
    return aspects if aspects else ['none']

def _get_combined_pred_for_row(row: pd.Series) -> List[str]:
    typ = str(row.get('type','explicit')).lower()
    if typ == 'implicit':
        pred = row.get('cleaned_combined_aspects_kg', ['none'])
        return pred if isinstance(pred, list) else ['none']
    base = row.get('cleaned_explicit_paraphrases', ['none'])
    if not isinstance(base, list): base = ['none']
    base_real   = [v for v in base if not _is_null_value(v) and v != 'none']
    kg_contrib  = aggregate_kg(row)
    kg_specific = [a for a in kg_contrib if a != 'general']
    kg_new      = [a for a in kg_specific if a not in base_real]
    combined    = base_real + kg_new
    return combined if combined else ['none']

def _get_gold_for_row(row: pd.Series) -> List[str]:
    typ = str(row.get('type','explicit')).lower()
    if typ == 'implicit':
        cat = row.get('category', ['none'])
        if not isinstance(cat, list): cat = ['none']
        return _extract_aspect_from_category(cat)
    gold = row.get('aspect', ['none'])
    return gold if isinstance(gold, list) else ['none']

def evaluate_combined(df: pd.DataFrame,
                      jaccard_threshold: float = 0.6) -> Dict:
    """
    Type-aware combined evaluation using original prf1_for_example.
    explicit/mixed/objective → gold=aspect, pred=pipeline
    implicit                 → gold=category aspect-part, pred=combined+KG
    """
    tot_tp = tot_fp = tot_fn = 0
    p_list, r_list, f_list = [], [], []
    for _, row in df.iterrows():
        gold = _get_gold_for_row(row)
        pred = _get_combined_pred_for_row(row)
        # Use original prf1_for_example for consistent matching
        m = prf1_for_example(gold, pred, thresh=jaccard_threshold)
        tot_tp += m['tp']; tot_fp += m['fp']; tot_fn += m['fn']
        p_list.append(m['precision'])
        r_list.append(m['recall'])
        f_list.append(m['f1'])
    mp = tot_tp/(tot_tp+tot_fp) if (tot_tp+tot_fp) else 0.0
    mr = tot_tp/(tot_tp+tot_fn) if (tot_tp+tot_fn) else 0.0
    mf = 2*mp*mr/(mp+mr)        if (mp+mr)         else 0.0
    return {
        "micro": {"P":round(mp,4),"R":round(mr,4),"F1":round(mf,4),
                  "TP":tot_tp,"FP":tot_fp,"FN":tot_fn},
        "macro": {"P":round(float(np.mean(p_list)),4),
                  "R":round(float(np.mean(r_list)),4),
                  "F1":round(float(np.mean(f_list)),4)},
    }

# ================================================================
# 7. Print helpers
# ================================================================
def _print_aspect_row(label, result, width=20):
    m  = result["micro"]
    mc = result["macro"]
    print(f"    {label:<{width}} | "
          f"Micro P={m['P']:.4f} R={m['R']:.4f} F1={m['F1']:.4f} "
          f"(TP={m['TP']} FP={m['FP']} FN={m['FN']}) | "
          f"Macro F1={mc['F1']:.4f}")

def _print_combined_row(label, res, width=25):
    m  = res["micro"]
    mc = res["macro"]
    print(f"    {label:<{width}} | "
          f"Micro P={m['P']:.4f} R={m['R']:.4f} F1={m['F1']:.4f} "
          f"(TP={m['TP']} FP={m['FP']} FN={m['FN']}) | "
          f"Macro F1={mc['F1']:.4f}")

# ================================================================
# 8. Pipeline stage definitions
# ================================================================
EXPLICIT_STAGES = [
    ("FT",         "cleaned_baseline"),
    ("FT+IG",      "cleaned_explicit_iterative"),
    ("FT+IG+PARA", "cleaned_explicit_paraphrases"),
]
TYPE_ORDER = ['explicit','implicit','mixed','objective']

# ================================================================
# 9. Report functions
# ================================================================
def report_approach1(df, dataset_name, jaccard_threshold=0.6):
    print(f"\n{'#'*70}")
    print(f"# APPROACH 1 — SEPARATE METRICS")
    print(f"# DATASET: {dataset_name}  ({len(df)} rows)")
    print(f"{'#'*70}")

    stages        = [(l,c) for l,c in EXPLICIT_STAGES if c in df.columns]
    types_present = df['type'].dropna().unique() if 'type' in df.columns else []

    # ── Metric A: Explicit extraction ────────────────────────────
    print(f"\n{'─'*60}")
    print(f"  METRIC A — EXPLICIT EXTRACTION (gold=aspect)")
    print(f"{'─'*60}")

    print(f"\n  A.1 Overall ({len(df)} rows)")
    for lbl, col in stages:
        res = evaluate_aspect(df, col, gold_col='aspect',
                              jaccard_threshold=jaccard_threshold)
        _print_aspect_row(lbl, res)

    for typ in TYPE_ORDER:
        if typ not in types_present: continue
        sub = df[df['type'] == typ]
        print(f"\n  A.2 {typ.upper()} ({len(sub)} rows)")
        for lbl, col in stages:
            if col not in sub.columns: continue
            res = evaluate_aspect(sub, col, gold_col='aspect',
                                  jaccard_threshold=jaccard_threshold)
            _print_aspect_row(lbl, res)

    if 'domain' in df.columns:
        domains = sorted(df['domain'].dropna().unique())
        if len(domains) > 1:
            print(f"\n  A.3 PER DOMAIN")
            for dom in domains:
                sub = df[df['domain'] == dom]
                print(f"\n    Domain: {dom}  ({len(sub)} rows)")
                for lbl, col in stages:
                    if col not in sub.columns: continue
                    res = evaluate_aspect(sub, col, gold_col='aspect',
                                          jaccard_threshold=jaccard_threshold)
                    _print_aspect_row(lbl, res, width=22)

    # ── Metric B: Implicit KG extraction ─────────────────────────
    print(f"\n{'─'*60}")
    print(f"  METRIC B — IMPLICIT KG (gold=category, implicit rows)")
    print(f"{'─'*60}")

    if 'type' not in df.columns or 'kg_final_pairs' not in df.columns:
        print("  ⚠ Missing 'type' or 'kg_final_pairs' — skipping Metric B")
        return

    impl = df[df['type'] == 'implicit']
    print(f"\n  B.1 Overall implicit ({len(impl)} rows)")
    for relaxed, mode in [(False,"STRICT"),(True,"RELAXED")]:
        r = evaluate_kg_pairs(impl, relaxed=relaxed)
        print(f"    KG {mode:<8} | "
              f"P={r['P']:.4f} R={r['R']:.4f} F1={r['F1']:.4f} "
              f"(TP={r['TP']} FP={r['FP']} FN={r['FN']}) "
              f"Any-TP@row={r['Any-TP@row']:.4f} ({r['rows']} rows)")

    if 'kg_step' in impl.columns:
        print(f"\n  B.2 KG step distribution (implicit)")
        for step, cnt in impl['kg_step'].value_counts().items():
            print(f"    {step:<35}: {cnt:4d} ({100*cnt/len(impl):.1f}%)")

    if 'domain' in impl.columns:
        domains = sorted(impl['domain'].dropna().unique())
        if len(domains) > 1:
            print(f"\n  B.3 KG by domain (implicit rows)")
            for dom in domains:
                sub = impl[impl['domain'] == dom]
                print(f"\n    Domain: {dom}  ({len(sub)} rows)")
                for relaxed, mode in [(False,"STRICT"),(True,"RELAXED")]:
                    r = evaluate_kg_pairs(sub, relaxed=relaxed)
                    print(f"      KG {mode:<8} | "
                          f"P={r['P']:.4f} R={r['R']:.4f} F1={r['F1']:.4f} "
                          f"(TP={r['TP']} FP={r['FP']} FN={r['FN']})")


def report_approach2(df, dataset_name, jaccard_threshold=0.6):
    print(f"\n{'#'*70}")
    print(f"# APPROACH 2 — TYPE-AWARE COMBINED EVALUATION")
    print(f"# DATASET: {dataset_name}  ({len(df)} rows)")
    print(f"# explicit/mixed/objective → gold=aspect")
    print(f"# implicit                 → gold=category (aspect part)")
    print(f"{'#'*70}")

    stages        = [(l,c) for l,c in EXPLICIT_STAGES if c in df.columns]
    types_present = df['type'].dropna().unique() if 'type' in df.columns else []

    print(f"\n{'─'*60}")
    print(f"  1. OVERALL ({len(df)} rows)")
    print(f"{'─'*60}")
    print(f"  [Explicit-only baselines — gold=aspect, all rows]")
    for lbl, col in stages:
        if col not in df.columns: continue
        res = evaluate_aspect(df, col, gold_col='aspect',
                              jaccard_threshold=jaccard_threshold)
        _print_combined_row(lbl, res)
    print(f"\n  [Full system — type-aware gold]")
    res = evaluate_combined(df, jaccard_threshold)
    _print_combined_row("FT+IG+PARA+KG", res)

    for typ in TYPE_ORDER:
        if typ not in types_present: continue
        sub       = df[df['type'] == typ]
        gold_note = 'category→aspect' if typ == 'implicit' else 'aspect'
        print(f"\n{'─'*60}")
        print(f"  2. {typ.upper()} ({len(sub)} rows) — gold={gold_note}")
        print(f"{'─'*60}")
        print(f"  [Explicit-only baselines — gold=aspect]")
        for lbl, col in stages:
            if col not in sub.columns: continue
            res = evaluate_aspect(sub, col, gold_col='aspect',
                                  jaccard_threshold=jaccard_threshold)
            _print_combined_row(lbl, res)
        print(f"\n  [Full system — type-aware gold]")
        res = evaluate_combined(sub, jaccard_threshold)
        _print_combined_row("FT+IG+PARA+KG", res)

    if 'domain' in df.columns:
        domains = sorted(df['domain'].dropna().unique())
        if len(domains) > 1:
            print(f"\n{'─'*60}")
            print(f"  3. PER DOMAIN — type-aware combined")
            print(f"{'─'*60}")
            for dom in domains:
                sub = df[df['domain'] == dom]
                print(f"\n    Domain: {dom}  ({len(sub)} rows)")
                print(f"    [Explicit-only baselines]")
                for lbl, col in stages:
                    if col not in sub.columns: continue
                    res = evaluate_aspect(sub, col, gold_col='aspect',
                                          jaccard_threshold=jaccard_threshold)
                    _print_combined_row(lbl, res, width=22)
                print(f"    [Full system]")
                res = evaluate_combined(sub, jaccard_threshold)
                _print_combined_row("FT+IG+PARA+KG", res, width=22)

# ================================================================
# 10. Save composite CSV
# ================================================================
def save_composite_csv(datasets: Dict[str, pd.DataFrame]) -> None:
    frames = []
    for ds_name, ds_df in datasets.items():
        df = ds_df.copy()
        df['dataset_label']     = ds_name
        df['combined_gold_str'] = df.apply(
            lambda r: str(_get_gold_for_row(r)), axis=1)
        df['combined_pred_str'] = df.apply(
            lambda r: str(_get_combined_pred_for_row(r)), axis=1)
        frames.append(df)

    composite = pd.concat(frames, ignore_index=True)
    list_cols = [
        'aspect','category','kg_preds','kg_final_pairs',
        'cleaned_baseline','cleaned_explicit_iterative',
        'cleaned_explicit_paraphrases','cleaned_combined_aspects',
        'cleaned_combined_aspects_kg',
    ]
    for col in list_cols:
        if col in composite.columns:
            composite[col] = composite[col].apply(
                lambda v: str(v) if isinstance(v, list) else v)

    out_path = "/content/composite_predictions_all_datasets.csv"
    composite.to_csv(out_path, index=False, encoding='utf-8-sig')
    print(f"\n✅ Composite CSV saved: {out_path}")
    print(f"   Rows: {len(composite)}")
    print(f"   Columns: {composite.columns.tolist()}")

    from google.colab import files
    files.download(out_path)

# ================================================================
# 11. Run
# ================================================================
DATASETS = {
    "SemEval14":       test_14,
    "SemEval15":       test_15,
    "SemEval16":       test_16,
    "ACOS_restaurant": test_acos_rest,
    "ACOS_laptop":     test_acos_lap,
    "MABSA_EN":      test_mabsa,
}

print(f"\n{'='*70}")
print(f"APPROACH 1 — SEPARATE TWO-METRIC EVALUATION")
print(f"{'='*70}")
for ds_name, ds_df in DATASETS.items():
    report_approach1(ds_df, dataset_name=ds_name)

# print(f"\n\n{'='*70}")
# print(f"APPROACH 2 — TYPE-AWARE COMBINED EVALUATION")
# print(f"{'='*70}")
# for ds_name, ds_df in DATASETS.items():
#     report_approach2(ds_df, dataset_name=ds_name)

save_composite_csv(DATASETS)
```
